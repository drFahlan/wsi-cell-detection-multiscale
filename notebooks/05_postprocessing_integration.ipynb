```python
import os
import numpy as np
import pandas as pd
from torch.utils.data import Dataset
from PIL import Image
import albumentations as A
from albumentations.pytorch import ToTensorV2
from pathlib import Path

class CellDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform=None):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)


    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        mask_path = self.mask_paths[idx]

        image = np.array(Image.open(self.image_paths[idx]).convert("RGB"))  # [H, W, 3]
        mask = np.load(self.mask_paths[idx])  # [2, H, W]
        mask = np.transpose(mask, (1, 2, 0))        # shape: (H, W, 2)

        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']           # Tensor: [3, H, W]
            mask = augmented['mask']             # Tensor: [H, W, 2]
            mask = mask.permute(2, 0, 1)         # Convert to [2, H, W]

        # Ensure both have the same filename stem
        img_stem = Path(img_path).stem
        mask_stem = Path(mask_path).stem
        assert img_stem == mask_stem, f"Filename mismatch: {img_stem} vs {mask_stem}"

        return image, mask, img_stem

```

```python
import glob
from torch.utils.data import DataLoader

def get_dataloaders(base_img_path, base_mask_path, patch_size, batch_size):
    def get_paths(split):
        img_dir = os.path.join(base_img_path, split, "cell")
        mask_dir = os.path.join(base_mask_path, split, "cell")


        image_paths = sorted(glob.glob(f"{img_dir}/*.jpg"))  # or .png
        mask_paths = []


        for img_path in image_paths:
            filename = os.path.splitext(os.path.basename(img_path))[0]
            mask_paths.append(os.path.join(mask_dir, filename + ".npy"))


        return image_paths, mask_paths

    train_img, train_mask = get_paths("train")
    val_img, val_mask = get_paths("test")

    train_dataset = CellDataset(train_img, train_mask, transform=get_train_transform(patch_size))
    val_dataset = CellDataset(val_img, val_mask, transform=get_val_transform(patch_size))

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2,worker_init_fn=seed_worker, generator=g)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

    return train_loader, val_loader
```

```python
def get_train_transform(patch_size):
    return A.Compose([
        A.RandomCrop(width=patch_size, height=patch_size),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.Normalize(mean=(0.485, 0.456, 0.406),
                    std=(0.229, 0.224, 0.225)),
        ToTensorV2()
    ])

def get_val_transform(patch_size):
    return A.Compose([
        #A.CenterCrop(height=patch_size, width=patch_size),# ini masalahnya
        A.Normalize(mean=(0.485, 0.456, 0.406),
                    std=(0.229, 0.224, 0.225)),
        ToTensorV2()
    ])
```

#model def

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torch.nn.functional as F

class AttentionBlock(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super(AttentionBlock, self).__init__()
        self.W_g = nn.Sequential(
            nn.Conv2d(F_g, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )

        self.W_x = nn.Sequential(
            nn.Conv2d(F_l, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )

        self.psi = nn.Sequential(
            nn.Conv2d(F_int, 1, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(1),
            nn.Sigmoid()
        )

        self.relu = nn.ReLU(inplace=True)

    def forward(self, g, x):
        g1 = self.W_g(g)
        x1 = self.W_x(x)
        psi = self.relu(g1 + x1)
        psi = self.psi(psi)
        return x * psi


class UpConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super(UpConv, self).__init__()
        self.up = nn.ConvTranspose2d(in_ch, out_ch, kernel_size=2, stride=2)

    def forward(self, x):
        x = self.up(x)
        return x


class AttentionUNet(nn.Module):
    def __init__(self, encoder='resnet34', num_classes=2, dropout=0.0, pretrained=True):
        super(AttentionUNet, self).__init__()

        assert encoder == 'resnet34', "Only resnet34 is supported currently"

        resnet = models.resnet34(pretrained=pretrained)
        self.dropout = dropout  # Store for conditional dropout if needed

        # Encoder layers
        self.encoder0 = nn.Sequential(resnet.conv1, resnet.bn1, resnet.relu)  # 64, /2
        self.encoder1 = nn.Sequential(resnet.maxpool, resnet.layer1)          # 64, /4
        self.encoder2 = resnet.layer2                                         # 128, /8
        self.encoder3 = resnet.layer3                                         # 256, /16
        self.encoder4 = resnet.layer4                                         # 512, /32

        # Decoder
        self.up4 = UpConv(512, 256)
        self.att4 = AttentionBlock(F_g=256, F_l=256, F_int=128)
        self.dec4 = self._make_decoder_block(512, 256)

        self.up3 = UpConv(256, 128)
        self.att3 = AttentionBlock(F_g=128, F_l=128, F_int=64)
        self.dec3 = self._make_decoder_block(256, 128)

        self.up2 = UpConv(128, 64)
        self.att2 = AttentionBlock(F_g=64, F_l=64, F_int=32)
        self.dec2 = self._make_decoder_block(128, 64)

        self.up1 = UpConv(64, 64)
        self.att1 = AttentionBlock(F_g=64, F_l=64, F_int=32)
        self.dec1 = self._make_decoder_block(128, 64)

        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def _make_decoder_block(self, in_ch, out_ch):
        layers = [
            nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        ]
        if self.dropout > 0:
            layers.append(nn.Dropout2d(self.dropout))
        return nn.Sequential(*layers)

    def forward(self, x):
        e0 = self.encoder0(x)
        e1 = self.encoder1(e0)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)

        d4 = self.up4(e4)
        e3_att = self.att4(d4, e3)
        d4 = torch.cat((e3_att, d4), dim=1)
        d4 = self.dec4(d4)

        d3 = self.up3(d4)
        e2_att = self.att3(d3, e2)
        d3 = torch.cat((e2_att, d3), dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        e1_att = self.att2(d2, e1)
        d2 = torch.cat((e1_att, d2), dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        e0_att = self.att1(d1, e0)
        d1 = torch.cat((e0_att, d1), dim=1)
        d1 = self.dec1(d1)
        out = self.final_conv(d1)
        out = F.interpolate(out, size=x.shape[2:], mode='bilinear', align_corners=False)  # <--- Add this
        return out

```

```python
import numpy as np
from skimage.draw import disk

def accumulate_weighted_confidences_circular(all_coords, tissue_prob_map, sigma=7, r=11, apply_to_channel_0=True):
    """
    Extracts weighted confidence around each cell coordinate using circular Gaussian-weighted patch.

    Parameters:
        all_coords (list): [class_0_coords, class_1_coords] where each is list of (x, y)
        tissue_prob_map (np.ndarray): 2D float32 array in [0, 1]
        sigma (float): standard deviation of Gaussian
        r (int): radius of circular region
        apply_to_channel_0 (bool): whether to evaluate for channel 0 with 1 - tissue_prob_map

    Returns:
        dict: {'class_0': [...], 'class_1': [...]}
    """
    h, w = tissue_prob_map.shape

    # Precompute a (2r+1)x(2r+1) Gaussian kernel centered at (r, r)
    yx = np.mgrid[-r:r+1, -r:r+1]
    distance_sq = yx[0]**2 + yx[1]**2
    gaussian_kernel = np.exp(-distance_sq / (2 * sigma**2))

    # Create circular mask (same shape as kernel)
    circular_mask = distance_sq <= r**2
    gaussian_kernel *= circular_mask
    gaussian_kernel /= gaussian_kernel.sum()  # normalize inside the disk only

    def extract_confidences(prob_map, coords_list):
        confidences = []
        for x, y in coords_list:
            x = int(round(x))
            y = int(round(y))

            xmin = max(0, x - r)
            xmax = min(w, x + r + 1)
            ymin = max(0, y - r)
            ymax = min(h, y + r + 1)

            patch = prob_map[ymin:ymax, xmin:xmax]

            # Determine the slice of the Gaussian kernel to match patch shape
            k_ymin = r - (y - ymin)
            k_ymax = r + (ymax - y)
            k_xmin = r - (x - xmin)
            k_xmax = r + (xmax - x)

            weights = gaussian_kernel[k_ymin:k_ymax, k_xmin:k_xmax]

            if patch.shape != weights.shape:
                continue  # Safety

            weighted_prob = (patch * weights).sum()
            confidence = weighted_prob  # Already normalized
            confidences.append(confidence)

        return confidences

    output = {}
    if apply_to_channel_0:
        prob_map_ch0 = 1.0 - tissue_prob_map
        output["class_0"] = extract_confidences(prob_map_ch0, all_coords[0])
    else:
        output["class_0"] = []

    output["class_1"] = extract_confidences(tissue_prob_map, all_coords[1])
    return output


def blend_confidences(all_confidences, spatial_confidences, lambda_weight=0.5):
    """
    Blend detection confidence and spatial confidence, with fallback if spatial is missing.

    Parameters:
        all_confidences (dict): {0: [...], 1: [...]} detection confidence
        spatial_confidences (dict): {"class_0": [...], "class_1": [...]} spatial confidence
        lambda_weight (float): blend weight for detection confidence

    Returns:
        dict: {0: [...], 1: [...]} combined confidence per class
    """
    final_confidences = {}

    for c in range(2):
        original = all_confidences[c]
        spatial = spatial_confidences.get(f"class_{c}", [])

        if len(original) != len(spatial) and len(spatial) == 0:
            # Fallback: use original detection confidence only
            final_confidences[c] = original
        else:
            combined = [
                lambda_weight * o + (1 - lambda_weight) * s
                for o, s in zip(original, spatial)
            ]
            final_confidences[c] = combined

    return final_confidences
```

```python
import matplotlib.pyplot as plt
import random
import torchvision.transforms.functional as TF
import torch
import numpy as np
import os
from skimage.feature import peak_local_max

def detect_local_maxima(prob_map, min_distance=6, threshold_abs=0.1):
    peaks = peak_local_max(prob_map, min_distance=min_distance, threshold_abs=threshold_abs)
    coords = [(x, y) for y, x in peaks]  # swap (y, x) to (x, y)
    return coords

def denormalize(tensor, mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)):
    # Tensor shape: (3, H, W)
    mean = torch.tensor(mean).view(3, 1, 1).to(tensor.device)
    std = torch.tensor(std).view(3, 1, 1).to(tensor.device)
    return tensor * std + mean
def extract_coords_from_mask(mask_channel, min_distance=6, threshold_abs=0.1):
    return detect_local_maxima(mask_channel, min_distance=min_distance, threshold_abs=threshold_abs)

def denormalize(tensor, mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)):
    mean = torch.tensor(mean).view(3, 1, 1).to(tensor.device)
    std = torch.tensor(std).view(3, 1, 1).to(tensor.device)
    return tensor * std + mean

def load_gt_coords_from_csv(csv_path, class_channel):
    coords = []
    if not os.path.exists(csv_path):
        return coords
    with open(csv_path, 'r') as f:
        for line in f.readlines():
            parts = line.strip().split(',')
            if len(parts) != 3:
                continue
            x, y, c = int(parts[0]), int(parts[1]), int(parts[2])
            if c == class_channel + 1:  # 1=non-tumor, 2=tumor
                coords.append((x, y))
    return coords

def match_predictions_to_gt(pred_coords, gt_coords, radius=15):
    matched_gt = set()
    tp_coords = []
    fp_coords = []

    for px, py in pred_coords:
        found = False
        min_dist = float('inf')
        nearest_idx = -1

        for idx, (gx, gy) in enumerate(gt_coords):
            if idx in matched_gt:
                continue
            dist = np.sqrt((px - gx) ** 2 + (py - gy) ** 2)
            if dist < radius and dist < min_dist:
                found = True
                min_dist = dist
                nearest_idx = idx

        if found:
            tp_coords.append((px, py))
            matched_gt.add(nearest_idx)
        else:
            fp_coords.append((px, py))

    fn = len(gt_coords) - len(matched_gt)
    return tp_coords, fp_coords, fn

def extract_confidences_from_coords(pred_mask, pred_coords):
    if len(pred_coords) == 0:
        return np.array([])  # Return empty confidence array

    h, w = pred_mask.shape
    coords = np.rint(pred_coords).astype(np.intp)  # round and convert to int

    # Ensure coords is 2D
    coords = coords.reshape(-1, 2)

    coords[:, 0] = np.clip(coords[:, 0], 0, w - 1)  # x
    coords[:, 1] = np.clip(coords[:, 1], 0, h - 1)  # y
    return pred_mask[coords[:, 1], coords[:, 0]]
```

```python
import os
import numpy as np
import torch
from tqdm import tqdm
import matplotlib.pyplot as plt

def load_gt_coords_from_csv(csv_path, class_channel):
    coords = []
    with open(csv_path, 'r') as f:
        for line in f.readlines():
            parts = line.strip().split(',')
            if len(parts) != 3:
                continue
            x, y, c = int(parts[0]), int(parts[1]), int(parts[2])
            if c == class_channel + 1:  # CSV uses 1 for non-tumor, 2 for tumor
                coords.append((x, y))
    return coords

def compute_detection_confusion(pred_coords, gt_coords, radius=15):
    matched_gt = set()
    tp, fp = 0, 0

    for px, py in pred_coords:
        found = False
        min_dist = float('inf')
        nearest_idx = -1

        for idx, (gx, gy) in enumerate(gt_coords):
            if idx in matched_gt:
                continue
            dist = np.sqrt((px - gx) ** 2 + (py - gy) ** 2)
            if dist < radius and dist < min_dist:
                found = True
                min_dist = dist
                nearest_idx = idx

        if found:
            tp += 1
            matched_gt.add(nearest_idx)
        else:
            fp += 1

    fn = len(gt_coords) - len(matched_gt)
    return tp, fp, fn

def filter_overlapping_predictions(all_coords, all_confidences, min_distance=7):
    """
    Filters overlapping predicted coordinates between classes based on confidence.

    Parameters
    ----------
    all_coords : dict[int, List[Tuple[int, int]]]
        Dictionary of predicted coordinates per class.
    all_confidences : dict[int, List[float]]
        Dictionary of confidence scores per class.
    min_distance : int
        Minimum distance threshold to consider two predictions as overlapping.

    Returns
    -------
    filtered_coords : dict[int, List[Tuple[int, int]]]
        Filtered predicted coordinates per class.
    filtered_confidences : dict[int, List[float]]
        Corresponding confidence scores for the filtered coordinates.
    """
    # Collect all predictions with their class and index
    all_predictions = []
    for class_id in [0, 1]:
        for idx, (coord, conf) in enumerate(zip(all_coords[class_id], all_confidences[class_id])):
            all_predictions.append({
                'coord': coord,
                'confidence': conf,
                'class': class_id,
                'original_idx': idx
            })

    # Sort by confidence (highest first)
    all_predictions.sort(key=lambda x: x['confidence'], reverse=True)

    # Keep track of which predictions to keep
    kept_predictions = []

    for current_pred in all_predictions:
        keep_current = True

        # Check against all already kept predictions
        for kept_pred in kept_predictions:
            dist = np.linalg.norm(np.array(current_pred['coord']) - np.array(kept_pred['coord']))

            if dist < min_distance:
                # If they're overlapping, we don't keep the current one
                # (since we're processing in order of decreasing confidence,
                # the kept one has higher or equal confidence)
                keep_current = False
                break

        if keep_current:
            kept_predictions.append(current_pred)

    # Reconstruct the filtered results
    filtered_coords = {0: [], 1: []}
    filtered_confidences = {0: [], 1: []}

    for pred in kept_predictions:
        class_id = pred['class']
        filtered_coords[class_id].append(pred['coord'])
        filtered_confidences[class_id].append(pred['confidence'])

    return filtered_coords, filtered_confidences
```

```python
# Config
patch_size = 512
stride = int(patch_size // 2)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
base_csv_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/val/cell"

train_loader, val_loader = get_dataloaders(
    base_img_path="/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images",
    base_mask_path="/content/drive/MyDrive/research/ocelot/dataset/GT/gaussian masks/type3",
    patch_size=patch_size,
    batch_size=4
)

model = AttentionUNet(encoder='resnet34', num_classes=2, dropout=0.0)
model.to(device)
model.load_state_dict(torch.load("/content/drive/MyDrive/research/ocelot/Method/cell detection/attention UNet-pytorch/model/baseline/best_copper-sweep-1_rank1_epoch176_f1=0.6179.pt"))
model.eval()

# Collect all validation samples
all_samples = []
for images, masks, filenames in val_loader:
    for i in range(len(images)):
        all_samples.append((images[i], masks[i], filenames[i]))

# Randomly sample 3 distinct images every time
sampled = random.sample(all_samples, 3)


for img_tensor, mask_tensor, filename in sampled:
    image = img_tensor.unsqueeze(0).to(device)
    mask = mask_tensor.cpu().numpy()
    filename = filename if isinstance(filename, str) else filename.decode()
    print(f"Image filename: {filename}")

    # Strip extension to match GT CSV
    base_filename = filename.rsplit('.', 1)[0]
    csv_path = f"/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/val/cell/{base_filename}.csv"
    print(f"Expected GT CSV path: {csv_path}")

    # Load GT coordinates
    gt_coords = []
    if os.path.exists(csv_path):
        with open(csv_path, 'r') as f:
            for line in f:
                x, y, cls = map(int, line.strip().split(','))
                gt_coords.append((x, y, cls))
    else:
        print(f"GT CSV not found for {filename}")

    H, W = image.shape[2:]

    probs_full = torch.zeros((2, H, W), dtype=torch.float32).to(device)
    count_map = torch.zeros((1, H, W), dtype=torch.float32).to(device)

    for y in range(0, H, stride):
        for x in range(0, W, stride):
            y1, y2 = y, min(y + patch_size, H)
            x1, x2 = x, min(x + patch_size, W)
            patch = image[:, :, y1:y2, x1:x2]
            out_patch = model(patch)
            prob_patch = torch.sigmoid(out_patch)[0, :, :y2 - y1, :x2 - x1]
            probs_full[:, y1:y2, x1:x2] += prob_patch
            count_map[:, y1:y2, x1:x2] += 1

    probs_full = probs_full / count_map
    probs_np = probs_full.detach().cpu().numpy()
    img_denorm = denormalize(img_tensor.to(device)).cpu().numpy().transpose(1, 2, 0)
    img_np = np.clip(img_denorm, 0, 1)

    fig, axes = plt.subplots(2, 6, figsize=(24, 8))

    axes[0, 0].imshow(img_np)
    axes[0, 0].set_title("Input Image")

    axes[0, 1].imshow(mask[0], cmap='hot', alpha=0.6)
    axes[0, 1].set_title("GT Mask - Non-Tumor")
    axes[1, 1].imshow(mask[1], cmap='hot', alpha=0.6)
    axes[1, 1].set_title("GT Mask - Tumor")

    axes[0, 2].imshow(probs_np[0], cmap='hot')
    axes[0, 2].set_title("Pred Mask - Non-Tumor")
    axes[1, 2].imshow(probs_np[1], cmap='hot')
    axes[1, 2].set_title("Pred Mask - Tumor")

    all_coords = {}
    all_confidences = {}
    for c in range(2):
        # --- Prediction ---
        pred_mask = probs_np[c]
        pred_coords = detect_local_maxima(pred_mask, min_distance=6, threshold_abs=0.2)
        confidences = extract_confidences_from_coords(pred_mask, pred_coords)
        if len(pred_coords) > 0:
            sorted_idx = np.argsort(confidences)[::-1]
            sorted_coords = [pred_coords[i] for i in sorted_idx]
            sorted_confidences = [confidences[i] for i in sorted_idx]
        else:
            sorted_coords = []
            sorted_confidences = []

        all_coords[c] = sorted_coords
        all_confidences[c] = sorted_confidences

    # --- Filter overlapping predictions ---
    filtered_coords, filtered_confidences = filter_overlapping_predictions(all_coords, all_confidences, min_distance=7)

    for c in [0, 1]:
        # --- GT from CSV ---
        gt_coords_csv = load_gt_coords_from_csv(csv_path, c)
        tp_csv, fp_csv, fn_csv = match_predictions_to_gt(filtered_coords[c], gt_coords_csv, radius=15)

        # --- GT from mask ---
        mask_channel = mask_tensor[c].numpy()
        gt_coords_mask = extract_coords_from_mask(mask_channel, min_distance=6, threshold_abs=0.1)
        tp_mask, fp_mask, fn_mask = match_predictions_to_gt(filtered_coords[c], gt_coords_mask, radius=15)

        # --- Plot GT from mask ---
        ax_mask = axes[c, 3]
        canvas_mask = np.ones((H, W, 3))
        ax_mask.imshow(canvas_mask)

        if gt_coords_mask is not None and len(gt_coords_mask) > 0:
            mx, my = zip(*gt_coords_mask)
            ax_mask.scatter(mx, my, c='blue', s=10, label='GT-Mask')

        if tp_mask:
            tx, ty = zip(*tp_mask)
            ax_mask.scatter(tx, ty, c='green', s=10, label='TP')

        if fp_mask:
            fx, fy = zip(*fp_mask)
            ax_mask.scatter(fx, fy, c='red', s=10, label='FP')

        ax_mask.set_title(f"Pred vs GT-Mask (Class {c})")
        ax_mask.axis('off')
        ax_mask.legend(loc='lower left', fontsize=7)

        # --- Plot GT from CSV ---
        ax_csv = axes[c, 4]
        canvas_csv = np.ones((H, W, 3))
        ax_csv.imshow(canvas_csv)

        if gt_coords_csv:
            gx, gy = zip(*gt_coords_csv)
            ax_csv.scatter(gx, gy, c='blue', s=10, label='GT-CSV')

        if tp_csv:
            tx, ty = zip(*tp_csv)
            ax_csv.scatter(tx, ty, c='green', s=10, label='TP')

        if fp_csv:
            fx, fy = zip(*fp_csv)
            ax_csv.scatter(fx, fy, c='red', s=10, label='FP')

        ax_csv.set_title(f"Pred vs GT-CSV (Class {c})")
        ax_csv.axis('off')
        ax_csv.legend(loc='lower left', fontsize=7)

        # --- Compute F1 ---

        tp_csv_n = len(tp_csv)
        fp_csv_n = len(fp_csv)
        fn_csv_n = fn_csv
        tp_mask_n = len(tp_mask)
        fp_mask_n = len(fp_mask)
        fn_mask_n = fn_mask

        f1_csv = 2 * tp_csv_n / (2 * tp_csv_n + fp_csv_n + fn_csv_n + 1e-8) if (tp_csv_n + fp_csv_n + fn_csv_n) > 0 else 1.0
        f1_mask = 2 * tp_mask_n / (2 * tp_mask_n + fp_mask_n + fn_mask_n + 1e-8) if (tp_mask_n + fp_mask_n + fn_mask_n) > 0 else 1.0

        print(f"[{filename}] Class {c} F1 CSV : {f1_csv:.4f} (TP={tp_csv_n}, FP={fp_csv_n}, FN={fn_csv_n})")
        print(f"[{filename}] Class {c} F1 Mask: {f1_mask:.4f} (TP={tp_mask_n}, FP={fp_mask_n}, FN={fn_mask_n})")

                # --- Plot Predicted Coordinates Only ---
        ax_pred = axes[c, 5]
        canvas_pred = np.ones((H, W, 3))
        ax_pred.imshow(canvas_pred)

        if filtered_coords[c]:
            px, py = zip(*filtered_coords[c])
            ax_pred.scatter(px, py, c='orange', s=10, label='Pred')

        ax_pred.set_title(f"Filtered Predictions (Class {c})")
        ax_pred.axis('off')
        ax_pred.legend(loc='lower left', fontsize=7)
```

```python
# Config
patch_size = 512
stride = int(patch_size // 2)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
base_csv_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/test/cell"
base_img_path="/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images"
base_mask_path="/content/drive/MyDrive/research/ocelot/dataset/GT/gaussian masks/type3"
weights_path="/content/drive/MyDrive/research/ocelot/Method/cell detection/attention UNet-pytorch/model/baseline/best_copper-sweep-1_rank1_epoch176_f1=0.6179.pt"
TISSUE_PROB_BASE_PATH = "/content/drive/MyDrive/research/ocelot/dataset/integration/softmax_unet++(se_resnet50_acc0.8915)/test"
radius_thresh = 15

#injection config
sigma=7
r=15
apply_to_channel_0 = True
lambda_weight = 0.5
```

```python
# Initialize counters
global_tp = {0: 0, 1: 0}
global_fp = {0: 0, 1: 0}
global_fn = {0: 0, 1: 0}

train_loader, val_loader = get_dataloaders(
    base_img_path=base_img_path,
    base_mask_path=base_mask_path,
    patch_size=patch_size,
    batch_size=4
)

model = AttentionUNet(encoder='resnet34', num_classes=2, dropout=0.0)
model.to(device)
model.load_state_dict(torch.load("/content/drive/MyDrive/research/ocelot/Method/cell detection/attention UNet-pytorch/model/baseline/best_copper-sweep-1_rank1_epoch176_f1=0.6179.pt"))
model.eval()

with torch.no_grad():
    for images, masks, filenames in tqdm(val_loader):
        for i in range(len(images)):
            image = images[i].unsqueeze(0).to(device)
            filename = filenames[i] if isinstance(filenames[i], str) else filenames[i].decode()
            H, W = image.shape[2:]

            tissue_prob_path = f"{TISSUE_PROB_BASE_PATH}/{filename}.npy"
            tissue_prob_map = np.load(tissue_prob_path).astype(np.float32)

            probs_full = torch.zeros((2, H, W), dtype=torch.float32).to(device)
            count_map = torch.zeros((1, H, W), dtype=torch.float32).to(device)

            # Sliding window inference
            for y in range(0, H, stride):
                for x in range(0, W, stride):
                    y1, y2 = y, min(y + patch_size, H)
                    x1, x2 = x, min(x + patch_size, W)
                    patch = image[:, :, y1:y2, x1:x2]

                    out_patch = model(patch)
                    prob_patch = torch.sigmoid(out_patch)[0, :, :y2 - y1, :x2 - x1]

                    probs_full[:, y1:y2, x1:x2] += prob_patch
                    count_map[:, y1:y2, x1:x2] += 1

            probs_full = probs_full / count_map
            probs_np = probs_full.cpu().numpy()

            all_coords = {}
            all_confidences = {}

            for c in range(2):  # for both classes
                pred_mask = probs_np[c]
                pred_coords = detect_local_maxima(pred_mask, min_distance=15, threshold_abs=0.1)
                confidences = extract_confidences_from_coords(pred_mask, pred_coords)

                if len(pred_coords) > 0:
                    sorted_idx = np.argsort(confidences)[::-1]
                    sorted_coords = [pred_coords[i] for i in sorted_idx]
                    sorted_confidences = [confidences[i] for i in sorted_idx]
                else:
                    sorted_coords = []
                    sorted_confidences = []

                all_coords[c] = sorted_coords
                all_confidences[c] = sorted_confidences

#add output injection here using all_coords
            spatial_confidences = accumulate_weighted_confidences_circular(
                all_coords, tissue_prob_map, sigma=sigma, r=r, apply_to_channel_0=apply_to_channel_0
            )

            combined_confidences = blend_confidences(all_confidences, spatial_confidences, lambda_weight)

            # --- Filter overlapping predictions ---
            filtered_coords, filtered_confidences = filter_overlapping_predictions(all_coords, combined_confidences, min_distance=15)

            # --- Evaluate filtered predictions ---
            for c in [0, 1]:
                csv_path = os.path.join(base_csv_path, f"{filename}.csv")
                if not os.path.exists(csv_path):
                    continue

                gt_coords = load_gt_coords_from_csv(csv_path, class_channel=c)
                tp, fp, fn = compute_detection_confusion(filtered_coords[c], gt_coords, radius=radius_thresh)
                global_tp[c] += tp
                global_fp[c] += fp
                global_fn[c] += fn

# Compute F1 scores
f1_per_class = {}
precision_per_class = {}
recall_per_class = {}


print("\n" + "="*60)
print("EVALUATION RESULTS - 4-Channel Model (RGB + Tissue)")
print("="*60)

for c in [0, 1]:
    tp, fp, fn = global_tp[c], global_fp[c], global_fn[c]


    precision = tp / (tp + fp + 1e-8) if (tp + fp) > 0 else 0.0
    recall = tp / (tp + fn + 1e-8) if (tp + fn) > 0 else 0.0
    f1 = 2 * tp / (2 * tp + fp + fn + 1e-8) if (tp + fp + fn) > 0 else 1.0


    precision_per_class[c] = precision
    recall_per_class[c] = recall
    f1_per_class[c] = f1

    class_name = "Non-Tumor" if c == 0 else "Tumor"
    print(f"\n{class_name} (Class {c}):")
    print(f"  TP: {tp:4d}")
    print(f"  FP: {fp:4d}")
    print(f"  FN: {fn:4d}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1 Score:  {f1:.4f}")

print("Class-wise F1 Scores:")
print(f"Non-Tumor (Class 0): F1 = {f1_per_class[0]:.4f}")
print(f"Tumor     (Class 1): F1 = {f1_per_class[1]:.4f}")

# Overall metrics
overall_f1 = np.mean(list(f1_per_class.values()))
overall_precision = np.mean(list(precision_per_class.values()))
overall_recall = np.mean(list(recall_per_class.values()))

print(f"\nOVERALL METRICS:")
print(f"  Average Precision: {overall_precision:.4f}")
print(f"  Average Recall:    {overall_recall:.4f}")
print(f"  Average F1 Score:  {overall_f1:.4f}")
print("="*60)
```
