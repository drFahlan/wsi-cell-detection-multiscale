### data loader

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import numpy as np
import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2
import os
import glob
import random
import wandb
from PIL import Image
import pandas as pd
import numpy as np
from tqdm import tqdm

```

```python
class CellDetectionPatchDataset(Dataset):
    def __init__(self,
                 image_dir,
                 softmax_dir,
                 mask_dir,
                 csv_dir,
                 config,
                 split='train',
                 num_workers=4):

        self.image_dir = image_dir
        self.softmax_dir = softmax_dir
        self.mask_dir = mask_dir
        self.csv_dir = csv_dir
        self.config = config
        self.split = split
        self.patch_size = config.patch_size

        # Find all complete file triplets
        self.file_triplets = self._find_complete_triplets()
        print(f"Found {len(self.file_triplets)} complete file triplets for {split}")

        # NEW - create the transform methods we actually use
        self.spatial_transforms = self._get_spatial_transforms()
        self.image_post_transforms = self._get_image_post_transforms()
        self.softmax_post_transforms = self._get_softmax_post_transforms()

        #self.softmax_mean, self.softmax_std = self.compute_dataset_softmax_stats()

    def compute_dataset_softmax_stats(self):
        """Compute mean/std for all softmax outputs in the dataset."""
        all_softmax = []

        # Load all softmax files (use first 1000 if dataset is large)
        for triplet in tqdm(self.file_triplets[:1000], desc="Computing softmax stats"):
            softmax = np.load(triplet['softmax'])
            if softmax.ndim == 2:  # Ensure shape (H,W,C)
                softmax = np.expand_dims(softmax, -1)
            all_softmax.append(softmax)

        # Stack and compute stats
        all_softmax = np.concatenate(all_softmax, axis=0)  # (N_samples, H, W, C)
        mean = np.mean(all_softmax, axis=(0,1,2))          # (C,)
        std = np.std(all_softmax, axis=(0,1,2))            # (C,)

        return mean, std

    def _find_complete_triplets(self):
        """Find images that have corresponding softmax and mask files"""
        triplets = []

        # Get all image files
        image_pattern = os.path.join(self.image_dir, self.split, "cell", "*.jpg")
        image_files = glob.glob(image_pattern)

        for img_path in image_files:
            basename = os.path.splitext(os.path.basename(img_path))[0]

            # Check for corresponding softmax file
            softmax_path = os.path.join(self.softmax_dir, self.split, f"{basename}.npy")

            # Check for corresponding mask file
            mask_path = os.path.join(self.mask_dir, self.split, "cell", f"{basename}.npy")

            if os.path.exists(softmax_path) and os.path.exists(mask_path):
                triplets.append({
                    'image': img_path,
                    'softmax': softmax_path,
                    'mask': mask_path,
                    'basename': basename
                })
            else:
                print(f"Skipping {basename}: missing softmax or mask file")

        return triplets

    def _get_spatial_transforms(self):
        """Get spatial transforms that need to be synchronized"""
        if self.split == 'train':
            return A.ReplayCompose([
                A.HorizontalFlip(p=0.5),
                A.VerticalFlip(p=0.5),
                A.RandomRotate90(p=0.5)
            ])
        else:
            return A.ReplayCompose([])  # No spatial transforms for val/test

    def _get_image_post_transforms(self):
        """Get image-specific transforms (color + normalization)"""
        if self.split == 'train':
            return A.Compose([
                A.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.05, p=0.3),
                A.RandomGamma(gamma_limit=(80, 120), p=0.2),
                A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
                ToTensorV2()
            ])
        else:
            return A.Compose([
                A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
                ToTensorV2()
            ])

    def _get_softmax_post_transforms(self):
        """Get softmax-specific transforms (just tensor conversion)"""
        return A.Compose([ToTensorV2()])

    def __len__(self):
        return len(self.file_triplets)

    def __getitem__(self, idx):

        triplet = self.file_triplets[idx]

        try:
            # Load full images
            image = cv2.imread(triplet['image'])
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            softmax = np.load(triplet['softmax'])  # (1024, 1024)
            mask = np.load(triplet['mask'])        # (2, 1024, 1024)

            # no stain normalization

            # === STEP 1: Synchronized Random Crop (only for training) ===
            if self.split == 'train':
                crop_transform = A.ReplayCompose([
                    A.RandomCrop(width=self.patch_size, height=self.patch_size)
                ])

                # Apply crop to image
                crop_result = crop_transform(image=image)
                image = crop_result['image']
                replay_data = crop_result['replay']

                # Replay crop to softmax
                softmax = A.ReplayCompose.replay(replay_data, image=softmax)['image']

                # Replay crop to mask
                mask_for_crop = mask.transpose(1, 2, 0)  # (2, H, W) -> (H, W, 2)
                cropped_mask = A.ReplayCompose.replay(replay_data, image=mask_for_crop)['image']
                mask = cropped_mask.transpose(2, 0, 1)  # Back to (2, H, W)

                # Ensure softmax is 2D
                if softmax.ndim == 3:
                    softmax = softmax.squeeze()

            # Else (val/test): skip cropping, use original full image/softmax/mask

            # STEP 2: Synchronized Spatial Transforms
            spatial_transform = self._get_spatial_transforms()

            # Apply spatial transforms to first component
            spatial_result = spatial_transform(image=image)
            spatially_transformed_image = spatial_result['image']
            spatial_replay = spatial_result['replay']

            # Apply same spatial transforms to other components
            spatially_transformed_softmax = A.ReplayCompose.replay(spatial_replay, image=softmax)['image']

            mask_for_spatial = mask.transpose(1, 2, 0)  # (2, H, W) -> (H, W, 2)
            spatially_transformed_mask = A.ReplayCompose.replay(spatial_replay, image=mask_for_spatial)['image']
            spatially_transformed_mask = spatially_transformed_mask.transpose(2, 0, 1)  # (H, W, 2) -> (2, H, W)
                                                                                                              #print(f"DEBUG - shape after transform: {spatially_transformed_mask.shape}")

            # STEP 3: Component-Specific Transforms
            # Image: Color transforms + normalization + tensor
            image_post_transform = self._get_image_post_transforms()
            image_transformed = image_post_transform(image=spatially_transformed_image)['image']

            # Softmax: normalize + Just tensor conversion
            #spatially_transformed_softmax = (spatially_transformed_softmax - self.softmax_mean) / self.softmax_std

            softmax_post_transform = self._get_softmax_post_transforms()
            softmax_transformed = softmax_post_transform(image=spatially_transformed_softmax)['image']

            # Mask: Manual tensor conversion (no albumentations)
            mask_transformed = torch.from_numpy(spatially_transformed_mask).float()

            # Add channel dimension to softmax: (H, W) -> (1, H, W)
            if len(softmax_transformed.shape) == 2:
                softmax_transformed = softmax_transformed.unsqueeze(0)

            # Concatenate image and softmax: (3, H, W) + (1, H, W) -> (4, H, W)
            combined_input = torch.cat([image_transformed, softmax_transformed], dim=0)
            # Right before return, add these debug prints:
                                                                                                  # print(f"DEBUG - Final combined_input shape: {combined_input.shape}")
                                                                                                  # print(f"DEBUG - Final mask_transformed shape: {mask_transformed.shape}")
                                                                                                  # print(f"DEBUG - combined_input type: {type(combined_input)}")
                                                                                                  # print(f"DEBUG - mask_transformed type: {type(mask_transformed)}")
            #debug_after_transforms(image_transformed, softmax_transformed, mask_transformed, triplet['basename'], debug_chance=0.2)

            return combined_input, mask_transformed, triplet['basename']  # Add filename

        except Exception as e:
            print(f"Error loading patch {idx} from {triplet['basename']}: {str(e)}")
            # NEW - returns 3 values to match expected output
            dummy_input = torch.zeros(4, self.patch_size, self.patch_size)
            dummy_mask = torch.zeros(2, self.patch_size, self.patch_size)
            return dummy_input, dummy_mask, triplet['basename']
```

```python
def create_dataloaders(config, num_workers=2):
    """Create train, val, test dataloaders based on wandb config"""

    # Define paths (update these to your actual paths)
    base_image_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images"
    base_softmax_path = "/content/drive/MyDrive/research/ocelot/dataset/integration/softmax_unet++(se_resnet50_acc0.8915)"
    base_mask_path = "/content/drive/MyDrive/research/ocelot/dataset/GT/gaussian masks/type3"
    base_csv_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations"

    datasets = {}
    dataloaders = {}

    for split in ['train', 'val', 'test']:
        datasets[split] = CellDetectionPatchDataset(
            image_dir=base_image_path,
            softmax_dir=base_softmax_path,
            mask_dir=base_mask_path,
            csv_dir=base_csv_path,
            config=config,
            split=split,
            num_workers=num_workers
        )

        # Different batch size and shuffle for different splits
        shuffle = (split == 'train')

        dataloaders[split] = DataLoader(
            datasets[split],
            batch_size=config.batch_size,
            shuffle=shuffle,
            num_workers=num_workers,
            pin_memory=True,
            drop_last=(split == 'train')  # Drop last incomplete batch for training
        )

    return dataloaders, datasets
```

### model definition

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torch.nn.functional as F

class AttentionBlock(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super(AttentionBlock, self).__init__()
        self.W_g = nn.Sequential(
            nn.Conv2d(F_g, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )

        self.W_x = nn.Sequential(
            nn.Conv2d(F_l, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )

        self.psi = nn.Sequential(
            nn.Conv2d(F_int, 1, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(1),
            nn.Sigmoid()
        )

        self.relu = nn.ReLU(inplace=True)

    def forward(self, g, x):
        g1 = self.W_g(g)
        x1 = self.W_x(x)
        psi = self.relu(g1 + x1)
        psi = self.psi(psi)
        return x * psi


class UpConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super(UpConv, self).__init__()
        self.up = nn.ConvTranspose2d(in_ch, out_ch, kernel_size=2, stride=2)

    def forward(self, x):
        x = self.up(x)
        return x


class AttentionUNet(nn.Module):
    def __init__(self, encoder='resnet34', num_classes=2, dropout=0.0, pretrained=True, in_channels=4):
        super(AttentionUNet, self).__init__()

        assert encoder == 'resnet34', "Only resnet34 is supported currently"

        resnet = models.resnet34(pretrained=pretrained)
        self.dropout = dropout  # Store for conditional dropout if needed

        # Modify first conv layer for 4-channel input (RGB + tissue)
        if in_channels != 3:
            # Create new conv1 layer with 4 input channels
            original_conv1 = resnet.conv1
            new_conv1 = nn.Conv2d(
                in_channels=in_channels,
                out_channels=original_conv1.out_channels,
                kernel_size=original_conv1.kernel_size,
                stride=original_conv1.stride,
                padding=original_conv1.padding,
                bias=original_conv1.bias
            )

            # Initialize new conv1 weights
            with torch.no_grad():
                if pretrained:
                    # Copy RGB weights and initialize tissue channel
                    new_conv1.weight[:, :3, :, :] = original_conv1.weight
                    # Initialize tissue channel (4th channel) with average of RGB channels
                    new_conv1.weight[:, 3:, :, :] = original_conv1.weight.mean(dim=1, keepdim=True)
                else:
                    # Random initialization for all channels
                    nn.init.kaiming_normal_(new_conv1.weight, mode='fan_out', nonlinearity='relu')

            resnet.conv1 = new_conv1

        # Encoder layers (identical to original)
        self.encoder0 = nn.Sequential(resnet.conv1, resnet.bn1, resnet.relu)  # 64, /2
        self.encoder1 = nn.Sequential(resnet.maxpool, resnet.layer1)          # 64, /4
        self.encoder2 = resnet.layer2                                         # 128, /8
        self.encoder3 = resnet.layer3                                         # 256, /16
        self.encoder4 = resnet.layer4                                         # 512, /32

        # Decoder (identical to original)
        self.up4 = UpConv(512, 256)
        self.att4 = AttentionBlock(F_g=256, F_l=256, F_int=128)
        self.dec4 = self._make_decoder_block(512, 256)

        self.up3 = UpConv(256, 128)
        self.att3 = AttentionBlock(F_g=128, F_l=128, F_int=64)
        self.dec3 = self._make_decoder_block(256, 128)

        self.up2 = UpConv(128, 64)
        self.att2 = AttentionBlock(F_g=64, F_l=64, F_int=32)
        self.dec2 = self._make_decoder_block(128, 64)

        self.up1 = UpConv(64, 64)
        self.att1 = AttentionBlock(F_g=64, F_l=64, F_int=32)
        self.dec1 = self._make_decoder_block(128, 64)

        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def _make_decoder_block(self, in_ch, out_ch):
        layers = [
            nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        ]
        if self.dropout > 0:
            layers.append(nn.Dropout2d(self.dropout))
        return nn.Sequential(*layers)

    def forward(self, x):
        e0 = self.encoder0(x)
        e1 = self.encoder1(e0)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)

        d4 = self.up4(e4)
        e3_att = self.att4(d4, e3)
        d4 = torch.cat((e3_att, d4), dim=1)
        d4 = self.dec4(d4)

        d3 = self.up3(d4)
        e2_att = self.att3(d3, e2)
        d3 = torch.cat((e2_att, d3), dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        e1_att = self.att2(d2, e1)
        d2 = torch.cat((e1_att, d2), dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        e0_att = self.att1(d1, e0)
        d1 = torch.cat((e0_att, d1), dim=1)
        d1 = self.dec1(d1)

        out = self.final_conv(d1)
        out = F.interpolate(out, size=x.shape[2:], mode='bilinear', align_corners=False)
        return out
```

```python
def create_model(config):
    """
    Create model based on wandb config
    """
    model = AttentionUNet(
        encoder='resnet34',
        num_classes=2,                    # Binary cell detection (background + cell)
        dropout=config.dropout,           # From wandb config
        pretrained=True,                  # Use ImageNet pretrained weights
        in_channels=4                     # RGB (3) + tissue softmax (1)
    )

    print(f"âœ… Model created with:")
    print(f"   - Input channels: 4 (RGB + tissue)")
    print(f"   - Output classes: 2")
    print(f"   - Dropout rate: {config.dropout}")
    print(f"   - Pretrained encoder: True")

    return model
```

### loss function and optimizers

```python
class FocalLoss(nn.Module):
    def __init__(self, alpha=1.0, gamma=2.0, reduction='mean'):
        """
        Focal Loss for addressing class imbalance

        Args:
            alpha (float): Weighting factor for rare class (cells vs background)
            gamma (float): Focusing parameter (higher = more focus on hard examples)
            reduction (str): 'mean', 'sum', or 'none'
        """
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        """
        Args:
            inputs: (N, C, H, W) - Raw logits from model
            targets: (N, C, H, W) - Target probability maps (0-1)
        """
        # Convert logits to probabilities
        probs = torch.sigmoid(inputs)

        # Probability-weighted focal loss
        # Weight positive contribution by target probability
        pos_loss = -self.alpha * (1 - probs) ** self.gamma * targets * torch.log(probs + 1e-8)

        # Weight negative contribution by (1 - target probability)
        neg_loss = -(1 - self.alpha) * probs ** self.gamma * (1 - targets) * torch.log(1 - probs + 1e-8)

        # Combine losses
        focal_loss = pos_loss + neg_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss
```

```python
class BCETverskyLoss(nn.Module):
    def __init__(self, alpha=0.7, bce_weight=0.5):
        """
        Combined BCE and Tversky Loss

        Args:
            alpha (float): Weight for false negatives in Tversky (higher = penalize missed cells more)
            beta (float): Weight for false positives in Tversky (higher = penalize false detections more)
            bce_weight (float): Weight for BCE component (0.5 = equal weighting)
        """
        super(BCETverskyLoss, self).__init__()
        self.alpha = alpha
        self.beta = 1.0-alpha
        self.bce_weight = bce_weight
        self.tversky_weight = 1.0 - bce_weight

    def forward(self, inputs, targets):
        """
        Args:
            inputs: (N, C, H, W) - Raw logits from model
            targets: (N, C, H, W) - Target probability maps (0-1)
        """
        # BCE component
        bce_loss = F.binary_cross_entropy_with_logits(inputs, targets, reduction='mean')

        # Convert logits to probabilities for Tversky
        probs = torch.sigmoid(inputs)

        # Calculate Tversky for each sample and class
        tversky_loss = 0
        batch_size, num_classes = inputs.shape[0], inputs.shape[1]

        for i in range(batch_size):
            for c in range(num_classes):
                pred_flat = probs[i, c].flatten()
                target_flat = targets[i, c].flatten()

                # Tversky components
                tp = (pred_flat * target_flat).sum()
                fp = (pred_flat * (1 - target_flat)).sum()
                fn = ((1 - pred_flat) * target_flat).sum()

                # Tversky index
                tversky_index = tp / (tp + self.alpha * fn + self.beta * fp + 1e-8)
                tversky_loss += (1 - tversky_index)

        # Average over batch and classes
        tversky_loss = tversky_loss / (batch_size * num_classes)

        # Combine losses
        total_loss = self.bce_weight * bce_loss + self.tversky_weight * tversky_loss

        return total_loss
```

```python
def create_loss_function(config):
    """
    Create loss function based on wandb config

    Args:
        config: wandb config object with loss function parameters

    Returns:
        loss_fn: Configured loss function
    """

    if config.loss_function == 'focal':
        loss_fn = FocalLoss(
            alpha=config.focal_alpha,
            gamma=config.focal_gamma,
            reduction='mean'
        )

        print(f"âœ… Focal Loss created:")
        print(f"   - Alpha (class weight): {config.focal_alpha}")
        print(f"   - Gamma (focus): {config.focal_gamma}")

    elif config.loss_function == 'bce_tversky':
        loss_fn = BCETverskyLoss(
            alpha=0.7,  # Default to your proven value
            bce_weight=config.get('bce_weight', 0.5)  # Default to equal weighting
        )

        print(f"âœ… BCE+Tversky Loss created:")
        print(f"   - Tversky Alpha (FN weight): {0.7}")
        print(f"   - Tversky Beta (FP weight): {0.3}")
        print(f"   - BCE Weight: {0.5}")

    else:
        raise ValueError(f"Unsupported loss function: {config.loss_function}")

    return loss_fn
```

```python
def create_optimizer(model, config):
    """
    Create optimizer based on wandb config

    Args:
        model: PyTorch model
        config: wandb config object with optimizer parameters

    Returns:
        optimizer: Configured optimizer
    """

    if config.optimizer == 'adam':
        optimizer = optim.Adam(
            model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay,
            betas=(0.9, 0.999),  # Standard Adam betas
            eps=1e-8             # Numerical stability
        )

        print(f"âœ… Adam Optimizer created:")
        print(f"   - Learning rate: {config.learning_rate}")
        print(f"   - Weight decay: {config.weight_decay}")
        print(f"   - Betas: (0.9, 0.999)")

    elif config.optimizer == 'adamw':
        optimizer = optim.AdamW(
            model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay,
            betas=(0.9, 0.999),  # Standard AdamW betas
            eps=1e-8             # Numerical stability
        )

        print(f"âœ… AdamW Optimizer created:")
        print(f"   - Learning rate: {config.learning_rate}")
        print(f"   - Weight decay: {config.weight_decay}")
        print(f"   - Betas: (0.9, 0.999)")

    else:
        raise ValueError(f"Unsupported optimizer: {config.optimizer}")

    return optimizer
```

```python
import torch.optim as optim

def create_scheduler(optimizer, config):
    """
    Create learning rate scheduler based on wandb config

    Args:
        optimizer: PyTorch optimizer
        config: wandb config object with scheduler parameters

    Returns:
        scheduler: Configured scheduler or None
    """

    if not hasattr(config, 'scheduler') or config.scheduler == 'none':
        print("âœ… No learning rate scheduler")
        return None

    elif config.scheduler == 'plateau':
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer,
            mode='min',           # For metrics like F1 score
            factor=0.5,           # Reduce by 50%
            patience=40,          # Wait 10 epochs
            verbose=True
        )
        print(f"âœ… Plateau scheduler created (factor=0.5, patience=40)")

    else:
        raise ValueError(f"Unsupported scheduler: {config.scheduler}")

    return scheduler
```

### train and validate

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn.functional as F
from tqdm import tqdm
import wandb
from sklearn.metrics import precision_score, recall_score, f1_score
from scipy import ndimage
from skimage.feature import peak_local_max
from scipy.spatial import cKDTree


def detect_local_maxima(prob_map, min_distance=15, threshold_abs=0.1):
    peaks = peak_local_max(prob_map, min_distance=min_distance, threshold_abs=threshold_abs)
    coords = [(x, y) for y, x in peaks]  # swap (y, x) to (x, y)
    return coords

def compute_detection_confusion(pred_coords, gt_coords, radius=15):
    matched_gt = set()
    tp, fp = 0, 0

    for px, py in pred_coords:
        found = False
        min_dist = float('inf')
        nearest_idx = -1

        for idx, (gx, gy) in enumerate(gt_coords):
            if idx in matched_gt:
                continue
            dist = np.sqrt((px - gx) ** 2 + (py - gy) ** 2)
            if dist < radius and dist < min_dist:
                found = True
                min_dist = dist
                nearest_idx = idx

        if found:
            tp += 1
            matched_gt.add(nearest_idx)
        else:
            fp += 1

    fn = len(gt_coords) - len(matched_gt)
    return tp, fp, fn

def extract_confidences_from_coords(pred_mask, pred_coords):
    if len(pred_coords) == 0:
        return np.array([])  # Return empty confidence array

    h, w = pred_mask.shape
    coords = np.rint(pred_coords).astype(np.intp)  # round and convert to int

    # Ensure coords is 2D
    coords = coords.reshape(-1, 2)

    coords[:, 0] = np.clip(coords[:, 0], 0, w - 1)  # x
    coords[:, 1] = np.clip(coords[:, 1], 0, h - 1)  # y
    return pred_mask[coords[:, 1], coords[:, 0]]


def load_gt_coords_from_csv(csv_path, class_channel):
    coords = []
    if not os.path.exists(csv_path) or os.path.getsize(csv_path) == 0:
        return []
    with open(csv_path, 'r') as f:
        for line in f.readlines():
            line = line.strip()  # âœ… Remove leading/trailing whitespace
            if not line:
              continue

            parts = [p.strip() for p in line.split(',')]  # âœ… Remove extra spaces around each value
            if len(parts) != 3:
                continue
            # x, y, c = int(parts[0]), int(parts[1]), int(parts[2])
            try:
                x = int(parts[0])
                y = int(parts[1])
                c = int(parts[2])
            except ValueError:
                continue  # Skip lines with non-integer values

            if c == class_channel + 1:  # 1=non-tumor, 2=tumor
                coords.append((x, y))
    return coords

def filter_overlapping_predictions(all_coords, all_confidences, min_distance=7):
    """
    Filters overlapping predicted coordinates between classes based on confidence.

    Parameters
    ----------
    all_coords : dict[int, List[Tuple[int, int]]]
        Dictionary of predicted coordinates per class.
    all_confidences : dict[int, List[float]]
        Dictionary of confidence scores per class.
    min_distance : int
        Minimum distance threshold to consider two predictions as overlapping.

    Returns
    -------
    filtered_coords : dict[int, List[Tuple[int, int]]]
        Filtered predicted coordinates per class.
    filtered_confidences : dict[int, List[float]]
        Corresponding confidence scores for the filtered coordinates.
    """
    # Collect all predictions with their class and index
    all_predictions = []
    for class_id in [0, 1]:
        for idx, (coord, conf) in enumerate(zip(all_coords[class_id], all_confidences[class_id])):
            all_predictions.append({
                'coord': coord,
                'confidence': conf,
                'class': class_id,
                'original_idx': idx
            })

    # Sort by confidence (highest first)
    all_predictions.sort(key=lambda x: x['confidence'], reverse=True)

    # Keep track of which predictions to keep
    kept_predictions = []

    for current_pred in all_predictions:
        keep_current = True

        # Check against all already kept predictions
        for kept_pred in kept_predictions:
            dist = np.linalg.norm(np.array(current_pred['coord']) - np.array(kept_pred['coord']))

            if dist < min_distance:
                # If they're overlapping, we don't keep the current one
                # (since we're processing in order of decreasing confidence,
                # the kept one has higher or equal confidence)
                keep_current = False
                break

        if keep_current:
            kept_predictions.append(current_pred)

    # Reconstruct the filtered results
    filtered_coords = {0: [], 1: []}
    filtered_confidences = {0: [], 1: []}

    for pred in kept_predictions:
        class_id = pred['class']
        filtered_coords[class_id].append(pred['coord'])
        filtered_confidences[class_id].append(pred['confidence'])

    return filtered_coords, filtered_confidences
```

```python
import torch
import torch.optim as optim
import numpy as np
import os
from tqdm import tqdm
import wandb

def train_model(model, train_loader, val_loader, base_csv_path, optimizer, scheduler, loss_fn, device, config):
    """
    Training function for cell detection with 4-channel input (RGB + tissue)

    Args:
        model: Attention U-Net model
        train_loader: DataLoader for training (returns combined_inputs, masks, filenames)
        val_loader: DataLoader for validation (returns combined_inputs, masks, filenames)
        base_csv_path: Path to CSV annotations for validation
        optimizer: Optimizer (Adam/AdamW)
        scheduler: Learning rate scheduler (cosine/step/plateau/none)
        loss_fn: Loss function (focal/bce_tversky)
        device: CUDA device
        config: Wandb config with hyperparameters
    """
    model.to(device)
    best_f1 = 0

    patch_size = config.patch_size
    stride = int(patch_size/2)  # Can set smaller for overlap

    best_models = []  # Will store tuples: (val_f1, epoch, model_state_dict)
    TOP_K = 3

    low_f1_patience = 10
    low_f1_threshold = 0.25
    low_f1_counter = 0

    run_name = wandb.run.name  # e.g., "experiment-42"

    for epoch in range(config.epochs):
        model.train()
        train_loss = 0.0

        # Training loop
        for combined_inputs, masks, _ in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
            combined_inputs, masks = combined_inputs.to(device), masks.to(device)
            optimizer.zero_grad()
            outputs = model(combined_inputs)
            loss = loss_fn(outputs, masks)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        avg_train_loss = train_loss / len(train_loader)

        ##################################################################################
        # Validation
        ##################################################################################

        model.eval()
        val_loss = 0.0
        all_f1 = []

        global_tp = {0: 0, 1: 0}
        global_fp = {0: 0, 1: 0}
        global_fn = {0: 0, 1: 0}

        ENABLE_FILTERING = False

        with torch.no_grad():
            for combined_inputs, masks, filenames in tqdm(val_loader, desc="Validation", leave=False):
                for i in range(len(combined_inputs)):
                    combined_input = combined_inputs[i].to(device).unsqueeze(0)  # 4-channel input
                    mask = masks[i].to(device).unsqueeze(0)
                    filename = filenames[i]

                    H, W = combined_input.shape[2:]
                    probs_full = torch.zeros((2, H, W), dtype=torch.float32).to(device)
                    count_map = torch.zeros((1, H, W), dtype=torch.float32).to(device)

                    # Sliding window inference
                    for y in range(0, H, stride):
                        for x in range(0, W, stride):
                            y1 = y
                            y2 = min(y + patch_size, H)
                            x1 = x
                            x2 = min(x + patch_size, W)

                            # Extract patch from 4-channel input
                            input_patch = combined_input[:, :, y1:y2, x1:x2]

                            out_patch = model(input_patch)
                            prob_patch = torch.sigmoid(out_patch)[0, :, :y2 - y1, :x2 - x1]

                            probs_full[:, y1:y2, x1:x2] += prob_patch
                            count_map[:, y1:y2, x1:x2] += 1

                    probs_full = probs_full / count_map
                    loss = loss_fn(probs_full.unsqueeze(0), mask)
                    val_loss += loss.item()

                    probs_np = probs_full.cpu().numpy()

                    # Extract predictions for both classes
                    all_coords = {}
                    all_confidences = {}

                    for c in range(2):
                        pred_mask = probs_np[c]
                        pred_coords = detect_local_maxima(pred_mask, min_distance=config.min_distance, threshold_abs=0.2) #atur
                        confidences = extract_confidences_from_coords(pred_mask, pred_coords)

                        all_coords[c] = pred_coords
                        all_confidences[c] = confidences

                    if ENABLE_FILTERING:
                        # Apply overlap filtering across classes
                        filtered_coords, filtered_confidences = filter_overlapping_predictions(
                            all_coords, all_confidences, min_distance=config.min_distance
                        )
                    else:
                        filtered_coords, filtered_confidences = all_coords, all_confidences

                    csv_path = os.path.join(base_csv_path, f"{filename}.csv")
                    if not os.path.exists(csv_path):
                        continue

                    for c in range(2):
                        gt_coords = load_gt_coords_from_csv(csv_path, class_channel=c)

                        # Use filtered predictions
                        pred_coords_filtered = filtered_coords[c]
                        confidences_filtered = filtered_confidences[c]

                        # Sort by confidence (highest first)
                        if len(confidences_filtered) > 0:
                            sorted_idx = np.argsort(confidences_filtered)[::-1]
                            sorted_coords = [pred_coords_filtered[i] for i in sorted_idx]
                        else:
                            sorted_coords = []

                        tp, fp, fn = compute_detection_confusion(sorted_coords, gt_coords)
                        global_tp[c] += tp
                        global_fp[c] += fp
                        global_fn[c] += fn

        f1_per_class = {}
        precision_per_class = {}
        recall_per_class = {}

        for c in [0, 1]:
            tp = global_tp[c]
            fp = global_fp[c]
            fn = global_fn[c]

            precision = tp / (tp + fp + 1e-8) if (tp + fp) > 0 else 1.0
            recall = tp / (tp + fn + 1e-8) if (tp + fn) > 0 else 1.0
            f1 = 2 * precision * recall / (precision + recall + 1e-8) if (precision + recall) > 0 else 1.0


            precision_per_class[c] = precision
            recall_per_class[c] = recall
            f1_per_class[c] = f1

        avg_val_loss = val_loss / len(val_loader)
        avg_val_f1 = np.mean(list(f1_per_class.values()))

        # Check if avg_val_f1 is below the threshold
        if avg_val_f1 < low_f1_threshold:
            low_f1_counter += 1
            print(f"Epoch {epoch}: val_f1={avg_val_f1:.4f} < {low_f1_threshold}, counter={low_f1_counter}")
        else:
            low_f1_counter = 0  # reset if above threshold

        # Trigger early stopping
        if low_f1_counter >= low_f1_patience:
            print(f"Early stopping: val_f1 stayed below {low_f1_threshold} for {low_f1_patience} epochs.")
            break

        # Step scheduler (only if plateau)
        if scheduler and config.scheduler == 'plateau':
            scheduler.step(avg_val_loss)  # Use validation loss

# lagi diganti ke recall 1

        # Keep track of top K models
        if len(best_models) < TOP_K or recall_per_class[1] > min(m[0] for m in best_models):
            model_cpu_state = {k: v.cpu() for k, v in model.state_dict().items()}
            best_models.append((recall_per_class[1], epoch, model_cpu_state))
            best_models = sorted(best_models, key=lambda x: x[0], reverse=True)[:TOP_K]

        # Get current learning rate for logging
        current_lr = optimizer.param_groups[0]['lr']

        wandb.log({
            "epoch": epoch,
            "train_loss": avg_train_loss,
            "val_loss": avg_val_loss,
            "val_detection_f1": avg_val_f1,
            "learning_rate": current_lr,
            "val_f1_class_0": f1_per_class[0],
            "val_f1_class_1": f1_per_class[1],
            "val_precision_class_0": precision_per_class[0],
            "val_precision_class_1": precision_per_class[1],
            "val_recall_class_0": recall_per_class[0],
            "val_recall_class_1": recall_per_class[1]
        })

        print(f"Epoch {epoch+1}/{config.epochs}")
        print(f"Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}")
        print(f"Val F1: {avg_val_f1:.4f} (Class 0: {f1_per_class[0]:.4f}, Class 1: {f1_per_class[1]:.4f})")
        print(f"Learning Rate: {current_lr:.6f}")
        print("-" * 50)
#lagi diganti ke recall 1
    # Save best models
    for rank, (metric, epoch, state_dict) in enumerate(best_models, start=1):
        checkpoint_path = f"best_{run_name}_rank{rank}_epoch{epoch+1}_recallTC{metric:.4f}.pt"
        torch.save(state_dict, checkpoint_path)

        artifact = wandb.Artifact(f"model_rank{rank}", type="model")
        artifact.add_file(checkpoint_path)
        wandb.log_artifact(artifact)

    print(f"\nðŸŽ‰ Training completed!")
    print(f"Best validation F1: {best_models[0][0]:.4f} at epoch {best_models[0][1]+1}")
    return best_models[0][2]  # Return best model state dict
```

```python
import matplotlib.pyplot as plt
import random
import torchvision.transforms.functional as TF
import torch
import numpy as np
import os
from skimage.feature import peak_local_max

def detect_local_maxima(prob_map, min_distance=6, threshold_abs=0.1):
    peaks = peak_local_max(prob_map, min_distance=min_distance, threshold_abs=threshold_abs)
    coords = [(x, y) for y, x in peaks]  # swap (y, x) to (x, y)
    return coords

def denormalize(tensor, mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)):
    # Tensor shape: (3, H, W)
    mean = torch.tensor(mean).view(3, 1, 1).to(tensor.device)
    std = torch.tensor(std).view(3, 1, 1).to(tensor.device)
    return tensor * std + mean
def extract_coords_from_mask(mask_channel, min_distance=6, threshold_abs=0.1):
    return detect_local_maxima(mask_channel, min_distance=min_distance, threshold_abs=threshold_abs)

def denormalize(tensor, mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)):
    mean = torch.tensor(mean).view(3, 1, 1).to(tensor.device)
    std = torch.tensor(std).view(3, 1, 1).to(tensor.device)
    return tensor * std + mean

def load_gt_coords_from_csv(csv_path, class_channel):
    coords = []
    if not os.path.exists(csv_path):
        return coords
    with open(csv_path, 'r') as f:
        for line in f.readlines():
            parts = line.strip().split(',')
            if len(parts) != 3:
                continue
            x, y, c = int(parts[0]), int(parts[1]), int(parts[2])
            if c == class_channel + 1:  # 1=non-tumor, 2=tumor
                coords.append((x, y))
    return coords

def match_predictions_to_gt(pred_coords, gt_coords, radius=15):
    matched_gt = set()
    tp_coords = []
    fp_coords = []

    for px, py in pred_coords:
        found = False
        min_dist = float('inf')
        nearest_idx = -1

        for idx, (gx, gy) in enumerate(gt_coords):
            if idx in matched_gt:
                continue
            dist = np.sqrt((px - gx) ** 2 + (py - gy) ** 2)
            if dist < radius and dist < min_dist:
                found = True
                min_dist = dist
                nearest_idx = idx

        if found:
            tp_coords.append((px, py))
            matched_gt.add(nearest_idx)
        else:
            fp_coords.append((px, py))

    fn = len(gt_coords) - len(matched_gt)
    return tp_coords, fp_coords, fn
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import random
import os
import cv2
import glob
from types import SimpleNamespace

def load_full_scale_sample(image_dir, softmax_dir, mask_dir, split='val'):
    """
    Load full-scale 1024x1024 images (not patches)

    Returns list of (combined_input, mask, filename) for full images
    """
    samples = []

    # Find all complete triplets
    image_pattern = os.path.join(image_dir, split, "cell", "*.jpg")
    image_files = glob.glob(image_pattern)

    for img_path in image_files:
        basename = os.path.splitext(os.path.basename(img_path))[0]


        # Check for corresponding files
        softmax_path = os.path.join(softmax_dir, split, f"{basename}.npy")
        mask_path = os.path.join(mask_dir, split, "cell", f"{basename}.npy")

        if os.path.exists(softmax_path) and os.path.exists(mask_path):
            try:
                # Load full images (1024x1024)
                image = cv2.imread(img_path)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                softmax = np.load(softmax_path)  # (1024, 1024)
                mask = np.load(mask_path)        # (2, 1024, 1024)

                # no stain normalization

                # Ensure softmax is 2D
                if len(softmax.shape) == 3:
                    softmax = softmax.squeeze()

                # Convert to tensors and normalize
                # Normalize image (ImageNet normalization)
                # NEW - ensures float32
                image_norm = image.astype(np.float32) / 255.0
                mean = np.array([0.485, 0.456, 0.406], dtype=np.float32)  # â† Add dtype
                std = np.array([0.229, 0.224, 0.225], dtype=np.float32)   # â† Add dtype
                image_norm = (image_norm - mean) / std
                image_tensor = torch.from_numpy(image_norm).float().permute(2, 0, 1)  # â† Add .float()

                # Convert softmax to tensor
                softmax_tensor = torch.from_numpy(softmax.astype(np.float32)).float().unsqueeze(0)  # â† Add .astype()

                # Combine RGB + tissue
                combined_input = torch.cat([image_tensor, softmax_tensor], dim=0)  # (4, 1024, 1024)

                # Convert mask to tensor
                mask_tensor = torch.from_numpy(mask).float()  # (2, 1024, 1024)

                samples.append((combined_input, mask_tensor, basename))

            except Exception as e:
                print(f"Error loading {basename}: {e}")
                continue

    return samples
# Create gaussian weight map for blending - match actual patch size
def create_gaussian_weight(height, width):
    """Create 2D gaussian weight map for actual patch dimensions"""
    center_y, center_x = height // 2, width // 2
    y, x = np.ogrid[:height, :width]
    sigma_y, sigma_x = height / 6, width / 6
    weight = np.exp(-((x - center_x)**2 / (2 * sigma_x**2) + (y - center_y)**2 / (2 * sigma_y**2)))
    return torch.from_numpy(weight).float()
```

```python
import torch
import numpy as np
from tqdm import tqdm
import os
import cv2
import glob
from types import SimpleNamespace

# =============================================================================
# EVALUATION FOR 4-CHANNEL APPROACH
# =============================================================================

def evaluate_4channel_model():
    """
    Evaluate 4-channel model (RGB + tissue) on validation set
    """

    # Config
    patch_size = 512  # Your model's patch size
    stride = int(patch_size // 2)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    base_csv_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/test/cell"
    radius_thresh = 15

    # Data paths
    image_dir = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images"
    softmax_dir = "/content/drive/MyDrive/research/ocelot/dataset/integration/softmax_unet++(se_resnet50_acc0.8915)"
    mask_dir = "/content/drive/MyDrive/research/ocelot/dataset/GT/gaussian masks/type3"

    # Initialize counters
    global_tp = {0: 0, 1: 0}
    global_fp = {0: 0, 1: 0}
    global_fn = {0: 0, 1: 0}

    # Load model (4-channel input)
    config = SimpleNamespace(dropout=0.0)  # Your model's dropout
    model = create_model(config)  # 4-channel model
    model.load_state_dict(torch.load("/content/drive/MyDrive/research/ocelot/Method/cell detection/attention UNet-pytorch/model/enhanced/input injection/best_radiant-sweep-1_rank1_epoch185_f1=0.6919.pt"))
    model.to(device)
    model.eval()

    # Load full-scale validation samples
    val_samples = load_full_scale_sample(image_dir, softmax_dir, mask_dir, split='test')
    print(f"Loaded {len(val_samples)} validation samples")

    with torch.no_grad():
        for sample_idx, (combined_input_tensor, mask_tensor, filename) in enumerate(tqdm(val_samples, desc="Evaluating")):

            # Prepare 4-channel input
            combined_input = combined_input_tensor.unsqueeze(0).to(device)  # (1, 4, 1024, 1024)
            H, W = combined_input.shape[2:]

            print(f"Processing {filename} ({H}x{W})")

            probs_full = torch.zeros((2, H, W), dtype=torch.float32).to(device)
            count_map = torch.zeros((1, H, W), dtype=torch.float32).to(device)

            # Sliding window inference with 4-channel input
            for y in range(0, H, stride):
                for x in range(0, W, stride):
                    y1, y2 = y, min(y + patch_size, H)
                    x1, x2 = x, min(x + patch_size, W)

                    # Extract 4-channel patch
                    patch = combined_input[:, :, y1:y2, x1:x2]

                    # Model inference
                    out_patch = model(patch)
                    prob_patch = torch.sigmoid(out_patch)[0, :, :y2 - y1, :x2 - x1]

                    # Accumulate predictions
                    probs_full[:, y1:y2, x1:x2] += prob_patch
                    count_map[:, y1:y2, x1:x2] += 1

            # Average overlapping predictions
            probs_full = probs_full / count_map
            probs_np = probs_full.cpu().numpy()

            # Extract predictions for both classes
            all_coords = {}
            all_confidences = {}

            for c in range(2):  # for both classes
                pred_mask = probs_np[c]
                pred_coords = detect_local_maxima(pred_mask, min_distance=15, threshold_abs=0.1)
                confidences = extract_confidences_from_coords(pred_mask, pred_coords)

                if len(pred_coords) > 0:
                    sorted_idx = np.argsort(confidences)[::-1]
                    sorted_coords = [pred_coords[i] for i in sorted_idx]
                    sorted_confidences = [confidences[i] for i in sorted_idx]
                else:
                    sorted_coords = []
                    sorted_confidences = []

                all_coords[c] = sorted_coords
                all_confidences[c] = sorted_confidences

            # Filter overlapping predictions
            filtered_coords, filtered_confidences = filter_overlapping_predictions(
                all_coords, all_confidences, min_distance=15
            )

            # Evaluate filtered predictions
            for c in [0, 1]:
                csv_path = os.path.join(base_csv_path, f"{filename}.csv")
                if not os.path.exists(csv_path):
                    print(f"âš ï¸ CSV not found for {filename}")
                    continue

                gt_coords = load_gt_coords_from_csv(csv_path, class_channel=c)
                tp, fp, fn = compute_detection_confusion(
                    filtered_coords[c], gt_coords, radius=radius_thresh
                )

                global_tp[c] += tp
                global_fp[c] += fp
                global_fn[c] += fn

                print(f"  Class {c}: TP={tp}, FP={fp}, FN={fn}")

    # Compute F1 scores
    f1_per_class = {}
    precision_per_class = {}
    recall_per_class = {}

    print("\n" + "="*60)
    print("EVALUATION RESULTS - 4-Channel Model (RGB + Tissue)")
    print("="*60)

    for c in [0, 1]:
        tp, fp, fn = global_tp[c], global_fp[c], global_fn[c]

        precision = tp / (tp + fp + 1e-8) if (tp + fp) > 0 else 0.0
        recall = tp / (tp + fn + 1e-8) if (tp + fn) > 0 else 0.0
        f1 = 2 * tp / (2 * tp + fp + fn + 1e-8) if (tp + fp + fn) > 0 else 1.0

        precision_per_class[c] = precision
        recall_per_class[c] = recall
        f1_per_class[c] = f1

        class_name = "Non-Tumor" if c == 0 else "Tumor"
        print(f"\n{class_name} (Class {c}):")
        print(f"  TP: {tp:4d}")
        print(f"  FP: {fp:4d}")
        print(f"  FN: {fn:4d}")
        print(f"  Precision: {precision:.4f}")
        print(f"  Recall:    {recall:.4f}")
        print(f"  F1 Score:  {f1:.4f}")

    # Overall metrics
    overall_f1 = np.mean(list(f1_per_class.values()))
    overall_precision = np.mean(list(precision_per_class.values()))
    overall_recall = np.mean(list(recall_per_class.values()))

    print(f"\nOVERALL METRICS:")
    print(f"  Average Precision: {overall_precision:.4f}")
    print(f"  Average Recall:    {overall_recall:.4f}")
    print(f"  Average F1 Score:  {overall_f1:.4f}")
    print("="*60)

    return f1_per_class, precision_per_class, recall_per_class


def load_full_scale_sample(image_dir, softmax_dir, mask_dir, split='val'):
    """
    Load full-scale 1024x1024 images for evaluation
    """
    samples = []

    # Find all complete triplets
    image_pattern = os.path.join(image_dir, split, "cell", "*.jpg")
    image_files = glob.glob(image_pattern)

    for img_path in image_files:
        basename = os.path.splitext(os.path.basename(img_path))[0]

        # Check for corresponding files
        softmax_path = os.path.join(softmax_dir, split, f"{basename}.npy")
        mask_path = os.path.join(mask_dir, split, "cell", f"{basename}.npy")

        if os.path.exists(softmax_path) and os.path.exists(mask_path):
            try:
                # Load full images (1024x1024)
                image = cv2.imread(img_path)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                softmax = np.load(softmax_path)  # (1024, 1024)
                mask = np.load(mask_path)        # (2, 1024, 1024)

                # No stain normalization applied

                # Ensure softmax is 2D
                if len(softmax.shape) == 3:
                    softmax = softmax.squeeze()

                # Convert to tensors and normalize
                # Normalize image (ImageNet normalization)
                image_norm = image.astype(np.float32) / 255.0
                mean = np.array([0.485, 0.456, 0.406], dtype=np.float32)
                std = np.array([0.229, 0.224, 0.225], dtype=np.float32)
                image_norm = (image_norm - mean) / std
                image_tensor = torch.from_numpy(image_norm).float().permute(2, 0, 1)  # (3, 1024, 1024)

                # Convert softmax to tensor
                softmax_tensor = torch.from_numpy(softmax.astype(np.float32)).float().unsqueeze(0)  # (1, 1024, 1024)

                # Combine RGB + tissue
                combined_input = torch.cat([image_tensor, softmax_tensor], dim=0)  # (4, 1024, 1024)

                # Convert mask to tensor
                mask_tensor = torch.from_numpy(mask.astype(np.float32)).float()  # (2, 1024, 1024)

                samples.append((combined_input, mask_tensor, basename))

            except Exception as e:
                print(f"Error loading {basename}: {e}")
                continue

    return samples
```
