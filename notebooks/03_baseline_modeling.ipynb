```python
import pandas as pd
import scipy.io as sio
import os
import numpy as np
import glob

def process_csv_folder(csv_folder_path, output_dir):
    """
    Process all CSV files in a folder, converting them to MATLAB files separated by class.
    Assumes CSV files have no headers and columns are in order: x, y, class.

    Args:
        csv_folder_path (str): Path to the folder containing CSV files
        output_dir (str): Base directory for output folders
    """
    # Create output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Create class-specific directories
    class1_dir = os.path.join(output_dir, 'class1')
    class2_dir = os.path.join(output_dir, 'class2')

    if not os.path.exists(class1_dir):
        os.makedirs(class1_dir)
    if not os.path.exists(class2_dir):
        os.makedirs(class2_dir)

    # Get all CSV files in the folder
    csv_files = glob.glob(os.path.join(csv_folder_path, "*.csv"))

    if not csv_files:
        return f"No CSV files found in {csv_folder_path}"

    files_processed = 0
    total_class1_points = 0
    total_class2_points = 0
    errors = 0

    # Process each CSV file
    for csv_path in csv_files:
        try:
            # Read the CSV file without headers
            df = pd.read_csv(csv_path, header=None, names=['x', 'y', 'class'])

            # Split by class
            class1_data = df[df['class'] == 1][['x', 'y']].values
            class2_data = df[df['class'] == 2][['x', 'y']].values

            total_class1_points += len(class1_data)
            total_class2_points += len(class2_data)

            # Get base filename without extension
            base_filename = os.path.splitext(os.path.basename(csv_path))[0]

            # Save as MATLAB files
            if len(class1_data) > 0:
                class1_path = os.path.join(class1_dir, f"{base_filename}.mat")
                sio.savemat(class1_path, {'centroids': class1_data})

            if len(class2_data) > 0:
                class2_path = os.path.join(class2_dir, f"{base_filename}.mat")
                sio.savemat(class2_path, {'centroids': class2_data})

            files_processed += 1

        except Exception as e:
            errors += 1

    # Create summary log
    summary = (
        f"Processed {files_processed} CSV files in {csv_folder_path}\n"
        f"Total class 1 points: {total_class1_points}\n"
        f"Total class 2 points: {total_class2_points}\n"
        f"Errors encountered: {errors}\n"
        f"Output saved to {output_dir}"
    )

    return summary

def process_all_sets(base_input_path, base_output_path):
    """
    Process CSV files from train, val, and test sets.

    Args:
        base_input_path (str): Base path to the dataset
        base_output_path (str): Base path for output
    """
    # Define set names
    sets = ["train", "val", "test"]

    for set_name in sets:
        input_path = os.path.join(base_input_path, set_name, "cell")
        output_path = os.path.join(base_output_path, set_name, "cell")

        print(f"\n{'='*50}")
        print(f"Processing {set_name} set")
        print(f"{'='*50}")

        if os.path.exists(input_path):
            summary = process_csv_folder(input_path, output_path)
            print(summary)
        else:
            print(f"Directory not found: {input_path}")

```

### gaussian mask


```python
import os
import numpy as np
import pandas as pd
import glob
from tqdm import tqdm

def generate_gaussian_mask(shape, centers, sigma=4, radius=11):
    """
    Generate a combined Gaussian mask from multiple centers,
    using the maximum value at each pixel when overlaps occur.
    """
    h, w = shape
    mask = np.zeros((h, w), dtype=np.float32)

    for center in centers:
        y, x = center
        y, x = int(round(y)), int(round(x))

        y_min = max(0, y - radius)
        y_max = min(h, y + radius + 1)
        x_min = max(0, x - radius)
        x_max = min(w, x + radius + 1)

        y_grid, x_grid = np.meshgrid(np.arange(y_min, y_max), np.arange(x_min, x_max), indexing='ij')

        # Gaussian kernel
        gaussian = np.exp(-((x_grid - x) ** 2 + (y_grid - y) ** 2) / (2 * sigma ** 2))

        # Apply circular mask
        circle_mask = ((x_grid - x) ** 2 + (y_grid - y) ** 2) <= radius ** 2
        gaussian *= circle_mask

        # Apply max operation in the mask region
        mask_region = mask[y_min:y_max, x_min:x_max]
        mask[y_min:y_max, x_min:x_max] = np.maximum(mask_region, gaussian)

    return mask


def process_annotation_folder(annotation_root, output_root, sigma=4, radius=11, shape=(1024, 1024)):
    subsets = ['train', 'val', 'test']
    os.makedirs(output_root, exist_ok=True)

    for subset in subsets:
        input_folder = os.path.join(annotation_root, subset, "cell")
        output_folder = os.path.join(output_root, subset, "cell")
        os.makedirs(output_folder, exist_ok=True)

        csv_paths = glob.glob(os.path.join(input_folder, "*.csv"))
        print(f"Processing {subset} - {len(csv_paths)} files")

        for csv_path in tqdm(csv_paths):
            if os.stat(csv_path).st_size == 0:
                mask = np.zeros((2, shape[0], shape[1]), dtype=np.float32)
            else:
                df = pd.read_csv(csv_path, header=None)
                df.columns = ['x', 'y', 'class']

                mask = np.zeros((2, shape[0], shape[1]), dtype=np.float32)

                for cls in [1, 2]:
                    class_df = df[df['class'] == cls]
                    centers = [(y, x) for x, y in zip(class_df['x'], class_df['y'])]  # y,x order
                    if centers:
                        mask[cls - 1] = generate_gaussian_mask(shape, centers, sigma=sigma, radius=radius)

            filename = os.path.splitext(os.path.basename(csv_path))[0] + ".npy"
            np.save(os.path.join(output_folder, filename), mask)

# Example usage:
annotation_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations"
output_path = "/content/drive/MyDrive/research/ocelot/dataset/GT/gaussian masks/type3"
process_annotation_folder(annotation_path, output_path)
```

```python
import os
import numpy as np
import pandas as pd
from torch.utils.data import Dataset
from PIL import Image
import albumentations as A
from albumentations.pytorch import ToTensorV2
from pathlib import Path

class CellDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform=None):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)


    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        mask_path = self.mask_paths[idx]

        image = np.array(Image.open(self.image_paths[idx]).convert("RGB"))  # [H, W, 3]
        mask = np.load(self.mask_paths[idx])  # [2, H, W]
        mask = np.transpose(mask, (1, 2, 0))        # shape: (H, W, 2)

        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']           # Tensor: [3, H, W]
            mask = augmented['mask']             # Tensor: [H, W, 2]
            mask = mask.permute(2, 0, 1)         # Convert to [2, H, W]

        # Ensure both have the same filename stem
        img_stem = Path(img_path).stem
        mask_stem = Path(mask_path).stem
        assert img_stem == mask_stem, f"Filename mismatch: {img_stem} vs {mask_stem}"

        return image, mask, img_stem

```

```python
import glob
from torch.utils.data import DataLoader

def get_dataloaders(base_img_path, base_mask_path, patch_size, batch_size):
    def get_paths(split):
        img_dir = os.path.join(base_img_path, split, "cell")
        mask_dir = os.path.join(base_mask_path, split, "cell")


        image_paths = sorted(glob.glob(f"{img_dir}/*.jpg"))  # or .png
        mask_paths = []


        for img_path in image_paths:
            filename = os.path.splitext(os.path.basename(img_path))[0]
            mask_paths.append(os.path.join(mask_dir, filename + ".npy"))


        return image_paths, mask_paths

    train_img, train_mask = get_paths("train")
    val_img, val_mask = get_paths("test")

    train_dataset = CellDataset(train_img, train_mask, transform=get_train_transform(patch_size))
    val_dataset = CellDataset(val_img, val_mask, transform=get_val_transform(patch_size))

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2,worker_init_fn=seed_worker, generator=g)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

    return train_loader, val_loader
```

```python
def get_train_transform(patch_size):
    return A.Compose([
        A.RandomCrop(width=patch_size, height=patch_size),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.Normalize(mean=(0.485, 0.456, 0.406),
                    std=(0.229, 0.224, 0.225)),
        ToTensorV2()
    ])

def get_val_transform(patch_size):
    return A.Compose([
        #A.CenterCrop(height=patch_size, width=patch_size),# ini masalahnya
        A.Normalize(mean=(0.485, 0.456, 0.406),
                    std=(0.229, 0.224, 0.225)),
        ToTensorV2()
    ])
```

### model definition

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torch.nn.functional as F

class AttentionBlock(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super(AttentionBlock, self).__init__()
        self.W_g = nn.Sequential(
            nn.Conv2d(F_g, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )

        self.W_x = nn.Sequential(
            nn.Conv2d(F_l, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )

        self.psi = nn.Sequential(
            nn.Conv2d(F_int, 1, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(1),
            nn.Sigmoid()
        )

        self.relu = nn.ReLU(inplace=True)

    def forward(self, g, x):
        g1 = self.W_g(g)
        x1 = self.W_x(x)
        psi = self.relu(g1 + x1)
        psi = self.psi(psi)
        return x * psi


class UpConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super(UpConv, self).__init__()
        self.up = nn.ConvTranspose2d(in_ch, out_ch, kernel_size=2, stride=2)

    def forward(self, x):
        x = self.up(x)
        return x


class AttentionUNet(nn.Module):
    def __init__(self, encoder='resnet34', num_classes=2, dropout=0.0, pretrained=True):
        super(AttentionUNet, self).__init__()

        assert encoder == 'resnet34', "Only resnet34 is supported currently"

        resnet = models.resnet34(pretrained=pretrained)
        self.dropout = dropout  # Store for conditional dropout if needed

        # Encoder layers
        self.encoder0 = nn.Sequential(resnet.conv1, resnet.bn1, resnet.relu)  # 64, /2
        self.encoder1 = nn.Sequential(resnet.maxpool, resnet.layer1)          # 64, /4
        self.encoder2 = resnet.layer2                                         # 128, /8
        self.encoder3 = resnet.layer3                                         # 256, /16
        self.encoder4 = resnet.layer4                                         # 512, /32

        # Decoder
        self.up4 = UpConv(512, 256)
        self.att4 = AttentionBlock(F_g=256, F_l=256, F_int=128)
        self.dec4 = self._make_decoder_block(512, 256)

        self.up3 = UpConv(256, 128)
        self.att3 = AttentionBlock(F_g=128, F_l=128, F_int=64)
        self.dec3 = self._make_decoder_block(256, 128)

        self.up2 = UpConv(128, 64)
        self.att2 = AttentionBlock(F_g=64, F_l=64, F_int=32)
        self.dec2 = self._make_decoder_block(128, 64)

        self.up1 = UpConv(64, 64)
        self.att1 = AttentionBlock(F_g=64, F_l=64, F_int=32)
        self.dec1 = self._make_decoder_block(128, 64)

        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def _make_decoder_block(self, in_ch, out_ch):
        layers = [
            nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        ]
        if self.dropout > 0:
            layers.append(nn.Dropout2d(self.dropout))
        return nn.Sequential(*layers)

    def forward(self, x):
        e0 = self.encoder0(x)
        e1 = self.encoder1(e0)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)

        d4 = self.up4(e4)
        e3_att = self.att4(d4, e3)
        d4 = torch.cat((e3_att, d4), dim=1)
        d4 = self.dec4(d4)

        d3 = self.up3(d4)
        e2_att = self.att3(d3, e2)
        d3 = torch.cat((e2_att, d3), dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        e1_att = self.att2(d2, e1)
        d2 = torch.cat((e1_att, d2), dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        e0_att = self.att1(d1, e0)
        d1 = torch.cat((e0_att, d1), dim=1)
        d1 = self.dec1(d1)
        out = self.final_conv(d1)
        out = F.interpolate(out, size=x.shape[2:], mode='bilinear', align_corners=False)  # <--- Add this
        return out

```

### loss and optimizers setup

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

def tversky_loss(probs, targets, beta=0.3, smooth=1e-6):
    probs = probs.view(-1)
    targets = targets.view(-1)

    TP = (probs * targets).sum()
    FP = ((1 - targets) * probs).sum()
    FN = (targets * (1 - probs)).sum()

    tversky = (TP + smooth) / (TP + beta * FP + (1 - beta) * FN + smooth)
    return 1 - tversky
```

```python
class ComboLoss(nn.Module):
    def __init__(self, alpha=0.5, tversky_beta=0.5):
        """
        Combination of BCE and Tversky losses.

        Args:
            alpha: weight factor for BCE vs Tversky (1.0 = pure BCE, 0.0 = pure Tversky)
            tversky_beta: beta parameter for the Tversky loss component
        """
        super().__init__()
        self.alpha = alpha
        self.tversky_beta = tversky_beta
        self.bce = nn.BCEWithLogitsLoss()

    def forward(self, logits, targets):
        # BCE component (works with logits)
        bce_loss = self.bce(logits, targets)

        # Tversky component (needs probabilities)
        probs = torch.sigmoid(logits)
        tversky = tversky_loss(probs, targets, beta=self.tversky_beta)

        # Weighted combination
        combo = self.alpha * bce_loss + (1 - self.alpha) * tversky
        return combo

class FocalLoss(nn.Module):
    def __init__(self, alpha=1.0, gamma=2.0, reduction='mean'):
        """
        Focal Loss for addressing class imbalance

        Args:
            alpha (float): Weighting factor for rare class (cells vs background)
            gamma (float): Focusing parameter (higher = more focus on hard examples)
            reduction (str): 'mean', 'sum', or 'none'
        """
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        """
        Args:
            inputs: (N, C, H, W) - Raw logits from model
            targets: (N, C, H, W) - Target probability maps (0-1)
        """
        # Convert logits to probabilities
        probs = torch.sigmoid(inputs)

        # Probability-weighted focal loss
        # Weight positive contribution by target probability
        pos_loss = -self.alpha * (1 - probs) ** self.gamma * targets * torch.log(probs + 1e-8)

        # Weight negative contribution by (1 - target probability)
        neg_loss = -(1 - self.alpha) * probs ** self.gamma * (1 - targets) * torch.log(1 - probs + 1e-8)

        # Combine losses
        focal_loss = pos_loss + neg_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss
```

```python
# Updated factory function to match your style
def get_loss_function(loss_name, config=None, alpha=0.5):
    """
    Create loss function with parameters from wandb config or defaults

    Args:
        loss_name: 'bce', 'combo', 'focal'
        config: wandb.config object (optional)
        alpha: default alpha for combo loss
    """
    if loss_name == 'bce':
        return nn.BCEWithLogitsLoss()
    elif loss_name == 'combo':
        return ComboLoss(alpha=alpha)
    elif loss_name == 'focal':
        if config is not None:
            # Use swept parameters from wandb
            return FocalLoss(
                alpha=config.focal_alpha,
                gamma=config.focal_gamma,
                reduction='mean'
            )
        else:
            # Use default values
            return FocalLoss(alpha=1.0, gamma=2.0)
    else:
        raise ValueError(f"Unsupported loss: {loss_name}")
```

```python
def get_optimizer(optimizer_name, model_params, learning_rate, weight_decay=0.0):
    """
    Create optimizer with optional weight decay and momentum (for SGD).

    Args:
        optimizer_name (str): 'adam', 'adamw', or 'sgd'
        model_params: model.parameters()
        learning_rate (float)
        weight_decay (float): L2 regularization
        momentum (float or None): required for SGD

    Returns:
        optimizer (torch.optim.Optimizer)
    """
    if optimizer_name == 'adam':
        return torch.optim.Adam(model_params, lr=learning_rate, weight_decay=weight_decay)

    elif optimizer_name == 'adamw':
        return torch.optim.AdamW(model_params, lr=learning_rate, weight_decay=weight_decay)
    else:
        raise ValueError(f"Unsupported optimizer: {optimizer_name}")
```

```python
import torch.optim as optim

def create_scheduler(optimizer, config):
    """
    Create learning rate scheduler based on wandb config

    Args:
        optimizer: PyTorch optimizer
        config: wandb config object with scheduler parameters

    Returns:
        scheduler: Configured scheduler or None
    """

    if not hasattr(config, 'scheduler') or config.scheduler == 'none':
        print("âœ… No learning rate scheduler")
        return None

    elif config.scheduler == 'plateau':
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer,
            mode='min',           # For metrics like F1 score
            factor=0.5,           # Reduce by 50%
            patience=35,          # Wait 10 epochs
            verbose=True
        )
        print(f"âœ… Plateau scheduler created (factor=0.5, patience=4)")

    else:
        raise ValueError(f"Unsupported scheduler: {config.scheduler}")

    return scheduler
```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn.functional as F
from tqdm import tqdm
import wandb
from sklearn.metrics import precision_score, recall_score, f1_score
from scipy import ndimage
from skimage.feature import peak_local_max
from scipy.spatial import cKDTree


def detect_local_maxima(prob_map, min_distance=6, threshold_abs=0.1):
    peaks = peak_local_max(prob_map, min_distance=min_distance, threshold_abs=threshold_abs)
    coords = [(x, y) for y, x in peaks]  # swap (y, x) to (x, y)
    return coords


def compute_detection_confusion(pred_coords, gt_coords, radius=15):
    matched_gt = set()
    tp, fp = 0, 0

    for px, py in pred_coords:
        found = False
        min_dist = float('inf')
        nearest_idx = -1

        for idx, (gx, gy) in enumerate(gt_coords):
            if idx in matched_gt:
                continue
            dist = np.sqrt((px - gx) ** 2 + (py - gy) ** 2)
            if dist < radius and dist < min_dist:
                found = True
                min_dist = dist
                nearest_idx = idx

        if found:
            tp += 1
            matched_gt.add(nearest_idx)
        else:
            fp += 1

    fn = len(gt_coords) - len(matched_gt)
    return tp, fp, fn

def extract_confidences_from_coords(pred_mask, pred_coords):
    if len(pred_coords) == 0:
        return np.array([])  # Return empty confidence array

    h, w = pred_mask.shape
    coords = np.rint(pred_coords).astype(np.intp)  # round and convert to int

    # Ensure coords is 2D
    coords = coords.reshape(-1, 2)

    coords[:, 0] = np.clip(coords[:, 0], 0, w - 1)  # x
    coords[:, 1] = np.clip(coords[:, 1], 0, h - 1)  # y
    return pred_mask[coords[:, 1], coords[:, 0]]

def load_gt_coords_from_csv(csv_path, class_channel):
    coords = []
    if not os.path.exists(csv_path) or os.path.getsize(csv_path) == 0:
        return []
    with open(csv_path, 'r') as f:
        for line in f.readlines():
            line = line.strip()  # âœ… Remove leading/trailing whitespace
            if not line:
              continue

            parts = [p.strip() for p in line.split(',')]  # âœ… Remove extra spaces around each value
            if len(parts) != 3:
                continue
            # x, y, c = int(parts[0]), int(parts[1]), int(parts[2])
            try:
                x = int(parts[0])
                y = int(parts[1])
                c = int(parts[2])
            except ValueError:
                continue  # Skip lines with non-integer values

            if c == class_channel + 1:  # 1=non-tumor, 2=tumor
                coords.append((x, y))
    return coords

def filter_overlapping_predictions(all_coords, all_confidences, min_distance=7):
    """
    Filters overlapping predicted coordinates between classes based on confidence.

    Parameters
    ----------
    all_coords : dict[int, List[Tuple[int, int]]]
        Dictionary of predicted coordinates per class.
    all_confidences : dict[int, List[float]]
        Dictionary of confidence scores per class.
    min_distance : int
        Minimum distance threshold to consider two predictions as overlapping.

    Returns
    -------
    filtered_coords : dict[int, List[Tuple[int, int]]]
        Filtered predicted coordinates per class.
    filtered_confidences : dict[int, List[float]]
        Corresponding confidence scores for the filtered coordinates.
    """
    # Collect all predictions with their class and index
    all_predictions = []
    for class_id in [0, 1]:
        for idx, (coord, conf) in enumerate(zip(all_coords[class_id], all_confidences[class_id])):
            all_predictions.append({
                'coord': coord,
                'confidence': conf,
                'class': class_id,
                'original_idx': idx
            })

    # Sort by confidence (highest first)
    all_predictions.sort(key=lambda x: x['confidence'], reverse=True)

    # Keep track of which predictions to keep
    kept_predictions = []

    for current_pred in all_predictions:
        keep_current = True

        # Check against all already kept predictions
        for kept_pred in kept_predictions:
            dist = np.linalg.norm(np.array(current_pred['coord']) - np.array(kept_pred['coord']))

            if dist < min_distance:
                # If they're overlapping, we don't keep the current one
                # (since we're processing in order of decreasing confidence,
                # the kept one has higher or equal confidence)
                keep_current = False
                break

        if keep_current:
            kept_predictions.append(current_pred)

    # Reconstruct the filtered results
    filtered_coords = {0: [], 1: []}
    filtered_confidences = {0: [], 1: []}

    for pred in kept_predictions:
        class_id = pred['class']
        filtered_coords[class_id].append(pred['coord'])
        filtered_confidences[class_id].append(pred['confidence'])

    return filtered_coords, filtered_confidences
```

```python
def train_model(model, train_loader, val_loader, base_csv_path, optimizer, scheduler, loss_fn, device, config):
        model.to(device)
        best_f1 = 0

        patch_size = config.patch_size
        stride = int(patch_size/2)  # Can set smaller for overlap

        best_models = []  # Will store tuples: (val_f1, epoch, model_state_dict)
        TOP_K = 1

        run_name = wandb.run.name  # e.g., "experiment-42"

        low_f1_patience = 10
        low_f1_threshold = 0.25
        low_f1_counter = 0

        for epoch in range(config.epochs):
            model.train()
            train_loss = 0.0

            for images, masks, _ in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
                images, masks = images.to(device), masks.to(device)
                optimizer.zero_grad()
                outputs = model(images)
                loss = loss_fn(outputs, masks)
                loss.backward()
                optimizer.step()
                train_loss += loss.item()

            avg_train_loss = train_loss / len(train_loader)

##################################################################################
            # Validation
##################################################################################

            model.eval()
            val_loss = 0.0
            all_f1 = []

            global_tp = {0: 0, 1: 0}
            global_fp = {0: 0, 1: 0}
            global_fn = {0: 0, 1: 0}

            ENABLE_FILTERING = False

            with torch.no_grad():
                for images, masks, filenames in tqdm(val_loader, desc="Validation", leave=False):
                    # for i in tqdm(range(len(images)), desc=f"Processing batch images", leave=False):
                    for i in range(len(images)):
                        image = images[i].to(device).unsqueeze(0)
                        mask = masks[i].to(device).unsqueeze(0)
                        filename = filenames[i]

                        H, W = image.shape[2:]
                        probs_full = torch.zeros((2, H, W), dtype=torch.float32).to(device)
                        count_map = torch.zeros((1, H, W), dtype=torch.float32).to(device)

                        # Sliding window inference
                        for y in range(0, H, stride):
                            for x in range(0, W, stride):
                                y1 = y
                                y2 = min(y + patch_size, H)
                                x1 = x
                                x2 = min(x + patch_size, W)

                                img_patch = image[:, :, y1:y2, x1:x2]

                                out_patch = model(img_patch)
                                prob_patch = torch.sigmoid(out_patch)[0, :, :y2 - y1, :x2 - x1]

                                probs_full[:, y1:y2, x1:x2] += prob_patch
                                count_map[:, y1:y2, x1:x2] += 1

                        probs_full = probs_full / count_map
                        loss = loss_fn(probs_full.unsqueeze(0), mask)
                        val_loss += loss.item()

                        probs_np = probs_full.cpu().numpy()

                        # Extract predictions for both classes
                        all_coords = {}
                        all_confidences = {}

                        for c in range(2):
                            pred_mask = probs_np[c]
                            pred_coords = detect_local_maxima(pred_mask, min_distance=config.min_distance, threshold_abs=0.1)
                            confidences = extract_confidences_from_coords(pred_mask, pred_coords)

                            all_coords[c] = pred_coords
                            all_confidences[c] = confidences

                        if ENABLE_FILTERING:
                            # Apply overlap filtering across classes
                            filtered_coords, filtered_confidences = filter_overlapping_predictions(
                                all_coords, all_confidences, min_distance=config.min_distance
                            )
                        else:
                            filtered_coords, filtered_confidences = all_coords, all_confidences

                        csv_path = os.path.join(base_csv_path, f"{filename}.csv")
                        if not os.path.exists(csv_path):
                            continue

                        for c in range(2):
                            gt_coords = load_gt_coords_from_csv(csv_path, class_channel=c)

                            # Use filtered predictions
                            pred_coords_filtered = filtered_coords[c]
                            confidences_filtered = filtered_confidences[c]

                            # Sort by confidence (highest first)
                            if len(confidences_filtered) > 0:
                                sorted_idx = np.argsort(confidences_filtered)[::-1]
                                sorted_coords = [pred_coords_filtered[i] for i in sorted_idx]
                            else:
                                sorted_coords = []

                            tp, fp, fn = compute_detection_confusion(sorted_coords, gt_coords)
                            global_tp[c] += tp
                            global_fp[c] += fp
                            global_fn[c] += fn

            f1_per_class = {}
            precision_per_class = {}
            recall_per_class = {}

            for c in [0, 1]:
                tp = global_tp[c]
                fp = global_fp[c]
                fn = global_fn[c]

                precision = tp / (tp + fp + 1e-8) if (tp + fp) > 0 else 1.0
                recall = tp / (tp + fn + 1e-8) if (tp + fn) > 0 else 1.0
                f1 = 2 * precision * recall / (precision + recall + 1e-8) if (precision + recall) > 0 else 1.0

                precision_per_class[c] = precision
                recall_per_class[c] = recall
                f1_per_class[c] = f1

            avg_val_loss = val_loss / len(val_loader)
            avg_val_f1 = np.mean(list(f1_per_class.values()))  # âœ… CORRECT

            # Check if avg_val_f1 is below the threshold
            if avg_val_f1 < low_f1_threshold:
                low_f1_counter += 1
                print(f"Epoch {epoch}: val_f1={avg_val_f1:.4f} < {low_f1_threshold}, counter={low_f1_counter}")
            else:
                low_f1_counter = 0  # reset if above threshold

            # Trigger early stopping
            if low_f1_counter >= low_f1_patience:
                print(f"Early stopping: val_f1 stayed below {low_f1_threshold} for {low_f1_patience} epochs.")
                break

            # Step scheduler (only if plateau)
            if scheduler and config.scheduler == 'plateau':
                scheduler.step(avg_val_loss)  # Use validation loss

            # Keep track of top 3 models
            if len(best_models) < TOP_K or avg_val_f1 > min(m[0] for m in best_models):
                model_cpu_state = {k: v.cpu() for k, v in model.state_dict().items()}
                best_models.append((avg_val_f1, epoch, model_cpu_state))
                best_models = sorted(best_models, key=lambda x: x[0], reverse=True)[:TOP_K]

            # Get current learning rate for logging
            current_lr = optimizer.param_groups[0]['lr']

            wandb.log({
                "epoch": epoch,
                "train_loss": avg_train_loss,
                "val_loss": avg_val_loss,
                "val_detection_f1": avg_val_f1,
                "learning_rate": current_lr,
                "val_f1_class_0": f1_per_class[0],
                "val_f1_class_1": f1_per_class[1],
                "val_precision_class_0": precision_per_class[0],
                "val_precision_class_1": precision_per_class[1],
                "val_recall_class_0": recall_per_class[0],
                "val_recall_class_1": recall_per_class[1]
            })

            print(f"Epoch {epoch+1}/{config.epochs}")
            print(f"Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}")
            print(f"Val F1: {avg_val_f1:.4f} (Class 0: {f1_per_class[0]:.4f}, Class 1: {f1_per_class[1]:.4f})")
            print(f"Learning Rate: {current_lr:.6f}")
            print("-" * 50)

        for rank, (f1, epoch, state_dict) in enumerate(best_models, start=1):
            checkpoint_path = f"best_{run_name}_rank{rank}_epoch{epoch+1}_f1={f1:.4f}.pt"
            torch.save(state_dict, checkpoint_path)

            artifact = wandb.Artifact(f"model_rank{rank}", type="model")
            artifact.add_file(checkpoint_path)
            wandb.log_artifact(artifact)

        print(f"\nðŸŽ‰ Training completed!")
        print(f"Best validation F1: {best_models[0][0]:.4f} at epoch {best_models[0][1]+1}")
        return best_models[0][2]  # Return best model state dict
```
