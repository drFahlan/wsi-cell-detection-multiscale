```python
import os
import torch
from torch.utils.data import Dataset
from PIL import Image
import albumentations as A
from albumentations.pytorch import ToTensorV2

class TissueDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform=None):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        mask_path = self.mask_paths[idx]
        filename = os.path.basename(image_path)

        # Load RGB image and grayscale mask
        image = Image.open(image_path).convert("RGB")
        mask = Image.open(mask_path).convert("L")

        image = np.array(image)
        mask = np.array(mask)

        # Remap mask values: 1â†’0, 2â†’1, 255â†’255 (ignore)
        mask = np.where(mask == 1, 0, mask)
        mask = np.where(mask == 2, 1, mask)

        # Apply transform if defined
        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']

        return {
            "image": image,  # Already float via Albumentations
            "mask": mask.float(),  # Instead of .long()
            "filename": filename
        }
```

```python
from torch.utils.data import DataLoader
import glob

def get_image_mask_paths(image_dir, mask_dir, ext_img="jpg", ext_mask="png"):
    image_paths = sorted(glob.glob(os.path.join(image_dir, f"*.{ext_img}")))
    mask_paths = sorted(glob.glob(os.path.join(mask_dir, f"*.{ext_mask}")))
    assert len(image_paths) == len(mask_paths), "Mismatch in number of images and masks"
    return image_paths, mask_paths

def get_dataloaders(train_img_dir, train_mask_dir, val_img_dir, val_mask_dir,
                    train_transform, val_transform, batch_size=8, num_workers=2):

    train_image_paths, train_mask_paths = get_image_mask_paths(train_img_dir, train_mask_dir)
    val_image_paths, val_mask_paths = get_image_mask_paths(val_img_dir, val_mask_dir)

    train_dataset = TissueDataset(train_image_paths, train_mask_paths, transform=train_transform)
    val_dataset = TissueDataset(val_image_paths, val_mask_paths, transform=val_transform)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)

    return train_loader, val_loader
```

```python
def get_transforms(patch_size=512):
    train_transform = A.Compose([
        A.RandomCrop(patch_size, patch_size),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.ElasticTransform(alpha=1, sigma=50, p=0.2),
        A.OneOf([
            A.GaussianBlur(p=0.2),
            A.MotionBlur(p=0.2),
        ], p=0.3),
        A.ColorJitter(p=0.3),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])

    val_transform = A.Compose([
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])

    return train_transform, val_transform
```

#### Attention U-Net

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torch.nn.functional as F
import timm
from torch import nn

class AttentionBlock(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super(AttentionBlock, self).__init__()
        self.W_g = nn.Sequential(
            nn.Conv2d(F_g, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )

        self.W_x = nn.Sequential(
            nn.Conv2d(F_l, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )

        self.psi = nn.Sequential(
            nn.Conv2d(F_int, 1, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(1),
            nn.Sigmoid()
        )

        self.relu = nn.ReLU(inplace=True)

    def forward(self, g, x):
        g1 = self.W_g(g)
        x1 = self.W_x(x)
        psi = self.relu(g1 + x1)
        psi = self.psi(psi)
        return x * psi


class UpConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super(UpConv, self).__init__()
        self.up = nn.ConvTranspose2d(in_ch, out_ch, kernel_size=2, stride=2)

    def forward(self, x):
        x = self.up(x)
        return x


class AttentionUNet(nn.Module):
    def __init__(self, encoder='se_resnet50', num_classes=1, dropout=0.0, pretrained=True):
        super(AttentionUNet, self).__init__()

        assert encoder == 'resnet34', "Only resnet34 is supported currently"

        resnet = models.resnet34(pretrained=pretrained)
        self.dropout = dropout  # Store for conditional dropout if needed

        # Encoder layers
        self.encoder0 = nn.Sequential(resnet.conv1, resnet.bn1, resnet.relu)  # 64, /2
        self.encoder1 = nn.Sequential(resnet.maxpool, resnet.layer1)          # 64, /4
        self.encoder2 = resnet.layer2                                         # 128, /8
        self.encoder3 = resnet.layer3                                         # 256, /16
        self.encoder4 = resnet.layer4                                         # 512, /32

        # Decoder
        self.up4 = UpConv(512, 256)
        self.att4 = AttentionBlock(F_g=256, F_l=256, F_int=128)
        self.dec4 = self._make_decoder_block(512, 256)

        self.up3 = UpConv(256, 128)
        self.att3 = AttentionBlock(F_g=128, F_l=128, F_int=64)
        self.dec3 = self._make_decoder_block(256, 128)

        self.up2 = UpConv(128, 64)
        self.att2 = AttentionBlock(F_g=64, F_l=64, F_int=32)
        self.dec2 = self._make_decoder_block(128, 64)

        self.up1 = UpConv(64, 64)
        self.att1 = AttentionBlock(F_g=64, F_l=64, F_int=32)
        self.dec1 = self._make_decoder_block(128, 64)

        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def _make_decoder_block(self, in_ch, out_ch):
        layers = [
            nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        ]
        if self.dropout > 0:
            layers.append(nn.Dropout2d(self.dropout))
        return nn.Sequential(*layers)

    def forward(self, x):
        e0 = self.encoder0(x)
        e1 = self.encoder1(e0)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)

        d4 = self.up4(e4)
        e3_att = self.att4(d4, e3)
        d4 = torch.cat((e3_att, d4), dim=1)
        d4 = self.dec4(d4)

        d3 = self.up3(d4)
        e2_att = self.att3(d3, e2)
        d3 = torch.cat((e2_att, d3), dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        e1_att = self.att2(d2, e1)
        d2 = torch.cat((e1_att, d2), dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        e0_att = self.att1(d1, e0)
        d1 = torch.cat((e0_att, d1), dim=1)
        d1 = self.dec1(d1)
        out = self.final_conv(d1)
        out = F.interpolate(out, size=x.shape[2:], mode='bilinear', align_corners=False)  # <--- Add this
        return out

```

### loss and optimizers

```python
import torch.nn as nn
class BCEWithLogitsDiceLoss(nn.Module):
    def __init__(self, alpha=0.5, smooth=1e-6, ignore_index=255):
        super().__init__()
        self.alpha = alpha
        self.smooth = smooth
        self.ignore_index = ignore_index

    def forward(self, logits, targets):

        # Ensure targets have shape [B, 1, H, W]
        if targets.dim() == 3:
            targets = targets.unsqueeze(1)

        assert logits.shape == targets.shape, f"Shape mismatch: {logits.shape} vs {targets.shape}"


        # Ignore mask
        valid_mask = (targets != self.ignore_index).float()
        targets = targets.float()

        # BCE with logits (no sigmoid needed)
        bce_loss = F.binary_cross_entropy_with_logits(
            logits, targets, reduction='none'
        )
        bce_loss = (bce_loss * valid_mask).sum() / (valid_mask.sum() + self.smooth)

        # Dice on probabilities (requires sigmoid)
        probs = torch.sigmoid(logits)
        probs = probs.view(-1)
        targets = targets.view(-1)
        valid_mask = valid_mask.view(-1)

        probs = probs * valid_mask
        targets = targets * valid_mask

        intersection = (probs * targets).sum()
        union = probs.sum() + targets.sum()
        dice_score = (2 * intersection + self.smooth) / (union + self.smooth)
        dice_loss = 1 - dice_score

        return self.alpha * bce_loss + (1 - self.alpha) * dice_loss
```

```python
def get_loss(config, smooth=1e-6, ignore_index=255):
    """
    Factory function to create loss function with alpha from config.

    Args:
        config: Configuration object containing alpha (e.g., wandb.config)
        smooth (float): Smoothing factor for Dice loss
        ignore_index (int): Index to ignore in loss computation

    Returns:
        BCEDiceLoss: Configured loss function with alpha from config
    """
    # Get alpha from config (swept parameter)
    alpha = config.alpha

    return BCEWithLogitsDiceLoss(alpha=alpha, smooth=smooth, ignore_index=ignore_index)
```

```python
import torch
import torch.optim as optim

def get_optimizer(model_parameters, config):
    """
    Factory function to create optimizer based on config.
    Supports Adam and AdamW with swept weight decay values.

    Args:
        model_parameters: Model parameters (model.parameters())
        config: Configuration object containing optimizer hyperparameters
                Expected config attributes:
                - optimizer: 'adam' or 'adamw'
                - learning_rate: Learning rate value
                - weight_decay: Weight decay value (swept parameter)

    Returns:
        torch.optim.Optimizer: Configured optimizer
    """

    if config.optimizer == 'adam':
        return optim.Adam(
            model_parameters,
            lr=config.learning_rate,
            weight_decay=config.weight_decay,
            betas=(0.9, 0.999),  # Default Adam betas
            eps=1e-8             # Default epsilon
        )
    elif config.optimizer == 'adamw':
        return optim.AdamW(
            model_parameters,
            lr=config.learning_rate,
            weight_decay=config.weight_decay,
            betas=(0.9, 0.999),  # Default AdamW betas
            eps=1e-8             # Default epsilon
        )
    else:
        raise ValueError(f"Unsupported optimizer: {config.optimizer}. Choose 'adam' or 'adamw'")
```

```python
import torch.optim as optim

def get_scheduler(optimizer, config):
    """
    Create learning rate scheduler based on wandb config

    Args:
        optimizer: PyTorch optimizer
        config: wandb config object with scheduler parameters

    Returns:
        scheduler: Configured scheduler or None
    """

    if not hasattr(config, 'scheduler') or config.scheduler == 'none':
        print("âœ… No learning rate scheduler")
        return None

    elif config.scheduler == 'plateau':
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer,
            mode='min',           # For metrics like F1 score
            factor=0.5,           # Reduce by 50%
            patience=40,          # Wait 10 epochs
            verbose=True
        )
        print(f"âœ… Plateau scheduler created (factor=0.5, patience=40)")

    else:
        raise ValueError(f"Unsupported scheduler: {config.scheduler}")

    return scheduler
```

```python
import torch
import torch.nn.functional as F
import wandb
import os
from pathlib import Path
import numpy as np
from torch.utils.data import DataLoader

def sliding_window_inference(model, image, patch_size, stride, device):
    """
    Perform sliding window inference for images larger than patch_size.

    Args:
        model: Trained model
        image: Input image tensor [1, C, H, W]
        patch_size: Size of patches for inference
        stride: Stride for sliding window
        device: Device to run inference on

    Returns:
        torch.Tensor: Prediction for full image [1, 1, H, W]
    """
    model.eval()
    _, _, H, W = image.shape

    if H <= patch_size and W <= patch_size:
        # Image is smaller than patch size, direct inference
        with torch.no_grad():
            return model(image.to(device))

    # Initialize prediction and count arrays
    prediction = torch.zeros(1, 1, H, W, device=device)
    count = torch.zeros(1, 1, H, W, device=device)

    # Sliding window inference
    for y in range(0, H - patch_size + 1, stride):
        for x in range(0, W - patch_size + 1, stride):
            # Extract patch
            patch = image[:, :, y:y+patch_size, x:x+patch_size].to(device)

            # Get prediction for patch
            with torch.no_grad():
                patch_pred = model(patch)

            # Add to prediction and count
            prediction[:, :, y:y+patch_size, x:x+patch_size] += patch_pred
            count[:, :, y:y+patch_size, x:x+patch_size] += 1

    # Handle edges if image size is not divisible by stride
    # Right edge
    if (W - patch_size) % stride != 0:
        x = W - patch_size
        for y in range(0, H - patch_size + 1, stride):
            patch = image[:, :, y:y+patch_size, x:x+patch_size].to(device)
            with torch.no_grad():
                patch_pred = model(patch)
            prediction[:, :, y:y+patch_size, x:x+patch_size] += patch_pred
            count[:, :, y:y+patch_size, x:x+patch_size] += 1

    # Bottom edge
    if (H - patch_size) % stride != 0:
        y = H - patch_size
        for x in range(0, W - patch_size + 1, stride):
            patch = image[:, :, y:y+patch_size, x:x+patch_size].to(device)
            with torch.no_grad():
                patch_pred = model(patch)
            prediction[:, :, y:y+patch_size, x:x+patch_size] += patch_pred
            count[:, :, y:y+patch_size, x:x+patch_size] += 1

    # Bottom-right corner
    if (H - patch_size) % stride != 0 and (W - patch_size) % stride != 0:
        y, x = H - patch_size, W - patch_size
        patch = image[:, :, y:y+patch_size, x:x+patch_size].to(device)
        with torch.no_grad():
            patch_pred = model(patch)
        prediction[:, :, y:y+patch_size, x:x+patch_size] += patch_pred
        count[:, :, y:y+patch_size, x:x+patch_size] += 1

    # Average overlapping predictions
    prediction = prediction / (count + 1e-8)

    return prediction

def validate_model(model, val_loader, loss_fn, device, patch_size=1024):
    """
    Validate model with sliding window inference if needed.
    """
    model.eval()
    val_losses = []
    total_tp = 0  # True Positives
    total_fp = 0  # False Positives
    total_fn = 0  # False Negatives

    stride = patch_size // 2  # stride = patch_size / 2

    with torch.no_grad():
        for batch in tqdm(val_loader, desc="Validation", leave=False):
            images = batch['image']  # [B, 3, H, W]
            masks = batch['mask']    # [B, H, W]

            batch_predictions = []

            # Process each image in the batch
            for i in range(images.shape[0]):
                image = images[i:i+1]  # [1, 3, H, W]

                # Use sliding window if image is larger than patch_size
                if image.shape[2] > patch_size or image.shape[3] > patch_size:
                    pred = sliding_window_inference(model, image, patch_size, stride, device)
                else:
                    pred = model(image.to(device))

                batch_predictions.append(pred)

            # Stack predictions
            predictions = torch.cat(batch_predictions, dim=0)  # [B, 1, H, W]
            masks = masks.to(device)

            # Compute loss
            loss = loss_fn(predictions, masks)
            val_losses.append(loss.item())

            # Convert predictions to binary masks
            pred_binary = (torch.sigmoid(predictions) > 0.5).float()  # [B, 1, H, W]
            pred_binary = pred_binary.squeeze(1)  # [B, H, W]

            # Flatten predictions and targets, excluding ignore_index pixels
            for i in range(predictions.shape[0]):
                pred_flat = pred_binary[i].flatten()  # [H*W]
                mask_flat = masks[i].flatten()        # [H*W]

                # Create mask for valid pixels (exclude ignore_index=255)
                valid_mask = mask_flat != 255
                pred_valid = pred_flat[valid_mask]
                mask_valid = mask_flat[valid_mask]

                # Accumulate confusion matrix components
                tp = ((pred_valid == 1) & (mask_valid == 1)).sum().item()
                fp = ((pred_valid == 1) & (mask_valid == 0)).sum().item()
                fn = ((pred_valid == 0) & (mask_valid == 1)).sum().item()

                total_tp += tp
                total_fp += fp
                total_fn += fn

    # Compute overall metrics from accumulated confusion matrix
    # Precision = TP / (TP + FP)
    precision = total_tp / (total_tp + total_fp) if (total_tp + total_fp) > 0 else 0.0

    # Recall = TP / (TP + FN)
    recall = total_tp / (total_tp + total_fn) if (total_tp + total_fn) > 0 else 0.0

    # F1 = 2 * (Precision * Recall) / (Precision + Recall)
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0

    # Average loss
    avg_metrics = {
        'val_loss': np.mean(val_losses),
        'val_precision': precision,
        'val_recall': recall,
        'val_f1': f1,
        # Optional: Add confusion matrix info for debugging
        'total_tp': total_tp,
        'total_fp': total_fp,
        'total_fn': total_fn
    }

    return avg_metrics

def train_epoch(model, train_loader, optimizer, loss_fn, device, epoch):
    """
    Train for one epoch.
    """
    model.train()
    train_losses = []

    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
        images = batch['image'].to(device)  # [B, 3, H, W]
        masks = batch['mask'].to(device)    # [B, H, W]

        # Forward pass
        predictions = model(images)
        loss = loss_fn(predictions, masks)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        train_losses.append(loss.item())

    return np.mean(train_losses)
```

```python
from tqdm import tqdm

def train_model(model, train_loader, val_loader, loss_fn, optimizer, scheduler, config, device):
    """
    Main training function with early stopping and model logging to W&B.

    Args:
        model: UNet++ model
        train_loader: Training data loader
        val_loader: Validation data loader
        config: Wandb config with hyperparameters
        device: Device to run training on
    """

    # Early stopping parameters
    best_f1 = 0.0
    best_models = []  # Will store tuples: (val_f1, epoch, model_state_dict)
    TOP_K = 1
    run_name = wandb.run.name

    epochs_without_improvement = 0
    max_epochs = config.epochs
    early_stop_patience = 10
    early_stop_threshold = 0.60

    print(f"Starting training for {max_epochs} epochs...")
    print(f"Early stopping: F1 < {early_stop_threshold} for {early_stop_patience} epochs")

    model.to(device)

    for epoch in range(max_epochs):
        # Training
        train_loss = train_epoch(model, train_loader, optimizer, loss_fn, device, epoch)

        # Validation with sliding window
        val_metrics = validate_model(model, val_loader, loss_fn, device, patch_size=config.patch_size)

        #save cp model
        current_f1 = val_metrics['val_f1']
        if current_f1 > best_f1:
            best_f1 = current_f1
            best_model_path = f"{run_name}_epoch{epoch+1}_f1={best_f1:.4f}.pt"
            torch.save(model.state_dict(), best_model_path)

            # Log to W&B
            artifact = wandb.Artifact("best_model", type="model")
            artifact.add_file(best_model_path)
            wandb.log_artifact(artifact)

        # Update learning rate
        if scheduler is not None:
            scheduler.step(val_metrics['val_f1'])

        # Keep track of top K models
        if len(best_models) < TOP_K or val_metrics['val_f1'] > min(m[0] for m in best_models):
            model_cpu_state = {k: v.cpu() for k, v in model.state_dict().items()}
            best_models.append((val_metrics['val_f1'], epoch, model_cpu_state))
            best_models = sorted(best_models, key=lambda x: x[0], reverse=True)[:TOP_K]

        # Get current learning rate
        current_lr = optimizer.param_groups[0]['lr']

        # Log metrics
        wandb.log({
            'epoch': epoch+1,
            'train_loss': train_loss,
            'val_loss': val_metrics['val_loss'],
            'val_f1': val_metrics['val_f1'],
            'val_precision': val_metrics['val_precision'],
            'val_recall': val_metrics['val_recall'],
            'learning_rate': current_lr,
            'best_f1': best_f1
        })

        print(f"Train Loss: {train_loss:.4f} | Val Loss: {val_metrics['val_loss']:.4f} | "
              f"Val F1: {val_metrics['val_f1']:.4f} | Val Precision: {val_metrics['val_precision']:.4f} | "
              f"Val Recall: {val_metrics['val_recall']:.4f} | LR: {current_lr:.6f}")

        # Check if avg_val_f1 is below the threshold
        if val_metrics['val_f1'] <= early_stop_threshold:
            epochs_without_improvement += 1
            print(f"Epoch {epoch+1}: val_f1={val_metrics['val_f1']:.4f} < {early_stop_threshold}, counter={epochs_without_improvement}")
        else:
            epochs_without_improvement = 0  # reset if above threshold

        # Early stopping check
        if val_metrics['val_f1'] < early_stop_threshold:
            if epochs_without_improvement >= early_stop_patience:
                print(f"Early stopping: F1 ({val_metrics['val_f1']:.4f}) < {early_stop_threshold} "
                      f"for {early_stop_patience} epochs")
                break

    # # Save best models
    # for rank, (f1, epoch, state_dict) in enumerate(best_models, start=1):
    #     checkpoint_path = f"best_{run_name}_rank{rank}_epoch{epoch+1}_f1={f1:.4f}.pt"
    #     torch.save(state_dict, checkpoint_path)

    #     artifact = wandb.Artifact(f"model_rank{rank}", type="model")
    #     artifact.add_file(checkpoint_path)
    #     wandb.log_artifact(artifact)

    print(f"Training completed! Best F1: {best_f1:.4f}")
    print(f"Best model logged to W&B as artifact: best_model_{wandb.run.id}")

    return best_f1
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from pathlib import Path
import cv2
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix
import argparse
from tqdm import tqdm
import json
from typing import Dict, List, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')

class TissueSegEvaluator:
    def __init__(self, model_path: str, root_image: str, root_annotation: str,
                 patch_size: int = 512, overlap: int = 256, device: str = 'cuda',
                 use_training_style: bool = True):
        """
        Initialize the evaluator

        Args:
            model_path: Path to the trained model weights
            root_image: Root directory for images
            root_annotation: Root directory for annotations
            patch_size: Size of patches for sliding window inference
            overlap: Overlap between patches
            device: Device to run inference on
        """
        self.model_path = model_path
        self.root_image = Path(root_image)
        self.root_annotation = Path(root_annotation)
        self.patch_size = patch_size
        self.overlap = overlap
        self.device = device
        self.use_training_style = use_training_style

        # Set stride based on training style
        if use_training_style:
            self.stride = patch_size // 2  # Match training validation
        else:
            self.stride = patch_size - overlap  # Original implementation

        # Load model (you'll need to import your TissueSegModel and config)
        self.model = TissueSegModel("se_resnet50", dropout_rate=0.0)
        self.model.load_state_dict(torch.load(model_path, map_location=device))
        self.model.to(device)
        self.model.eval()

    def encode_ground_truth(self, gt_mask: np.ndarray) -> np.ndarray:
        """
        Encode ground truth mask: 1->0 (non-tumor), 2->1 (tumor), ignore 255

        Args:
            gt_mask: Raw ground truth mask

        Returns:
            Encoded binary mask and valid mask
        """
        encoded_mask = np.zeros_like(gt_mask, dtype=np.uint8)
        valid_mask = np.ones_like(gt_mask, dtype=bool)

        # 1 -> 0 (non-tumor)
        encoded_mask[gt_mask == 1] = 0
        # 2 -> 1 (tumor)
        encoded_mask[gt_mask == 2] = 1
        # 255 -> ignore
        valid_mask[gt_mask == 255] = False

        return encoded_mask, valid_mask

    # Create 2D Gaussian weight window (same size as patch)
    def get_gaussian_weight(self, patch_size, sigma_scale=2.0):
        sigma = patch_size * sigma_scale
        ax = np.linspace(-(patch_size - 1) / 2., (patch_size - 1) / 2., patch_size)
        xx, yy = np.meshgrid(ax, ax)
        weight = np.exp(-(xx**2 + yy**2) / (2. * sigma**2))
        weight /= np.max(weight)
        return weight.astype(np.float32)  # âœ… Ensure it's numpy

    def sliding_window_inference(self, image: np.ndarray) -> np.ndarray:
        """
        Perform sliding window inference on large image

        Args:
            image: Input image (H, W, 3)

        Returns:
            Prediction mask (H, W)
        """
        h, w = image.shape[:2]

        # Precompute the weight window (shape: (H, W))
        weight_window = self.get_gaussian_weight(self.patch_size)

        # Use the stride determined in __init__
        stride = self.stride

        # Calculate number of patches
        n_patches_h = (h - self.patch_size) // stride + 1 if h > self.patch_size else 1
        n_patches_w = (w - self.patch_size) // stride + 1 if w > self.patch_size else 1

        # Ensure we cover the entire image
        if h > self.patch_size and (n_patches_h - 1) * stride + self.patch_size < h:
            n_patches_h += 1
        if w > self.patch_size and (n_patches_w - 1) * stride + self.patch_size < w:
            n_patches_w += 1

        # Initialize prediction arrays
        prediction_sum = np.zeros((h, w), dtype=np.float32)
        count_map = np.zeros((h, w), dtype=np.float32)

        with torch.no_grad():
            for i in range(n_patches_h):
                for j in range(n_patches_w):
                    # Calculate patch coordinates
                    start_h = min(i * stride, max(0, h - self.patch_size))
                    start_w = min(j * stride, max(0, w - self.patch_size))
                    end_h = min(start_h + self.patch_size, h)
                    end_w = min(start_w + self.patch_size, w)

                    # Extract patch
                    patch = image[start_h:end_h, start_w:end_w]

                    # Pad if necessary
                    if patch.shape[0] < self.patch_size or patch.shape[1] < self.patch_size:
                        padded_patch = np.zeros((self.patch_size, self.patch_size, 3), dtype=patch.dtype)
                        padded_patch[:patch.shape[0], :patch.shape[1]] = patch
                        patch = padded_patch

                    # Prepare input tensor (match training preprocessing)
                    patch_tensor = torch.from_numpy(patch).permute(2, 0, 1).float()

                    # Apply same normalization as training if needed
                    if self.use_training_style:
                        # Assuming your training uses ImageNet normalization
                        # Adjust these values to match your training preprocessing
                        mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
                        std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
                        patch_tensor = patch_tensor / 255.0
                        patch_tensor = (patch_tensor - mean) / std
                    else:
                        patch_tensor = patch_tensor / 255.0

                    patch_tensor = patch_tensor.unsqueeze(0).to(self.device)

                    # Get prediction
                    pred = self.model(patch_tensor)

                    # Handle different output formats - match training validation
                    if pred.dim() == 4:  # (B, C, H, W)
                        pred = torch.sigmoid(pred)  # Apply sigmoid like in training
                        if pred.shape[1] > 1:
                            pred = pred[:, 1]  # Take tumor class
                        else:
                            pred = pred.squeeze(1)
                    else:
                        pred = torch.sigmoid(pred)  # Apply sigmoid

                    pred_np = pred.cpu().numpy().squeeze().astype(np.float32)

                    # Only use the valid part of the patch
                    valid_h = min(self.patch_size, end_h - start_h)
                    valid_w = min(self.patch_size, end_w - start_w)
                    pred_np = pred_np[:valid_h, :valid_w]

                    # Use corresponding region of Gaussian weight
                    weight_crop = weight_window[:valid_h, :valid_w]

                    # Accumulate weighted prediction
                    prediction_sum[start_h:start_h+valid_h, start_w:start_w+valid_w] += pred_np * weight_crop
                    count_map[start_h:start_h+valid_h, start_w:start_w+valid_w] += weight_crop


        # Average overlapping predictions
        final_prediction = prediction_sum / np.maximum(count_map, 1)

        return final_prediction

    def compute_metrics(self, y_true: np.ndarray, y_pred: np.ndarray,
                       valid_mask: np.ndarray) -> Dict[str, float]:
        """
        Compute evaluation metrics

        Args:
            y_true: Ground truth binary mask
            y_pred: Predicted binary mask
            valid_mask: Mask indicating valid pixels (ignore 255 regions)

        Returns:
            Dictionary containing all metrics
        """
        # Apply valid mask
        y_true_valid = y_true[valid_mask]
        y_pred_valid = y_pred[valid_mask]

        if len(y_true_valid) == 0:
            return {
                'precision': 0.0, 'recall': 0.0, 'f1': 0.0,
                'tn_over_fp_tn': 0.0, 'tn_over_fn_tn': 0.0,
                'tp': 0, 'tn': 0, 'fp': 0, 'fn': 0,
                'total_pixels': int(len(y_true.flatten())),
                'valid_pixels': 0
            }

        # Compute binary (tumor class) precision, recall, f1
        precision_bin, recall_bin, f1_bin, _ = precision_recall_fscore_support(
            y_true_valid, y_pred_valid, average='binary', zero_division=0
        )

        # Compute macro precision, recall, f1 (average over both classes equally)
        precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
            y_true_valid, y_pred_valid, average='macro', zero_division=0
        )


        # Compute confusion matrix
        cm = confusion_matrix(y_true_valid, y_pred_valid, labels=[0, 1])

        # Handle cases where confusion matrix might not be 2x2
        if cm.shape == (2, 2):
            tn, fp, fn, tp = cm.ravel()
        elif cm.shape == (1, 1):
            # Only one class present
            if len(np.unique(y_true_valid)) == 1 and len(np.unique(y_pred_valid)) == 1:
                if y_true_valid[0] == 0 and y_pred_valid[0] == 0:
                    tn, fp, fn, tp = len(y_true_valid), 0, 0, 0
                elif y_true_valid[0] == 1 and y_pred_valid[0] == 1:
                    tn, fp, fn, tp = 0, 0, 0, len(y_true_valid)
                else:
                    tn, fp, fn, tp = 0, 0, 0, 0
            else:
                tn, fp, fn, tp = 0, 0, 0, 0
        else:
            # Fallback: manually compute
            tp = np.sum((y_true_valid == 1) & (y_pred_valid == 1))
            tn = np.sum((y_true_valid == 0) & (y_pred_valid == 0))
            fp = np.sum((y_true_valid == 0) & (y_pred_valid == 1))
            fn = np.sum((y_true_valid == 1) & (y_pred_valid == 0))

        # Compute additional metrics
        tn_over_fp_tn = tn / (fp + tn) if (fp + tn) > 0 else 0.0
        tn_over_fn_tn = tn / (fn + tn) if (fn + tn) > 0 else 0.0

        return {
            'precision': float(precision_bin),
            'recall': float(recall_bin),
            'f1': float(f1_bin),
            'macro_precision': float(precision_macro),
            'macro_recall': float(recall_macro),
            'macro_f1': float(f1_macro),

            'tn_over_fp_tn': float(tn_over_fp_tn),  # Specificity
            'tn_over_fn_tn': float(tn_over_fn_tn),  # TN / (FN + TN)
            'tp': int(tp), 'tn': int(tn), 'fp': int(fp), 'fn': int(fn),
            'total_pixels': int(len(y_true.flatten())),
            'valid_pixels': int(len(y_true_valid))
        }

    def evaluate_dataset(self, split: str, threshold: float = 0.5) -> Tuple[Dict[str, float], List[Dict]]:
        """
        Evaluate model on a specific dataset split

        Args:
            split: Dataset split ('train', 'val', 'test')
            threshold: Threshold for converting probabilities to binary predictions

        Returns:
            Overall metrics and per-image metrics
        """
        image_dir = self.root_image / split / 'tissue'
        annotation_dir = self.root_annotation / split / 'tissue'

        if not image_dir.exists():
            raise ValueError(f"Image directory {image_dir} does not exist")
        if not annotation_dir.exists():
            raise ValueError(f"Annotation directory {annotation_dir} does not exist")

        image_files = list(image_dir.glob('*.png')) + list(image_dir.glob('*.jpg')) + list(image_dir.glob('*.tif'))

        all_metrics = []
        total_tp, total_tn, total_fp, total_fn = 0, 0, 0, 0
        total_valid_pixels = 0

        print(f"Evaluating {split} set: {len(image_files)} images")

        for img_path in tqdm(image_files):
            # Find corresponding annotation
            ann_path = None
            for ext in ['.png', '.jpg', '.tif']:
                potential_ann = annotation_dir / (img_path.stem + ext)
                if potential_ann.exists():
                    ann_path = potential_ann
                    break

            if ann_path is None:
                print(f"Warning: No annotation found for {img_path.name}")
                continue

            # Load image and annotation
            image = cv2.imread(str(img_path))
            if image is None:
                print(f"Warning: Could not load image {img_path}")
                continue
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            annotation = cv2.imread(str(ann_path), cv2.IMREAD_GRAYSCALE)
            if annotation is None:
                print(f"Warning: Could not load annotation {ann_path}")
                continue

            # Encode ground truth
            gt_encoded, valid_mask = self.encode_ground_truth(annotation)

            # Perform inference
            pred_prob = self.sliding_window_inference(image)
            pred_binary = (pred_prob > threshold).astype(np.uint8)

            # Compute metrics for this image
            img_metrics = self.compute_metrics(gt_encoded, pred_binary, valid_mask)
            img_metrics['image_name'] = img_path.name
            img_metrics['threshold'] = threshold
            all_metrics.append(img_metrics)

            # Accumulate for overall metrics
            total_tp += img_metrics.get('tp', 0)
            total_tn += img_metrics.get('tn', 0)
            total_fp += img_metrics.get('fp', 0)
            total_fn += img_metrics.get('fn', 0)
            total_valid_pixels += img_metrics.get('valid_pixels', 0)

        # Compute overall metrics
        overall_precision = total_tp / (total_tp + total_fp) if (total_tp + total_fp) > 0 else 0.0
        overall_recall = total_tp / (total_tp + total_fn) if (total_tp + total_fn) > 0 else 0.0
        overall_f1 = 2 * (overall_precision * overall_recall) / (overall_precision + overall_recall) if (overall_precision + overall_recall) > 0 else 0.0
        overall_tn_over_fp_tn = total_tn / (total_fp + total_tn) if (total_fp + total_tn) > 0 else 0.0
        overall_tn_over_fn_tn = total_tn / (total_fn + total_tn) if (total_fn + total_tn) > 0 else 0.0

        overall_metrics = {
            'split': split,
            'num_images': len(all_metrics),
            'total_valid_pixels': total_valid_pixels,
            'precision': overall_precision,
            'recall': overall_recall,
            'f1': overall_f1,
            'tn_over_fp_tn': overall_tn_over_fp_tn,
            'tn_over_fn_tn': overall_tn_over_fn_tn,
            'tp': total_tp, 'tn': total_tn, 'fp': total_fp, 'fn': total_fn,
            'threshold': threshold
        }

        return overall_metrics, all_metrics

    def run_evaluation(self, thresholds: List[float] = [0.5],
                      splits: List[str] = ['train','val','test'],
                      save_results: bool = True,
                      output_dir: str = 'evaluation_results') -> Dict:
        """
        Run complete evaluation across all splits and thresholds

        Args:
            thresholds: List of thresholds to evaluate
            splits: List of dataset splits to evaluate
            save_results: Whether to save results to files
            output_dir: Directory to save results

        Returns:
            Complete evaluation results
        """
        results = {'overall': {}, 'per_image': {}, 'thresholds': thresholds}

        if save_results:
            Path(output_dir).mkdir(exist_ok=True)

        for split in splits:
            results['overall'][split] = {}
            results['per_image'][split] = {}

            for threshold in thresholds:
                print(f"\nEvaluating {split} set with threshold {threshold}")

                try:
                    overall_metrics, per_image_metrics = self.evaluate_dataset(split, threshold)

                    results['overall'][split][f'threshold_{threshold}'] = overall_metrics
                    results['per_image'][split][f'threshold_{threshold}'] = per_image_metrics

                    # Print summary
                    print(f"Results for {split} (threshold={threshold}):")
                    print(f"  F1 Score: {overall_metrics['f1']:.4f}")
                    print(f"  Macro F1 Score: {overall_metrics.get('macro_f1', 0.0):.4f}")
                    print(f"  Precision: {overall_metrics['precision']:.4f}")
                    print(f"  Recall: {overall_metrics['recall']:.4f}")
                    print(f"  TN/(FP+TN): {overall_metrics['tn_over_fp_tn']:.4f}")
                    print(f"  TN/(FN+TN): {overall_metrics['tn_over_fn_tn']:.4f}")
                    print(f"  Images: {overall_metrics['num_images']}")


                except Exception as e:
                    print(f"Error evaluating {split} with threshold {threshold}: {e}")
                    continue

        # Save results
        if save_results:
            with open(Path(output_dir) / 'evaluation_results.json', 'w') as f:
                json.dump(results, f, indent=2)
            print(f"\nResults saved to {output_dir}/evaluation_results.json")

        return results

# Example usage function for Colab/Jupyter
def run_evaluation_example():
    """
    Example function to run evaluation - modify these paths for your setup
    """
    # TODO: Update these paths for your specific setup
    MODEL_PATH = "/path/to/your/model.pth"
    ROOT_IMAGE = "/path/to/root_image"
    ROOT_ANNOTATION = "/path/to/root_annotation"

    # Initialize evaluator
    evaluator = TissueSegEvaluator(
        model_path=MODEL_PATH,
        root_image=ROOT_IMAGE,
        root_annotation=ROOT_ANNOTATION,
        patch_size=512,
        overlap=64,
        device='cuda' if torch.cuda.is_available() else 'cpu'
    )

    # Run evaluation
    results = evaluator.run_evaluation(
        thresholds=[0.3, 0.5, 0.7],
        splits=['train', 'val', 'test'],
        save_results=True,
        output_dir='evaluation_results'
    )

    return results

def main():
    """
    Main function that can be called from command line or modified for notebook use
    """
    import sys

    # Check if running in notebook environment
    if 'ipykernel' in sys.modules or 'google.colab' in sys.modules:
        print("Detected notebook environment. Please use the TissueSegEvaluator class directly.")
        print("Example:")
        print("""
# Initialize evaluator to match training validation
evaluator = TissueSegEvaluator(
    model_path="your_model_path.pth",
    root_image="path/to/root_image",
    root_annotation="path/to/root_annotation",
    patch_size=1024,  # Match training patch size
    use_training_style=True  # Use training-style sliding window
)

# Run evaluation
results = evaluator.run_evaluation()
        """)
        return

    # Command line argument parsing for standalone script
    parser = argparse.ArgumentParser(description='Evaluate Tissue Segmentation Model')
    parser.add_argument('--model_path', type=str, required=True, help='Path to model weights')
    parser.add_argument('--root_image', type=str, required=True, help='Root directory for images')
    parser.add_argument('--root_annotation', type=str, required=True, help='Root directory for annotations')
    parser.add_argument('--patch_size', type=int, default=512, help='Patch size for inference')
    parser.add_argument('--overlap', type=int, default=64, help='Overlap between patches')
    parser.add_argument('--device', type=str, default='cuda', help='Device for inference')
    parser.add_argument('--thresholds', nargs='+', type=float, default=[0.5], help='Thresholds for evaluation')
    parser.add_argument('--splits', nargs='+', type=str, default=['train', 'val', 'test'], help='Dataset splits to evaluate')
    parser.add_argument('--output_dir', type=str, default='evaluation_results', help='Output directory for results')

    args = parser.parse_args()

    # Initialize evaluator
    evaluator = TissueSegEvaluator(
        model_path=args.model_path,
        root_image=args.root_image,
        root_annotation=args.root_annotation,
        patch_size=args.patch_size,
        overlap=args.overlap,
        device=args.device
    )

    # Run evaluation
    results = evaluator.run_evaluation(
        thresholds=args.thresholds,
        splits=args.splits,
        save_results=True,
        output_dir=args.output_dir
    )

    print("\nEvaluation completed!")

if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from pathlib import Path
import cv2
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix
import argparse
from tqdm import tqdm
import json
from typing import Dict, List, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')

class TissueSegEvaluator:
    def __init__(self, model_path: str, root_image: str, root_annotation: str,
                 patch_size: int = 512, overlap: int = 256, device: str = 'cuda',
                 use_training_style: bool = True):
        """
        Initialize the evaluator

        Args:
            model_path: Path to the trained model weights
            root_image: Root directory for images
            root_annotation: Root directory for annotations
            patch_size: Size of patches for sliding window inference
            overlap: Overlap between patches
            device: Device to run inference on
        """
        self.model_path = model_path
        self.root_image = Path(root_image)
        self.root_annotation = Path(root_annotation)
        self.patch_size = patch_size
        self.overlap = overlap
        self.device = device
        self.use_training_style = use_training_style

        # Set stride based on training style
        if use_training_style:
            self.stride = patch_size // 2  # Match training validation
        else:
            self.stride = patch_size - overlap  # Original implementation

        # Load model (you'll need to import your TissueSegModel and config)
        self.model = TissueSegModel("se_resnet50", dropout_rate=0.0)
        self.model.load_state_dict(torch.load(model_path, map_location=device))
        self.model.to(device)
        self.model.eval()

    def encode_ground_truth(self, gt_mask: np.ndarray) -> np.ndarray:
        """
        Encode ground truth mask: 1->0 (non-tumor), 2->1 (tumor), ignore 255

        Args:
            gt_mask: Raw ground truth mask

        Returns:
            Encoded binary mask and valid mask
        """
        encoded_mask = np.zeros_like(gt_mask, dtype=np.uint8)
        valid_mask = np.ones_like(gt_mask, dtype=bool)

        # 1 -> 0 (non-tumor)
        encoded_mask[gt_mask == 1] = 0
        # 2 -> 1 (tumor)
        encoded_mask[gt_mask == 2] = 1
        # 255 -> ignore
        valid_mask[gt_mask == 255] = False

        return encoded_mask, valid_mask

    # Create 2D Gaussian weight window (same size as patch)
    def get_gaussian_weight(self, patch_size, sigma_scale=2.0):
        sigma = patch_size * sigma_scale
        ax = np.linspace(-(patch_size - 1) / 2., (patch_size - 1) / 2., patch_size)
        xx, yy = np.meshgrid(ax, ax)
        weight = np.exp(-(xx**2 + yy**2) / (2. * sigma**2))
        weight /= np.max(weight)
        return weight.astype(np.float32)  # âœ… Ensure it's numpy

    def sliding_window_inference(self, image: np.ndarray) -> np.ndarray:
        """
        Perform sliding window inference on large image

        Args:
            image: Input image (H, W, 3)

        Returns:
            Prediction mask (H, W)
        """
        h, w = image.shape[:2]

        # Precompute the weight window (shape: (H, W))
        weight_window = self.get_gaussian_weight(self.patch_size)

        # Use the stride determined in __init__
        stride = self.stride

        # Calculate number of patches
        n_patches_h = (h - self.patch_size) // stride + 1 if h > self.patch_size else 1
        n_patches_w = (w - self.patch_size) // stride + 1 if w > self.patch_size else 1

        # Ensure we cover the entire image
        if h > self.patch_size and (n_patches_h - 1) * stride + self.patch_size < h:
            n_patches_h += 1
        if w > self.patch_size and (n_patches_w - 1) * stride + self.patch_size < w:
            n_patches_w += 1

        # Initialize prediction arrays
        prediction_sum = np.zeros((h, w), dtype=np.float32)
        count_map = np.zeros((h, w), dtype=np.float32)

        with torch.no_grad():
            for i in range(n_patches_h):
                for j in range(n_patches_w):
                    # Calculate patch coordinates
                    start_h = min(i * stride, max(0, h - self.patch_size))
                    start_w = min(j * stride, max(0, w - self.patch_size))
                    end_h = min(start_h + self.patch_size, h)
                    end_w = min(start_w + self.patch_size, w)

                    # Extract patch
                    patch = image[start_h:end_h, start_w:end_w]

                    # Pad if necessary
                    if patch.shape[0] < self.patch_size or patch.shape[1] < self.patch_size:
                        padded_patch = np.zeros((self.patch_size, self.patch_size, 3), dtype=patch.dtype)
                        padded_patch[:patch.shape[0], :patch.shape[1]] = patch
                        patch = padded_patch

                    # Prepare input tensor (match training preprocessing)
                    patch_tensor = torch.from_numpy(patch).permute(2, 0, 1).float()

                    # Apply same normalization as training if needed
                    if self.use_training_style:
                        # Assuming your training uses ImageNet normalization
                        # Adjust these values to match your training preprocessing
                        mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
                        std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
                        patch_tensor = patch_tensor / 255.0
                        patch_tensor = (patch_tensor - mean) / std
                    else:
                        patch_tensor = patch_tensor / 255.0

                    patch_tensor = patch_tensor.unsqueeze(0).to(self.device)

                    # Get prediction
                    pred = self.model(patch_tensor)

                    # Handle different output formats - match training validation
                    if pred.dim() == 4:  # (B, C, H, W)
                        pred = torch.sigmoid(pred)  # Apply sigmoid like in training
                        if pred.shape[1] > 1:
                            pred = pred[:, 1]  # Take tumor class
                        else:
                            pred = pred.squeeze(1)
                    else:
                        pred = torch.sigmoid(pred)  # Apply sigmoid

                    pred_np = pred.cpu().numpy().squeeze().astype(np.float32)

                    # Only use the valid part of the patch
                    valid_h = min(self.patch_size, end_h - start_h)
                    valid_w = min(self.patch_size, end_w - start_w)
                    pred_np = pred_np[:valid_h, :valid_w]

                    # Use corresponding region of Gaussian weight
                    weight_crop = weight_window[:valid_h, :valid_w]

                    # Accumulate weighted prediction
                    prediction_sum[start_h:start_h+valid_h, start_w:start_w+valid_w] += pred_np * weight_crop
                    count_map[start_h:start_h+valid_h, start_w:start_w+valid_w] += weight_crop


        # Average overlapping predictions
        final_prediction = prediction_sum / np.maximum(count_map, 1)

        return final_prediction
    def compute_metrics(self, y_true: np.ndarray, y_pred: np.ndarray,
                        valid_mask: np.ndarray) -> Dict[str, float]:
        """
        Compute evaluation metrics

        Args:
            y_true: Ground truth binary mask
            y_pred: Predicted binary mask
            valid_mask: Mask indicating valid pixels (ignore 255 regions)

        Returns:
            Dictionary containing all metrics and pixel arrays for global metrics
        """
        # Apply valid mask
        y_true_valid = y_true[valid_mask]
        y_pred_valid = y_pred[valid_mask]

        if len(y_true_valid) == 0:
            return {
                'precision': 0.0, 'recall': 0.0, 'f1': 0.0,
                'macro_precision': 0.0, 'macro_recall': 0.0, 'macro_f1': 0.0,
                'tn_over_fp_tn': 0.0, 'tn_over_fn_tn': 0.0,
                'tp': 0, 'tn': 0, 'fp': 0, 'fn': 0,
                'total_pixels': int(len(y_true.flatten())),
                'valid_pixels': 0,
                'y_true_valid': y_true_valid,
                'y_pred_valid': y_pred_valid
            }

        # Binary metrics (tumor class = 1)
        precision_bin, recall_bin, f1_bin, _ = precision_recall_fscore_support(
            y_true_valid, y_pred_valid, average='binary', zero_division=0
        )

        # Macro metrics (averaged over both classes 0 and 1)
        precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
            y_true_valid, y_pred_valid, average='macro', zero_division=0
        )

        # Confusion matrix
        cm = confusion_matrix(y_true_valid, y_pred_valid, labels=[0, 1])
        if cm.shape == (2, 2):
            tn, fp, fn, tp = cm.ravel()
        else:
            tp = np.sum((y_true_valid == 1) & (y_pred_valid == 1))
            tn = np.sum((y_true_valid == 0) & (y_pred_valid == 0))
            fp = np.sum((y_true_valid == 0) & (y_pred_valid == 1))
            fn = np.sum((y_true_valid == 1) & (y_pred_valid == 0))

        # Additional metrics
        tn_over_fp_tn = tn / (fp + tn) if (fp + tn) > 0 else 0.0
        tn_over_fn_tn = tn / (fn + tn) if (fn + tn) > 0 else 0.0

        return {
            'precision': float(precision_bin),
            'recall': float(recall_bin),
            'f1': float(f1_bin),
            'macro_precision': float(precision_macro),
            'macro_recall': float(recall_macro),
            'macro_f1': float(f1_macro),
            'tn_over_fp_tn': float(tn_over_fp_tn),
            'tn_over_fn_tn': float(tn_over_fn_tn),
            'tp': int(tp), 'tn': int(tn), 'fp': int(fp), 'fn': int(fn),
            'total_pixels': int(len(y_true.flatten())),
            'valid_pixels': int(len(y_true_valid)),
            'y_true_valid': y_true_valid,      # âœ… Needed for dataset-wide macro F1
            'y_pred_valid': y_pred_valid       # âœ… Needed for dataset-wide macro F1
        }

    def evaluate_dataset(self, split: str, threshold: float = 0.5) -> Tuple[Dict[str, float], List[Dict]]:
        """
        Evaluate model on a specific dataset split

        Args:
            split: Dataset split ('train', 'val', 'test')
            threshold: Threshold for converting probabilities to binary predictions

        Returns:
            Overall metrics and per-image metrics
        """
        image_dir = self.root_image / split / 'tissue'
        annotation_dir = self.root_annotation / split / 'tissue'

        if not image_dir.exists():
            raise ValueError(f"Image directory {image_dir} does not exist")
        if not annotation_dir.exists():
            raise ValueError(f"Annotation directory {annotation_dir} does not exist")

        image_files = list(image_dir.glob('*.png')) + list(image_dir.glob('*.jpg')) + list(image_dir.glob('*.tif'))

        all_metrics = []
        total_tp, total_tn, total_fp, total_fn = 0, 0, 0, 0
        total_valid_pixels = 0

        # âœ… For dataset-level macro F1
        global_y_true = []
        global_y_pred = []

        print(f"Evaluating {split} set: {len(image_files)} images")

        for img_path in tqdm(image_files):
            # Find corresponding annotation
            ann_path = None
            for ext in ['.png', '.jpg', '.tif']:
                potential_ann = annotation_dir / (img_path.stem + ext)
                if potential_ann.exists():
                    ann_path = potential_ann
                    break

            if ann_path is None:
                print(f"Warning: No annotation found for {img_path.name}")
                continue

            # Load image and annotation
            image = cv2.imread(str(img_path))
            if image is None:
                print(f"Warning: Could not load image {img_path}")
                continue
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            annotation = cv2.imread(str(ann_path), cv2.IMREAD_GRAYSCALE)
            if annotation is None:
                print(f"Warning: Could not load annotation {ann_path}")
                continue

            # Encode ground truth
            gt_encoded, valid_mask = self.encode_ground_truth(annotation)

            # Perform inference
            pred_prob = self.sliding_window_inference(image)
            pred_binary = (pred_prob > threshold).astype(np.uint8)

            # Compute metrics for this image
            img_metrics = self.compute_metrics(gt_encoded, pred_binary, valid_mask)
            img_metrics['image_name'] = img_path.name
            img_metrics['threshold'] = threshold
            all_metrics.append(img_metrics)

            # Accumulate for dataset-wide metrics
            total_tp += img_metrics.get('tp', 0)
            total_tn += img_metrics.get('tn', 0)
            total_fp += img_metrics.get('fp', 0)
            total_fn += img_metrics.get('fn', 0)
            total_valid_pixels += img_metrics.get('valid_pixels', 0)

            # âœ… Accumulate per-pixel labels for macro F1
            global_y_true.append(img_metrics['y_true_valid'])
            global_y_pred.append(img_metrics['y_pred_valid'])

        # âœ… Dataset-wide macro F1 (over all valid pixels across images)
        if len(global_y_true) > 0:
            global_y_true = np.concatenate(global_y_true)
            global_y_pred = np.concatenate(global_y_pred)

            precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
                global_y_true, global_y_pred, average='macro', zero_division=0
            )
        else:
            precision_macro, recall_macro, f1_macro = 0.0, 0.0, 0.0

        # Binary overall (tumor class)
        overall_precision = total_tp / (total_tp + total_fp) if (total_tp + total_fp) > 0 else 0.0
        overall_recall = total_tp / (total_tp + total_fn) if (total_tp + total_fn) > 0 else 0.0
        overall_f1 = 2 * (overall_precision * overall_recall) / (overall_precision + overall_recall) if (overall_precision + overall_recall) > 0 else 0.0
        overall_tn_over_fp_tn = total_tn / (total_fp + total_tn) if (total_fp + total_tn) > 0 else 0.0
        overall_tn_over_fn_tn = total_tn / (total_fn + total_tn) if (total_fn + total_tn) > 0 else 0.0

        overall_metrics = {
            'split': split,
            'num_images': len(all_metrics),
            'total_valid_pixels': total_valid_pixels,
            'precision': overall_precision,
            'recall': overall_recall,
            'f1': overall_f1,
            'tn_over_fp_tn': overall_tn_over_fp_tn,
            'tn_over_fn_tn': overall_tn_over_fn_tn,
            'tp': total_tp, 'tn': total_tn, 'fp': total_fp, 'fn': total_fn,
            'threshold': threshold,
            'macro_precision': float(precision_macro),
            'macro_recall': float(recall_macro),
            'macro_f1': float(f1_macro)
        }

        return overall_metrics, all_metrics

    def run_evaluation(self, thresholds: List[float] = [0.5],
                      splits: List[str] = ['train','val','test'],
                      save_results: bool = True,
                      output_dir: str = 'evaluation_results') -> Dict:
        """
        Run complete evaluation across all splits and thresholds

        Args:
            thresholds: List of thresholds to evaluate
            splits: List of dataset splits to evaluate
            save_results: Whether to save results to files
            output_dir: Directory to save results

        Returns:
            Complete evaluation results
        """
        results = {'overall': {}, 'per_image': {}, 'thresholds': thresholds}

        if save_results:
            Path(output_dir).mkdir(exist_ok=True)

        for split in splits:
            results['overall'][split] = {}
            results['per_image'][split] = {}

            for threshold in thresholds:
                print(f"\nEvaluating {split} set with threshold {threshold}")

                try:
                    overall_metrics, per_image_metrics = self.evaluate_dataset(split, threshold)

                    results['overall'][split][f'threshold_{threshold}'] = overall_metrics
                    results['per_image'][split][f'threshold_{threshold}'] = per_image_metrics

                    # Print summary
                    print(f"Results for {split} (threshold={threshold}):")
                    print(f"  F1 Score: {overall_metrics['f1']:.4f}")
                    print(f"  Macro F1 Score: {overall_metrics.get('macro_f1', 0.0):.4f}")
                    print(f"  Precision: {overall_metrics['precision']:.4f}")
                    print(f"  Recall: {overall_metrics['recall']:.4f}")
                    print(f"  TN/(FP+TN): {overall_metrics['tn_over_fp_tn']:.4f}")
                    print(f"  TN/(FN+TN): {overall_metrics['tn_over_fn_tn']:.4f}")
                    print(f"  Images: {overall_metrics['num_images']}")


                except Exception as e:
                    print(f"Error evaluating {split} with threshold {threshold}: {e}")
                    continue

        # Save results
        if save_results:
            with open(Path(output_dir) / 'evaluation_results.json', 'w') as f:
                json.dump(results, f, indent=2)
            print(f"\nResults saved to {output_dir}/evaluation_results.json")

        return results

# Example usage function for Colab/Jupyter
def run_evaluation_example():
    """
    Example function to run evaluation - modify these paths for your setup
    """
    # TODO: Update these paths for your specific setup
    MODEL_PATH = "/path/to/your/model.pth"
    ROOT_IMAGE = "/path/to/root_image"
    ROOT_ANNOTATION = "/path/to/root_annotation"

    # Initialize evaluator
    evaluator = TissueSegEvaluator(
        model_path=MODEL_PATH,
        root_image=ROOT_IMAGE,
        root_annotation=ROOT_ANNOTATION,
        patch_size=512,
        overlap=64,
        device='cuda' if torch.cuda.is_available() else 'cpu'
    )

    # Run evaluation
    results = evaluator.run_evaluation(
        thresholds=[0.3, 0.5, 0.7],
        splits=['train', 'val', 'test'],
        save_results=True,
        output_dir='evaluation_results'
    )

    return results

def main():
    """
    Main function that can be called from command line or modified for notebook use
    """
    import sys

    # Check if running in notebook environment
    if 'ipykernel' in sys.modules or 'google.colab' in sys.modules:
        print("Detected notebook environment. Please use the TissueSegEvaluator class directly.")
        print("Example:")
        print("""
# Initialize evaluator to match training validation
evaluator = TissueSegEvaluator(
    model_path="your_model_path.pth",
    root_image="path/to/root_image",
    root_annotation="path/to/root_annotation",
    patch_size=1024,  # Match training patch size
    use_training_style=True  # Use training-style sliding window
)

# Run evaluation
results = evaluator.run_evaluation()
        """)
        return

    # Command line argument parsing for standalone script
    parser = argparse.ArgumentParser(description='Evaluate Tissue Segmentation Model')
    parser.add_argument('--model_path', type=str, required=True, help='Path to model weights')
    parser.add_argument('--root_image', type=str, required=True, help='Root directory for images')
    parser.add_argument('--root_annotation', type=str, required=True, help='Root directory for annotations')
    parser.add_argument('--patch_size', type=int, default=512, help='Patch size for inference')
    parser.add_argument('--overlap', type=int, default=64, help='Overlap between patches')
    parser.add_argument('--device', type=str, default='cuda', help='Device for inference')
    parser.add_argument('--thresholds', nargs='+', type=float, default=[0.5], help='Thresholds for evaluation')
    parser.add_argument('--splits', nargs='+', type=str, default=['train', 'val', 'test'], help='Dataset splits to evaluate')
    parser.add_argument('--output_dir', type=str, default='evaluation_results', help='Output directory for results')

    args = parser.parse_args()

    # Initialize evaluator
    evaluator = TissueSegEvaluator(
        model_path=args.model_path,
        root_image=args.root_image,
        root_annotation=args.root_annotation,
        patch_size=args.patch_size,
        overlap=args.overlap,
        device=args.device
    )

    # Run evaluation
    results = evaluator.run_evaluation(
        thresholds=args.thresholds,
        splits=args.splits,
        save_results=True,
        output_dir=args.output_dir
    )

    print("\nEvaluation completed!")

if __name__ == "__main__":
    main()
```

```python
import os
import random
import numpy as np
import torch
import torch.nn.functional as F
import cv2
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

# --- CONFIG ---
image_dir = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images/val/tissue"
mask_dir = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/val/tissue"
patch_size = 512
stride = 256
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- MODEL LOADING ---
model = TissueSegModel("se_resnet50", dropout_rate=0.0)
ckpt_path = "/content/drive/MyDrive/research/ocelot/Method/tissue segmentation/UNet++(se_resnet50)/eager-sweep-1_epoch146_f1=0.8916.pt"
model.load_state_dict(torch.load(ckpt_path, map_location=device))
model.to(device)
model.eval()

def sliding_window_inference(image_tensor, patch_size=512, stride=256):
    """
    Perform sliding window inference on image_tensor (C, H, W) with uniform weights.

    Args:
        image_tensor (torch.Tensor): Input tensor with shape (C, H, W)
        patch_size (int): Size of each square patch
        stride (int): Stride between patches

    Returns:
        np.ndarray: Final prediction map (H, W)
    """
    _, H, W = image_tensor.shape

    # Initialize output and count maps
    prob_map = torch.zeros((1, H, W), device=device)
    count_map = torch.zeros((1, H, W), device=device)

    # Sliding window loop
    for y in range(0, H - patch_size + 1, stride):
        for x in range(0, W - patch_size + 1, stride):
            patch = image_tensor[:, y:y+patch_size, x:x+patch_size].unsqueeze(0).to(device)

            with torch.no_grad():
                logits = model(patch)
                probs = torch.sigmoid(logits).squeeze(0)  # shape: (1, patch_size, patch_size)

            # Accumulate predictions
            prob_map[:, y:y+patch_size, x:x+patch_size] += probs
            count_map[:, y:y+patch_size, x:x+patch_size] += 1

    # Normalize overlapping regions
    avg_prob_map = prob_map / count_map.clamp(min=1e-6)

    return avg_prob_map.squeeze().cpu().numpy()

# --- MASK ENCODING ---
def encode_mask(mask):
    mask = np.array(mask)
    encoded = np.zeros_like(mask, dtype=np.uint8)
    encoded[mask == 2] = 1
    encoded[mask == 255] = 255  # will be ignored, but shown for visualization
    return encoded

# --- VISUALIZATION ---
def visualize(image_np, gt_mask, prob_map, pred_binary, title):
    # Masking ignore (255) for display
    gt_mask_vis = np.where(gt_mask == 255, 0.5, gt_mask).astype(np.float32)

    fig, axs = plt.subplots(1, 4, figsize=(20, 5))
    axs[0].imshow(image_np)
    axs[0].set_title('Original Image')
    axs[1].imshow(gt_mask_vis, cmap='gray', vmin=0, vmax=1)
    axs[1].set_title('Encoded GT Mask (255 dimmed)')
    axs[2].imshow(prob_map, cmap='viridis', vmin=0, vmax=1)
    axs[2].set_title('Predicted Probabilities')
    axs[3].imshow(pred_binary, cmap='gray')
    axs[3].set_title('Binary Prediction')
    for ax in axs:
        ax.axis('off')
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()


# --- MAIN LOOP ---
image_files = sorted(os.listdir(image_dir))
selected_files = random.sample(image_files, 3)

preprocess = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])


for fname in selected_files:
    image_path = os.path.join(image_dir, fname)
    mask_path = os.path.join(mask_dir, fname.replace(".jpg", ".png"))

    # Load image and mask
    image = Image.open(image_path).convert("RGB")
    image_tensor = preprocess(image)  # shape (3, 1024, 1024)

    mask = Image.open(mask_path)
    gt_encoded = encode_mask(mask)

    # Run inference
    prob_map = sliding_window_inference(image_tensor)
    binary_pred = (prob_map >= 0.5).astype(np.uint8)

    # Visualize
    visualize(np.array(image), gt_encoded, prob_map, binary_pred, title=f"Sample: {fname}")
```

```python
import os
import random
import numpy as np
import cv2
import matplotlib.pyplot as plt

# --- CONFIG ---
base_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1"
output_patch_dir = "/content/drive/MyDrive/research/ocelot/dataset/integration/tissuesoftmax_unet++(se_resnet50_acc0.8915)"  # .npy dir
split = "val"
target_size = 1024
crop_size = 256

# --- Metadata ---
import json
with open(f"{base_path}/metadata.json", "r") as f:
    metadata = json.load(f)["sample_pairs"]

val_keys = [k for k, v in metadata.items() if v["subset"] == "val"]
random_keys = random.sample(val_keys, 3)

# --- Visualization ---
def show_images(images, titles, cmap="viridis", vmin=0, vmax=1):
    n = len(images)
    plt.figure(figsize=(4 * n, 4))
    for i in range(n):
        plt.subplot(1, n, i + 1)
        img = images[i]
        if img.ndim == 3:
            plt.imshow(img)
        else:
            plt.imshow(img, cmap=cmap, vmin=vmin, vmax=vmax)
        plt.title(titles[i])
        plt.axis("off")
    plt.tight_layout()
    plt.show()

# --- Metric Computation ---
def compute_metrics(pred, gt):
    tp = np.logical_and(pred == 1, gt == 1).sum()
    fp = np.logical_and(pred == 1, gt == 0).sum()
    fn = np.logical_and(pred == 0, gt == 1).sum()
    tn = np.logical_and(pred == 0, gt == 0).sum()

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    recall    = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1        = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0
    acc       = (tp + tn) / (tp + tn + fp + fn) if (tp + tn + fp + fn) > 0 else 0.0

    return precision, recall, f1, acc, tp, fp, fn, tn

# --- Main Loop ---
for key in random_keys:
    print(f"\n=== Evaluating: {key} ===")

    npy_path = os.path.join(output_patch_dir, split, f"{key}.npy")
    if not os.path.exists(npy_path):
        print(f"Missing: {npy_path}")
        continue

    prob_1024 = np.load(npy_path)
    prob_256 = cv2.resize(prob_1024, (256, 256), interpolation=cv2.INTER_AREA)
    binary_256 = (prob_256 > 0.5).astype(np.uint8)

    # Load GT
    mask_path = os.path.join(base_path, "annotations", split, "tissue", f"{key}.png")
    mask_raw = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    mask_raw = np.where(mask_raw == 255, 0, mask_raw)  # ignore regions to background
    mask_remap = np.where(mask_raw == 2, 1, 0)         # cancer=1, non-cancer=0

    # Crop GT based on patch location
    sample = metadata[key]
    tissue_info = sample["tissue"]
    cell_info = sample["cell"]
    patch_x_offset = sample["patch_x_offset"]
    patch_y_offset = sample["patch_y_offset"]

    # Calculate crop dimensions
    tissue_w = tissue_info["x_end"] - tissue_info["x_start"]
    tissue_h = tissue_info["y_end"] - tissue_info["y_start"]
    cell_w = cell_info["x_end"] - cell_info["x_start"]
    cell_h = cell_info["y_end"] - cell_info["y_start"]
    ratio_x = cell_w / tissue_w
    ratio_y = cell_h / tissue_h
    crop_w = int(round(target_size * ratio_x))
    crop_h = int(round(target_size * ratio_y))

    center_x = int(round(target_size * patch_x_offset))
    center_y = int(round(target_size * patch_y_offset))
    left = max(0, center_x - crop_w // 2)
    top = max(0, center_y - crop_h // 2)
    right = min(left + crop_w, target_size)
    bottom = min(top + crop_h, target_size)

    gt_crop = mask_remap[top:bottom, left:right]
    if gt_crop.shape != binary_256.shape:
        gt_crop = cv2.resize(gt_crop, (256, 256), interpolation=cv2.INTER_NEAREST)

    # Load cell image
    cell_path = os.path.join(base_path, "images", split, "cell", f"{key}.jpg")
    cell_img = cv2.imread(cell_path)
    cell_img = cv2.cvtColor(cell_img, cv2.COLOR_BGR2RGB)
    cell_img = cv2.resize(cell_img, (256, 256))

    # Compute metrics
    precision, recall, f1, acc, tp, fp, fn, tn = compute_metrics(binary_256.astype(bool), gt_crop.astype(bool))

    # Display range + dtype
    print(f"  Prob 1024: shape={prob_1024.shape}, dtype={prob_1024.dtype}, range=({prob_1024.min():.3f}, {prob_1024.max():.3f})")
    print(f"  Prob 256 : shape={prob_256.shape}, dtype={prob_256.dtype}, range=({prob_256.min():.3f}, {prob_256.max():.3f})")
    print(f"  Binary   : unique={np.unique(binary_256)}")
    print(f"  GT       : unique={np.unique(gt_crop)}")
    print(f"  Metrics  : Precision={precision:.4f}, Recall={recall:.4f}, F1={f1:.4f}, accuracy={acc:.4f}, TP={tp}, FP={fp}, FN={fn}, TN={tn}")

    # Display
    show_images(
        [cell_img, prob_1024, prob_256, binary_256.astype(np.float32), gt_crop.astype(np.float32)],
        ["Cell Patch", "Prob Map 1024", "Resized 256", "Binary 256", "GT"],
        cmap="viridis",
        vmin=0,
        vmax=1
    )

```
