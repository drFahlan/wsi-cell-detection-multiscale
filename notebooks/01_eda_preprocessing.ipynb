```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Base paths
base_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1"
sets = ["train", "val", "test"]  # Different dataset splits

# Function to find matching annotation pairs in a given subset
def find_matching_annotations(split):
    image_dir = os.path.join(base_path, f"images/{split}/tissue")
    annotation_dir = os.path.join(base_path, f"annotations/{split}/tissue")

    # Get filenames without extensions
    image_files = {os.path.splitext(f)[0]: f for f in os.listdir(image_dir) if f.endswith(('.png', '.jpg', '.jpeg'))}
    annotation_files = {os.path.splitext(f)[0]: f for f in os.listdir(annotation_dir) if f.endswith('.png')}

    # Find common files
    common_files = set(image_files.keys()) & set(annotation_files.keys())

    return image_dir, annotation_dir, image_files, annotation_files, common_files

# Iterate over train, val, and test sets
for split in sets:
    image_dir, annotation_dir, image_files, annotation_files, common_files = find_matching_annotations(split)

    print(f"\n[{split.upper()} SET] Found {len(common_files)} matching annotation pairs.")

    # Display one image only for test set
    if split == "test" and common_files:
        random_file = random.choice(list(common_files))
        image_path = os.path.join(image_dir, image_files[random_file])
        annotation_path = os.path.join(annotation_dir, annotation_files[random_file])

        # Display Image & Annotation Side by Side
        fig, axes = plt.subplots(1, 2, figsize=(10, 5))

        try:
            # Load and display the image
            img = Image.open(image_path)
            axes[0].imshow(img)
            axes[0].set_title(f"Image: {image_files[random_file]}")
            axes[0].axis("off")

            # Load and display the annotation
            ann = Image.open(annotation_path)
            axes[1].imshow(ann, cmap="gray")
            axes[1].set_title(f"Annotation: {annotation_files[random_file]}")
            axes[1].axis("off")

            plt.show()
            print(f"Displayed test image '{image_files[random_file]}' and annotation '{annotation_files[random_file]}' successfully.")
        except Exception as e:
            print(f"Error displaying images: {e}")
```

```python
import os
import pandas as pd

def count_files_in_directory(directory, keyword):
    file_counts = []

    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        # Filter files by extension
        valid_files = [f for f in files if f.endswith(('.jpg', '.png', '.csv'))]

        # Check if the number of valid files is greater than 10
        if len(valid_files) > 10:
            # Extract the part of the path after the keyword
            if keyword in root:
                display_path = root.split(keyword)[1]
            else:
                display_path = root

            # Append the count of valid files and the display path
            file_counts.append((display_path, len(valid_files)))

    # Create a DataFrame from the file counts
    df = pd.DataFrame(file_counts, columns=['Directory', 'File Count'])
    return df

keyword = 'ocelot2023_v1.0.1'

# Get the file counts
file_count_df = count_files_in_directory(parent_directory, keyword)

# Display the file counts in a table
if not file_count_df.empty:
    print(file_count_df)
else:
    print("No directories with more than 10 valid files found.")
```

```python
import os
from collections import Counter

def find_minority_file_types(directory):
    inconsistencies = []

    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        # Skip directories containing "cell" in their name
        if "cell" in os.path.basename(root):
            continue

        # Filter files by extension and count occurrences
        file_extensions = [os.path.splitext(f)[1] for f in files if f.endswith(('.jpg', '.png', '.csv'))]
        extension_counts = Counter(file_extensions)

        if extension_counts:
            # Find the most common file type
            most_common = extension_counts.most_common(1)[0]

            # Identify minority file types
            minority_types = [ext for ext, count in extension_counts.items() if count < most_common[1]]

            if minority_types:
                inconsistencies.append((root, minority_types))

    return inconsistencies

# Find minority file types
inconsistencies = find_minority_file_types(parent_directory)

# Display the inconsistencies
if inconsistencies:
    print("Inconsistencies found (minority file types):")
    for directory, minority_types in inconsistencies:
        print(f"Directory: {directory}, Minority Types: {', '.join(minority_types)}")
else:
    print("No inconsistencies found.")
```

```python
import os
import pandas as pd
from PIL import Image

def analyze_directory_contents(parent_directory):
    directory_analysis = []

    for root, dirs, files in os.walk(parent_directory):
        if "cell" in os.path.basename(root).lower():
            continue

        file_info = {'Directory': root, 'Formats': set(), 'Data Types': set()}

        for file in files:
            file_path = os.path.join(root, file)
            try:
                if file.endswith('.csv'):
                    df = pd.read_csv(file_path, nrows=5)
                    file_info['Formats'].add('CSV')
                    if not df.empty:
                        columns_str = ', '.join(df.columns.astype(str))
                        file_info['Data Types'].add(f"Columns: {columns_str}")
                    else:
                        file_info['Data Types'].add("Empty CSV")
                    break

                elif file.endswith(('.jpg', '.png')):
                    img = Image.open(file_path)
                    width, height = img.size
                    file_info['Formats'].add(f"Image ({img.format})")
                    file_info['Data Types'].add(f"Mode: {img.mode}, Dimensions: {width}x{height}")
                    break

            except Exception as e:
                file_info['Formats'].add(f"Error: {str(e)}")
                file_info['Data Types'].add("N/A")
                break

        if file_info['Formats']:
            directory_analysis.append(file_info)

    return directory_analysis


# Example usage
#parent_directory = "/path/to/your/data"  # Replace with your path
analysis_results = analyze_directory_contents(parent_directory)

if analysis_results:
    print("Directory Analysis Results:")
    for result in analysis_results:
        print(f"\nDirectory: {result['Directory']}")
        print(f"  Formats: {', '.join(result['Formats'])}")
        print(f"  Data Types: {'; '.join(result['Data Types'])}")
else:
    print("No relevant files found for analysis.")
```

```python
import os
from PIL import Image

def analyze_image_sizes(parent_directory):
    directory_analysis = []

    # Walk through the directory
    for root, dirs, files in os.walk(parent_directory):
        # Skip directories containing "cell" in their name
        if "cell" in os.path.basename(root):
            continue

        image_sizes = set()
        all_sizes = []

        for file in files:
            file_path = os.path.join(root, file)
            try:
                if file.endswith(('.jpg', '.png')):
                    # Analyze image file
                    img = Image.open(file_path)
                    image_sizes.add(img.size)
                    all_sizes.append(img.size)
            except Exception as e:
                continue  # Skip errors

        # Check for inconsistencies
        if len(image_sizes) > 1:
            inconsistent_sizes = True
        else:
            inconsistent_sizes = False

        # Append the analysis results if there are relevant images
        if image_sizes:
            directory_analysis.append({
                'Directory': root,
                'Image Sizes': image_sizes,
                'Inconsistent Sizes': inconsistent_sizes
            })

    return directory_analysis

# Analyze image sizes
analysis_results = analyze_image_sizes(parent_directory)

# Display the analysis results
if analysis_results:
    print("Image Size Analysis Results:")
    for result in analysis_results:
        print(f"Directory: {result['Directory']}")
        print(f"  Image Sizes: {', '.join(map(str, result['Image Sizes']))}")
        print(f"  Inconsistent Sizes: {result['Inconsistent Sizes']}")
else:
    print("No relevant images found for analysis.")
```

```python
import os
import numpy as np
from PIL import Image
from collections import Counter
import pandas as pd

def get_pixel_counts(folder_path, subset_name):
    pixel_counts = Counter()

    for file in os.listdir(folder_path):
        if file.endswith(('.jpg', '.png', '.tif')):
            img_path = os.path.join(folder_path, file)
            try:
                img = Image.open(img_path).convert('L')
                img_array = np.array(img).flatten()
                pixel_counts.update(img_array)
            except Exception as e:
                print(f"Error processing {file} in {subset_name}: {e}")

    df = pd.DataFrame(sorted(pixel_counts.items()), columns=["Pixel Intensity", subset_name])
    return df

# Base paths
base_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations"
output_path = "/content/pixel_distributions_with_inline_percent.csv"

# Collect data
dfs = []
for subset in ["train", "val", "test"]:
    folder = os.path.join(base_path, subset, "tissue")
    if os.path.exists(folder):
        df = get_pixel_counts(folder, subset)
        dfs.append(df)
    else:
        print(f"Folder not found: {folder}")

# Merge
combined_df = dfs[0]
for df in dfs[1:]:
    combined_df = pd.merge(combined_df, df, on="Pixel Intensity", how="outer")

# Fill missing values and cast to int
combined_df = combined_df.fillna(0).astype({"train": int, "val": int, "test": int})

# Totals per subset
totals = {
    "train": combined_df["train"].sum(),
    "val": combined_df["val"].sum(),
    "test": combined_df["test"].sum()
}

# Inline format: count (xx.xx%)
def format_inline(count, total):
    percent = (count / total) * 100
    return f"{count} ({percent:.2f}%)"

# Apply formatting
combined_df["train"] = combined_df["train"].apply(lambda x: format_inline(x, totals["train"]))
combined_df["val"]   = combined_df["val"].apply(lambda x: format_inline(x, totals["val"]))
combined_df["test"]  = combined_df["test"].apply(lambda x: format_inline(x, totals["test"]))

# Save result
combined_df.to_csv(output_path, index=False)
print(f"Saved to: {output_path}")
```

```python
import os
import numpy as np
from collections import defaultdict
import glob

def inspect_dataset(base_path):
    """
    Inspect all .npy files in train/val/test subfolders
    """
    results = {
        'summary': {
            'total_files': 0,
            'file_formats': defaultdict(int),
            'data_types': defaultdict(int),
            'shapes': defaultdict(int),
            'value_ranges': []
        },
        'by_subset': {},
        'inconsistencies': [],
        'file_details': {}  # Store per-file details for inconsistency tracking
    }

    # Define subsets to check
    subsets = ['train', 'val', 'test']

    for subset in subsets:
        subset_path = os.path.join(base_path, subset)

        if not os.path.exists(subset_path):
            print(f"Warning: {subset_path} does not exist")
            continue

        print(f"\nInspecting {subset} subset...")

        # Find all .npy files
        npy_files = glob.glob(os.path.join(subset_path, "*.npy"))

        subset_results = {
            'file_count': len(npy_files),
            'data_types': defaultdict(int),
            'shapes': defaultdict(int),
            'value_ranges': [],
            'file_sizes_mb': [],
            'files': []  # Store individual file info
        }

        for file_path in npy_files:
            try:
                # Load the array
                data = np.load(file_path)

                # Get file info
                file_size_mb = os.path.getsize(file_path) / (1024 * 1024)
                file_name = os.path.basename(file_path)

                # Collect statistics
                data_type = str(data.dtype)
                shape = data.shape
                min_val = float(np.min(data))
                max_val = float(np.max(data))

                # Store file details
                file_info = {
                    'name': file_name,
                    'path': file_path,
                    'subset': subset,
                    'data_type': data_type,
                    'shape': shape,
                    'shape_str': str(shape),
                    'min_val': min_val,
                    'max_val': max_val,
                    'size_mb': file_size_mb
                }

                subset_results['files'].append(file_info)
                results['file_details'][file_path] = file_info

                # Update counters
                subset_results['data_types'][data_type] += 1
                subset_results['shapes'][str(shape)] += 1
                subset_results['value_ranges'].append((min_val, max_val))
                subset_results['file_sizes_mb'].append(file_size_mb)

                # Update global counters
                results['summary']['total_files'] += 1
                results['summary']['file_formats']['.npy'] += 1
                results['summary']['data_types'][data_type] += 1
                results['summary']['shapes'][str(shape)] += 1
                results['summary']['value_ranges'].append((min_val, max_val))

            except Exception as e:
                print(f"Error processing {file_path}: {str(e)}")
                continue

        results['by_subset'][subset] = subset_results
        print(f"  Found {len(npy_files)} .npy files")

    return results

def analyze_inconsistencies(results):
    """
    Analyze and report inconsistencies with specific file names
    """
    inconsistencies = []

    # Find most common data type and shape
    most_common_dtype = max(results['summary']['data_types'].items(), key=lambda x: x[1])[0] if results['summary']['data_types'] else None
    most_common_shape = max(results['summary']['shapes'].items(), key=lambda x: x[1])[0] if results['summary']['shapes'] else None

    # Check data type inconsistency
    if len(results['summary']['data_types']) > 1:
        inconsistent_files = []
        for file_path, file_info in results['file_details'].items():
            if file_info['data_type'] != most_common_dtype:
                inconsistent_files.append({
                    'file': file_info['name'],
                    'path': file_path,
                    'subset': file_info['subset'],
                    'found': file_info['data_type'],
                    'expected': most_common_dtype
                })

        inconsistencies.append({
            'type': 'Data Type Inconsistency',
            'details': dict(results['summary']['data_types']),
            'files': inconsistent_files
        })

    # Check shape inconsistency
    if len(results['summary']['shapes']) > 1:
        inconsistent_files = []
        for file_path, file_info in results['file_details'].items():
            if file_info['shape_str'] != most_common_shape:
                inconsistent_files.append({
                    'file': file_info['name'],
                    'path': file_path,
                    'subset': file_info['subset'],
                    'found': file_info['shape_str'],
                    'expected': most_common_shape
                })

        inconsistencies.append({
            'type': 'Shape Inconsistency',
            'details': dict(results['summary']['shapes']),
            'files': inconsistent_files
        })

    # Check value range consistency (for softmax, should be [0,1])
    value_ranges = results['summary']['value_ranges']
    if value_ranges:
        min_vals = [r[0] for r in value_ranges]
        max_vals = [r[1] for r in value_ranges]

        overall_min = min(min_vals)
        overall_max = max(max_vals)

        # Check if values are outside expected softmax range [0,1]
        problematic_files = []
        for file_path, file_info in results['file_details'].items():
            if file_info['min_val'] < -1e-6 or file_info['max_val'] > 1 + 1e-6:
                problematic_files.append({
                    'file': file_info['name'],
                    'path': file_path,
                    'subset': file_info['subset'],
                    'min_val': file_info['min_val'],
                    'max_val': file_info['max_val']
                })

        if problematic_files:
            inconsistencies.append({
                'type': 'Value Range Issue (Expected [0,1] for softmax)',
                'details': f'Overall range: [{overall_min:.6f}, {overall_max:.6f}]',
                'files': problematic_files
            })

        # Check for significant variation in ranges across files
        min_variation = max(min_vals) - min(min_vals)
        max_variation = max(max_vals) - min(max_vals)

        if min_variation > 1e-3 or max_variation > 1e-3:
            variable_files = []
            mean_min = np.mean(min_vals)
            mean_max = np.mean(max_vals)

            for file_path, file_info in results['file_details'].items():
                if abs(file_info['min_val'] - mean_min) > 1e-3 or abs(file_info['max_val'] - mean_max) > 1e-3:
                    variable_files.append({
                        'file': file_info['name'],
                        'path': file_path,
                        'subset': file_info['subset'],
                        'min_val': file_info['min_val'],
                        'max_val': file_info['max_val']
                    })

            inconsistencies.append({
                'type': 'Value Range Variation',
                'details': f'Min values vary by {min_variation:.6f}, Max values vary by {max_variation:.6f}',
                'files': variable_files
            })

    return inconsistencies

def check_corresponding_images(softmax_path, images_path):
    """
    Check if softmax files have corresponding cell images
    """
    print(f"\nChecking correspondence between softmax files and cell images...")

    correspondence_issues = {
        'missing_cell_images': [],
        'missing_softmax_files': [],
        'summary': {}
    }

    subsets = ['train', 'val', 'test']

    for subset in subsets:
        print(f"\nChecking {subset} subset...")

        # Get softmax files
        softmax_subset_path = os.path.join(softmax_path, subset)
        softmax_files = []
        if os.path.exists(softmax_subset_path):
            softmax_files = [f for f in os.listdir(softmax_subset_path) if f.endswith('.npy')]

        # Get cell image files
        cell_images_path = os.path.join(images_path, 'images', subset, 'cell')
        cell_files = []
        if os.path.exists(cell_images_path):
            # Common image extensions
            extensions = ['*.png', '*.jpg', '*.jpeg', '*.tif', '*.tiff', '*.bmp']
            for ext in extensions:
                cell_files.extend([os.path.basename(f) for f in glob.glob(os.path.join(cell_images_path, ext))])

        print(f"  Softmax files: {len(softmax_files)}")
        print(f"  Cell image files: {len(cell_files)}")

        # Convert to base names (without extensions) for comparison
        softmax_bases = set()
        for f in softmax_files:
            base_name = os.path.splitext(f)[0]
            softmax_bases.add(base_name)

        cell_bases = set()
        for f in cell_files:
            base_name = os.path.splitext(f)[0]
            cell_bases.add(base_name)

        # Find missing correspondences
        missing_cell = softmax_bases - cell_bases
        missing_softmax = cell_bases - softmax_bases

        if missing_cell:
            for base_name in missing_cell:
                # Find the actual softmax file name
                actual_softmax_file = None
                for sf in softmax_files:
                    if os.path.splitext(sf)[0] == base_name:
                        actual_softmax_file = sf
                        break

                correspondence_issues['missing_cell_images'].append({
                    'subset': subset,
                    'softmax_file': actual_softmax_file,
                    'expected_cell_base': base_name,
                    'softmax_path': os.path.join(softmax_subset_path, actual_softmax_file) if actual_softmax_file else None
                })

        if missing_softmax:
            for base_name in missing_softmax:
                # Find the actual cell file name
                actual_cell_file = None
                for cf in cell_files:
                    if os.path.splitext(cf)[0] == base_name:
                        actual_cell_file = cf
                        break

                correspondence_issues['missing_softmax_files'].append({
                    'subset': subset,
                    'cell_file': actual_cell_file,
                    'expected_softmax_base': base_name,
                    'cell_path': os.path.join(cell_images_path, actual_cell_file) if actual_cell_file else None
                })

        correspondence_issues['summary'][subset] = {
            'softmax_count': len(softmax_files),
            'cell_count': len(cell_files),
            'missing_cell_images': len(missing_cell),
            'missing_softmax_files': len(missing_softmax),
            'matched_pairs': len(softmax_bases & cell_bases)
        }

    return correspondence_issues

def print_report(results, correspondence_issues=None):
    """
    Print comprehensive report with file names for inconsistencies
    """
    print("\n" + "="*80)
    print("DATASET INSPECTION REPORT")
    print("="*80)

    # Overall summary
    print(f"\nOVERALL SUMMARY:")
    print(f"Total files analyzed: {results['summary']['total_files']}")

    print(f"\nFile formats:")
    for fmt, count in results['summary']['file_formats'].items():
        print(f"  {fmt}: {count} files")

    print(f"\nData types:")
    for dtype, count in results['summary']['data_types'].items():
        print(f"  {dtype}: {count} files")

    print(f"\nShapes (most common first):")
    sorted_shapes = sorted(results['summary']['shapes'].items(), key=lambda x: x[1], reverse=True)
    for shape, count in sorted_shapes:
        print(f"  {shape}: {count} files")

    if results['summary']['value_ranges']:
        value_ranges = results['summary']['value_ranges']
        min_vals = [r[0] for r in value_ranges]
        max_vals = [r[1] for r in value_ranges]
        print(f"\nValue ranges:")
        print(f"  Overall min: {min(min_vals):.6f}")
        print(f"  Overall max: {max(max_vals):.6f}")
        print(f"  Average min: {np.mean(min_vals):.6f}")
        print(f"  Average max: {np.mean(max_vals):.6f}")

    # Per-subset breakdown
    print(f"\nPER-SUBSET BREAKDOWN:")
    for subset, data in results['by_subset'].items():
        print(f"\n{subset.upper()} SET:")
        print(f"  Files: {data['file_count']}")

        if data['file_sizes_mb']:
            avg_size = np.mean(data['file_sizes_mb'])
            print(f"  Average file size: {avg_size:.2f} MB")

        print(f"  Data types: {dict(data['data_types'])}")
        print(f"  Unique shapes: {len(data['shapes'])}")

        if data['value_ranges']:
            subset_mins = [r[0] for r in data['value_ranges']]
            subset_maxs = [r[1] for r in data['value_ranges']]
            print(f"  Value range: [{min(subset_mins):.6f}, {max(subset_maxs):.6f}]")

    # Inconsistencies with file names
    inconsistencies = analyze_inconsistencies(results)

    if inconsistencies:
        print(f"\nâš ï¸  INCONSISTENCIES FOUND:")
        for i, issue in enumerate(inconsistencies, 1):
            print(f"\n{i}. {issue['type']}")
            print(f"   Overview: {issue['details']}")

            if 'files' in issue and issue['files']:
                print(f"   Affected files ({len(issue['files'])}):")
                for file_info in issue['files'][:10]:  # Show first 10 files
                    if 'expected' in file_info:
                        print(f"     â€¢ {file_info['file']} ({file_info['subset']}) - Found: {file_info['found']}, Expected: {file_info['expected']}")
                    else:
                        print(f"     â€¢ {file_info['file']} ({file_info['subset']}) - Range: [{file_info['min_val']:.6f}, {file_info['max_val']:.6f}]")

                if len(issue['files']) > 10:
                    print(f"     ... and {len(issue['files']) - 10} more files")
    else:
        print(f"\nâœ… NO MAJOR INCONSISTENCIES FOUND")

    # Correspondence report
    if correspondence_issues:
        print(f"\n" + "="*80)
        print("CORRESPONDENCE CHECK REPORT")
        print("="*80)

        print(f"\nSUMMARY:")
        for subset, summary in correspondence_issues['summary'].items():
            print(f"\n{subset.upper()} SET:")
            print(f"  Softmax files: {summary['softmax_count']}")
            print(f"  Cell image files: {summary['cell_count']}")
            print(f"  Matched pairs: {summary['matched_pairs']}")
            print(f"  Missing cell images: {summary['missing_cell_images']}")
            print(f"  Missing softmax files: {summary['missing_softmax_files']}")

        # Missing cell images
        if correspondence_issues['missing_cell_images']:
            print(f"\nâš ï¸  SOFTMAX FILES WITHOUT CORRESPONDING CELL IMAGES:")
            for item in correspondence_issues['missing_cell_images']:
                print(f"  â€¢ {item['softmax_file']} ({item['subset']}) - No matching cell image for base name: {item['expected_cell_base']}")

        # Missing softmax files
        if correspondence_issues['missing_softmax_files']:
            print(f"\nâš ï¸  CELL IMAGES WITHOUT CORRESPONDING SOFTMAX FILES:")
            for item in correspondence_issues['missing_softmax_files']:
                print(f"  â€¢ {item['cell_file']} ({item['subset']}) - No matching softmax file for base name: {item['expected_softmax_base']}")

        if not correspondence_issues['missing_cell_images'] and not correspondence_issues['missing_softmax_files']:
            print(f"\nâœ… ALL FILES HAVE CORRESPONDING PAIRS")

    print("\n" + "="*80)

# Main execution
if __name__ == "__main__":
    # Set your dataset paths
    softmax_path = "/content/drive/MyDrive/research/ocelot/dataset/integration/cropped tissue softmax_UNet"
    images_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1"

    print(f"Inspecting softmax dataset at: {softmax_path}")
    print(f"Checking correspondence with images at: {images_path}")

    # Check if paths exist
    if not os.path.exists(softmax_path):
        print(f"Error: Softmax path {softmax_path} does not exist!")
        exit(1)

    if not os.path.exists(images_path):
        print(f"Warning: Images path {images_path} does not exist! Skipping correspondence check.")
        images_path = None

    # Inspect the softmax dataset
    results = inspect_dataset(softmax_path)

    # Check correspondence with cell images
    correspondence_issues = None
    if images_path:
        correspondence_issues = check_corresponding_images(softmax_path, images_path)

    # Print comprehensive report
    print_report(results, correspondence_issues)
```

```python
 import numpy as np
import os

def pad_array_to_target(array, target_size=(1024, 1024), pad_value=0):
    """
    Pad array to target size with symmetric padding
    """
    current_shape = array.shape
    print(f"  Current shape: {current_shape}")

    if len(current_shape) == 2:
        # 2D array (H, W)
        h, w = current_shape
        target_h, target_w = target_size

        # Calculate padding needed
        pad_h = target_h - h
        pad_w = target_w - w

        print(f"  Padding needed: height +{pad_h}, width +{pad_w}")

        if pad_h < 0 or pad_w < 0:
            print(f"  Warning: Array is larger than target size!")
            return array

        # Symmetric padding (distribute padding evenly)
        pad_h_before = pad_h // 2
        pad_h_after = pad_h - pad_h_before
        pad_w_before = pad_w // 2
        pad_w_after = pad_w - pad_w_before

        print(f"  Padding distribution: top={pad_h_before}, bottom={pad_h_after}, left={pad_w_before}, right={pad_w_after}")

        padded = np.pad(array,
                       ((pad_h_before, pad_h_after), (pad_w_before, pad_w_after)),
                       mode='constant', constant_values=pad_value)

    elif len(current_shape) == 3:
        # 3D array - need to determine if it's (C, H, W) or (H, W, C)
        if current_shape[0] <= 10:  # Assume (C, H, W) if first dim is small
            c, h, w = current_shape
            target_h, target_w = target_size

            pad_h = target_h - h
            pad_w = target_w - w

            print(f"  Detected format: (C, H, W) with {c} channels")
            print(f"  Padding needed: height +{pad_h}, width +{pad_w}")

            if pad_h < 0 or pad_w < 0:
                print(f"  Warning: Array is larger than target size!")
                return array

            pad_h_before = pad_h // 2
            pad_h_after = pad_h - pad_h_before
            pad_w_before = pad_w // 2
            pad_w_after = pad_w - pad_w_before

            print(f"  Padding distribution: top={pad_h_before}, bottom={pad_h_after}, left={pad_w_before}, right={pad_w_after}")

            padded = np.pad(array,
                           ((0, 0), (pad_h_before, pad_h_after), (pad_w_before, pad_w_after)),
                           mode='constant', constant_values=pad_value)

        else:  # Assume (H, W, C)
            h, w, c = current_shape
            target_h, target_w = target_size

            pad_h = target_h - h
            pad_w = target_w - w

            print(f"  Detected format: (H, W, C) with {c} channels")
            print(f"  Padding needed: height +{pad_h}, width +{pad_w}")

            if pad_h < 0 or pad_w < 0:
                print(f"  Warning: Array is larger than target size!")
                return array

            pad_h_before = pad_h // 2
            pad_h_after = pad_h - pad_h_before
            pad_w_before = pad_w // 2
            pad_w_after = pad_w - pad_w_before

            print(f"  Padding distribution: top={pad_h_before}, bottom={pad_h_after}, left={pad_w_before}, right={pad_w_after}")

            padded = np.pad(array,
                           ((pad_h_before, pad_h_after), (pad_w_before, pad_w_after), (0, 0)),
                           mode='constant', constant_values=pad_value)

    print(f"  Final shape: {padded.shape}")
    return padded

def validate_softmax_after_padding(array, file_name):
    """
    Validate softmax properties after padding
    """
    print(f"  Validating {file_name}:")

    # Check value range
    min_val = np.min(array)
    max_val = np.max(array)
    print(f"    Value range: [{min_val:.6f}, {max_val:.6f}]")

    if min_val < -1e-6:
        print(f"    âš ï¸  Warning: Negative values found!")
    if max_val > 1 + 1e-6:
        print(f"    âš ï¸  Warning: Values > 1 found!")

    # Check data type
    print(f"    Data type: {array.dtype}")

    # For multi-class softmax, check if probabilities sum to 1
    if len(array.shape) == 3:
        if array.shape[0] <= 10:  # (C, H, W)
            if array.shape[0] > 1:
                prob_sums = np.sum(array, axis=0)
                mean_sum = np.mean(prob_sums)
                std_sum = np.std(prob_sums)
                print(f"    Probability sum - Mean: {mean_sum:.6f}, Std: {std_sum:.6f}")
                if abs(mean_sum - 1.0) > 1e-3:
                    print(f"    âš ï¸  Warning: Probabilities don't sum to ~1.0!")
        else:  # (H, W, C)
            if array.shape[2] > 1:
                prob_sums = np.sum(array, axis=2)
                mean_sum = np.mean(prob_sums)
                std_sum = np.std(prob_sums)
                print(f"    Probability sum - Mean: {mean_sum:.6f}, Std: {std_sum:.6f}")
                if abs(mean_sum - 1.0) > 1e-3:
                    print(f"    âš ï¸  Warning: Probabilities don't sum to ~1.0!")

def pad_specific_files(base_path, target_size=(1024, 1024), pad_value=0, backup=True):
    """
    Pad the two specific problematic files: 129.npy and 657.npy
    """
    # Define the problematic files
    files_to_fix = [
        {'file': '129.npy', 'subset': 'train'},
        {'file': '657.npy', 'subset': 'test'}
    ]

    print("="*60)
    print("PADDING SPECIFIC FILES")
    print("="*60)

    for file_info in files_to_fix:
        file_name = file_info['file']
        subset = file_info['subset']

        # Construct file path
        file_path = os.path.join(base_path, subset, file_name)

        print(f"\nProcessing: {file_name} in {subset} subset")
        print(f"Path: {file_path}")

        # Check if file exists
        if not os.path.exists(file_path):
            print(f"  âŒ File not found: {file_path}")
            continue

        try:
            # Load the array
            print(f"  ðŸ“‚ Loading {file_name}...")
            original_array = np.load(file_path)

            # Create backup if requested
            if backup:
                backup_path = file_path.replace('.npy', '_backup.npy')
                print(f"  ðŸ’¾ Creating backup: {os.path.basename(backup_path)}")
                np.save(backup_path, original_array)

            # Check if padding is actually needed
            current_shape = original_array.shape
            if len(current_shape) >= 2:
                h, w = current_shape[-2:]  # Get last two dimensions
                if (h, w) == target_size:
                    print(f"  âœ… File already has correct size {target_size}, skipping")
                    continue

            # Pad the array
            print(f"  ðŸ”§ Padding array...")
            padded_array = pad_array_to_target(original_array, target_size, pad_value)

            # Validate the result
            validate_softmax_after_padding(padded_array, file_name)

            # Save the padded array (overwrite original)
            print(f"  ðŸ’¾ Saving padded array...")
            np.save(file_path, padded_array)

            print(f"  âœ… Successfully padded {file_name}")

        except Exception as e:
            print(f"  âŒ Error processing {file_name}: {str(e)}")
            continue

    print(f"\n" + "="*60)
    print("PADDING COMPLETE")
    print("="*60)

def verify_fixed_files(base_path, target_size=(1024, 1024)):
    """
    Verify that the fixed files now have the correct size
    """
    files_to_check = [
        {'file': '129.npy', 'subset': 'train'},
        {'file': '657.npy', 'subset': 'test'}
    ]

    print(f"\nVERIFICATION:")
    print("-" * 40)

    all_good = True

    for file_info in files_to_check:
        file_name = file_info['file']
        subset = file_info['subset']
        file_path = os.path.join(base_path, subset, file_name)

        if os.path.exists(file_path):
            try:
                array = np.load(file_path)
                current_shape = array.shape

                if len(current_shape) >= 2:
                    h, w = current_shape[-2:]
                    if (h, w) == target_size:
                        print(f"âœ… {file_name} ({subset}): {current_shape} - Correct size!")
                    else:
                        print(f"âŒ {file_name} ({subset}): {current_shape} - Still wrong size!")
                        all_good = False
                else:
                    print(f"â“ {file_name} ({subset}): {current_shape} - Unexpected dimensions!")
                    all_good = False

            except Exception as e:
                print(f"âŒ {file_name} ({subset}): Error loading - {str(e)}")
                all_good = False
        else:
            print(f"âŒ {file_name} ({subset}): File not found!")
            all_good = False

    if all_good:
        print(f"\nðŸŽ‰ All files are now the correct size!")
    else:
        print(f"\nâš ï¸  Some files still have issues.")

# Main execution
if __name__ == "__main__":
    # Set your dataset path
    base_path = "/content/drive/MyDrive/research/ocelot/dataset/integration/cropped tissue softmax_UNet"

    print(f"Base path: {base_path}")

    # Check if base path exists
    if not os.path.exists(base_path):
        print(f"âŒ Error: Path {base_path} does not exist!")
        exit(1)

    # Pad the specific problematic files
    pad_specific_files(
        base_path=base_path,
        target_size=(1024, 1024),
        pad_value=0,        # Use 0 for background/no prediction
        backup=True         # Create backup files before modifying
    )

    # Verify the fixes
    verify_fixed_files(base_path, target_size=(1024, 1024))
```

```python
import cv2
import numpy as np
import os
import collections
import hashlib

# Set the base dataset path
base_folder = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images"

# Define subsets to iterate through
subsets = ["train", "val", "test"]

# Dictionary to store image hashes for duplicate detection
image_hashes = {}

for subset in subsets:
    folder_path = os.path.join(base_folder, subset, "cell")  # Adjust as needed

    if not os.path.exists(folder_path):
        print(f"Skipping {folder_path} (not found)")
        continue

    print(f"\n===== Analyzing {subset.upper()} Set =====")

    # Get all image files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.endswith((".jpg", ".png", ".jpeg", ".tif", ".bmp"))]

    if not image_files:
        print("No image files found in this subset.")
        continue

    dimensions = []
    dtypes = []
    formats = []
    max_values = []
    duplicate_images = {}

    for file in image_files:
        file_path = os.path.join(folder_path, file)

        # Load image
        image = cv2.imread(file_path)

        if image is None:
            print(f"Error loading image: {file}")
            continue

        # Get image metadata
        height, width, channels = image.shape
        dtype = image.dtype
        file_format = file.split(".")[-1]  # Extract format from extension
        max_value = image.max()

        # Store attributes
        dimensions.append((height, width, channels))
        dtypes.append(dtype)
        formats.append(file_format)
        max_values.append(max_value)

        # Compute hash for duplicate detection
        image_hash = hashlib.md5(image.tobytes()).hexdigest()
        if image_hash in image_hashes:
            duplicate_images.setdefault(image_hashes[image_hash], []).append(file)
        else:
            image_hashes[image_hash] = file

    # Determine the most common attributes
    most_common_dimension = collections.Counter(dimensions).most_common(1)[0][0]
    most_common_dtype = collections.Counter(dtypes).most_common(1)[0][0]
    most_common_format = collections.Counter(formats).most_common(1)[0][0]
    most_common_max_value = collections.Counter(max_values).most_common(1)[0][0]

    # Display results
    print(f"Most Common Image Size (HxWxC): {most_common_dimension}")
    print(f"Most Common Data Type: {most_common_dtype}")
    print(f"Most Common Format: {most_common_format}")
    print(f"Most Common Max Pixel Value: {most_common_max_value}")

    # Check for inconsistencies (excluding max value inconsistencies)
    inconsistent_dimensions = [d for d in dimensions if d != most_common_dimension]
    inconsistent_dtypes = [d for d in dtypes if d != most_common_dtype]
    inconsistent_formats = [f for f in formats if f != most_common_format]

    if any([inconsistent_dimensions, inconsistent_dtypes, inconsistent_formats]):
        print("\n===== Inconsistencies Found =====")
        if inconsistent_dimensions:
            print(f"Different Image Sizes (HxWxC): {set(inconsistent_dimensions)}")
        if inconsistent_dtypes:
            print(f"Different Data Types: {set(inconsistent_dtypes)}")
        if inconsistent_formats:
            print(f"Different Formats: {set(inconsistent_formats)}")
    else:
        print("No inconsistencies detected.")

    # Display duplicate images
    if duplicate_images:
        print("\n===== Duplicate Images Found =====")
        for original, duplicates in duplicate_images.items():
            print(f"Original Image: {original}")
            for dup in duplicates:
                print(f"- Duplicate: {dup}")

print("\nAnalysis Complete.")
```

```python
import os
import pandas as pd
import collections
import hashlib

# Set the base dataset path
base_folder = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations"

# Define subsets to iterate through
subsets = ["train", "val", "test"]

# Dictionary to store file hashes for duplicate detection
file_hashes = {}

for subset in subsets:
    folder_path = os.path.join(base_folder, subset, "cell")

    if not os.path.exists(folder_path):
        print(f"Skipping {folder_path} (not found)")
        continue

    print(f"\n===== Analyzing {subset.upper()} Set =====")

    # Get all .csv files in the folder
    csv_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]
    total_csv_files = len(csv_files)
    processed_files = 0

    if not csv_files:
        print("No CSV files found in this subset.")
        continue

    column_counts = {}
    dtypes_list = {}
    duplicate_files = {}

    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        try:
            df = pd.read_csv(file_path, header=None)  # Ensure no headers are read

            # Compute file hash for duplicate detection
            file_hash = hashlib.md5(pd.util.hash_pandas_object(df, index=False).values).hexdigest()
            if file_hash in file_hashes:
                duplicate_files.setdefault(file_hashes[file_hash], []).append(file)
            else:
                file_hashes[file_hash] = file

            # Collect statistics
            column_counts[file] = len(df.columns)
            dtypes_list[file] = tuple(df.dtypes.astype(str).tolist())

            processed_files += 1

        except Exception as e:
            print(f"Error loading {file}: {e}")
            continue

    # Determine the most common column count
    col_count_counter = collections.Counter(column_counts.values())
    most_common_col_count, col_count_freq = col_count_counter.most_common(1)[0] if col_count_counter else (None, 0)

    # Determine the most common data types structure
    dtype_counter = collections.Counter(dtypes_list.values())
    most_common_dtypes, dtype_freq = dtype_counter.most_common(1)[0] if dtype_counter else (None, 0)

    # Display results
    print(f"Total CSV Files Found: {total_csv_files}")
    print(f"Successfully Processed: {processed_files}")
    print(f"Most Common Column Count: {most_common_col_count} (Occurrences: {col_count_freq})")
    print(f"Most Common Data Types: {most_common_dtypes} (Occurrences: {dtype_freq})")

    # Check for inconsistencies
    inconsistent_col_files = [file for file, count in column_counts.items() if count != most_common_col_count]
    inconsistent_dtype_files = [file for file, dtypes in dtypes_list.items() if dtypes != most_common_dtypes]

    if inconsistent_col_files or inconsistent_dtype_files:
        print("\n===== Inconsistencies Found =====")
        if inconsistent_col_files:
            print(f"Files with Different Column Counts ({len(inconsistent_col_files)}):")
            for file in inconsistent_col_files:
                print(f"- {file} (Columns: {column_counts[file]})")

        if inconsistent_dtype_files:
            print(f"Files with Different Data Types ({len(inconsistent_dtype_files)}):")
            for file in inconsistent_dtype_files:
                print(f"- {file} (Data Types: {dtypes_list[file]})")
    else:
        print("No inconsistencies detected.")

    # Display duplicate files
    if duplicate_files:
        print("\n===== Duplicate Files Found =====")
        for original, duplicates in duplicate_files.items():
            print(f"Original File: {original}")
            for dup in duplicates:
                print(f"- Duplicate: {dup}")

print("\nAnalysis Complete.")
```

```python
import os
import pandas as pd

# Paths to dataset splits
base_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations"
splits = ["train/cell", "val/cell", "test/cell"]

# Expected number of columns (since there are no headers)
EXPECTED_NUM_COLUMNS = 3
VALID_CLASSES = {1, 2}  # Allowed class labels

# Logging issues
issues = []

def check_csv_structure(csv_path):
    """Check if a CSV file follows the expected format (no column names)."""
    try:
        # Read CSV without headers
        df = pd.read_csv(csv_path, header=None)

        # Check number of columns
        if df.shape[1] != EXPECTED_NUM_COLUMNS:
            issues.append(f"âŒ Incorrect column count in {csv_path}: Expected {EXPECTED_NUM_COLUMNS}, Found {df.shape[1]}")
            return  # Skip further checks if the structure is wrong

        # Assign column names for validation
        df.columns = ["x", "y", "class"]

        # Check for missing values
        if df.isnull().any().any():
            issues.append(f"âš ï¸ Missing values in {csv_path}")

        # Check data types
        if not pd.api.types.is_numeric_dtype(df["x"]) or not pd.api.types.is_numeric_dtype(df["y"]):
            issues.append(f"âŒ Non-numeric x/y values in {csv_path}")
        if not pd.api.types.is_integer_dtype(df["class"]):
            issues.append(f"âŒ Class column is not integer in {csv_path}")

        # Check valid class labels (only 1 or 2)
        invalid_classes = set(df["class"].unique()) - VALID_CLASSES
        if invalid_classes:
            issues.append(f"âŒ Invalid class values in {csv_path}: {invalid_classes}")

    except Exception as e:
        issues.append(f"âŒ Failed to read {csv_path}: {str(e)}")

def verify_structure():
    """Verify the annotation structure for train, val, and test."""
    for split in splits:
        split_path = os.path.join(base_path, split)

        # List all CSV files
        csv_files = [f for f in os.listdir(split_path) if f.endswith(".csv")]

        for csv_file in csv_files:
            csv_path = os.path.join(split_path, csv_file)
            check_csv_structure(csv_path)

    # Print or save issues
    if issues:
        print("\n".join(issues))
    else:
        print("âœ… All CSV annotation files are structurally valid!")

# Run verification
verify_structure()
```

```python
import os
import pandas as pd

# Paths to dataset splits
base_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1"
annotation_path = os.path.join(base_path, "annotations")
image_path = os.path.join(base_path, "images")  # Assuming images are stored here

splits = ["train/cell", "val/cell", "test/cell"]

# Image dimensions (fixed)
IMG_WIDTH, IMG_HEIGHT = 1024, 1024

# Allowed image extensions
IMAGE_EXTENSIONS = {".png", ".jpg", ".jpeg"}

# Logging issues
issues = []

def check_logical_consistency(csv_path, img_file):
    """Check if the CSV file is logically consistent."""
    try:
        # Read CSV file without headers (assumed format: x, y, class)
        df = pd.read_csv(csv_path, header=None, names=["x", "y", "class"])

        # Check if image exists
        if img_file is None or not os.path.exists(img_file):
            issues.append(f"âŒ Missing image for annotation: {csv_path}")
            return  # Skip further checks if image is missing

        # Check coordinate bounds
        out_of_bounds = df[(df["x"] < 0) | (df["x"] >= IMG_WIDTH) | (df["y"] < 0) | (df["y"] >= IMG_HEIGHT)]
        if not out_of_bounds.empty:
            issues.append(f"âš ï¸ Out-of-bounds coordinates in {csv_path}: {len(out_of_bounds)} entries")

        # Check duplicate points
        duplicate_points = df.duplicated(subset=["x", "y"], keep=False)
        if duplicate_points.sum() > 0:
            issues.append(f"âš ï¸ Duplicate (x, y) points found in {csv_path}: {duplicate_points.sum()} duplicates")

    except Exception as e:
        issues.append(f"âŒ Error processing {csv_path}: {str(e)}")

def verify_logical_consistency():
    """Verify logical integrity of annotations."""
    for split in splits:
        ann_dir = os.path.join(annotation_path, split)
        img_dir = os.path.join(image_path, split)

        # List all CSV files
        csv_files = [f for f in os.listdir(ann_dir) if f.endswith(".csv")]

        for csv_file in csv_files:
            csv_path = os.path.join(ann_dir, csv_file)

            # Find corresponding image file (same name, different extensions)
            img_base = os.path.splitext(csv_file)[0]  # Remove .csv extension
            img_file = None
            for ext in IMAGE_EXTENSIONS:
                if os.path.exists(os.path.join(img_dir, img_base + ext)):
                    img_file = os.path.join(img_dir, img_base + ext)
                    break

            check_logical_consistency(csv_path, img_file)

    # Print or save issues
    if issues:
        print("\n".join(issues))
    else:
        print("âœ… All annotations are logically valid!")

# Run verification
verify_logical_consistency()
```

```python
import os
import numpy as np
import pandas as pd

# === CONFIGURATION ===
csv_root = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations"
mask_root = "/content/drive/MyDrive/research/ocelot/dataset/GT/gaussian masks/type5"
splits = ["train", "val", "test"]
radius = 7  # Radius to check around center

# === STORAGE ===
report = {
    split: {
        "class_1": {"total": 0, "eq_1": 0, "between_09_1": 0, "between_08_09": 0, "lt_08": 0},
        "class_2": {"total": 0, "eq_1": 0, "between_09_1": 0, "between_08_09": 0, "lt_08": 0}
    }
    for split in splits
}

low_confidence_cells = {
    split: {
        "class_1": [],  # List of (x, y, value)
        "class_2": []
    }
    for split in splits
}

# === MAIN LOOP ===
for split in splits:
    csv_dir = os.path.join(csv_root, split, "cell")
    mask_dir = os.path.join(mask_root, split, "cell")

    for fname in os.listdir(csv_dir):
        if not fname.endswith(".csv"):
            continue

        csv_path = os.path.join(csv_dir, fname)
        mask_path = os.path.join(mask_dir, fname.replace(".csv", ".npy"))

        if not os.path.exists(mask_path):
            print(f"Missing mask for {fname}")
            continue

        # Load mask (shape: [2, H, W])
        mask = np.load(mask_path)
        H, W = mask.shape[1:]

        # Load CSV (format: x, y, class)
        df = pd.read_csv(csv_path, names=["x", "y", "class"])

        for _, row in df.iterrows():
            x, y, cls = row["x"], row["y"], int(row["class"])
            if cls not in [1, 2]:
                continue

            channel = cls - 1
            y_int, x_int = int(round(y)), int(round(x))

            # Define patch bounds
            y_min = max(0, y_int - radius)
            y_max = min(H, y_int + radius + 1)
            x_min = max(0, x_int - radius)
            x_max = min(W, x_int + radius + 1)

            local_patch = mask[channel, y_min:y_max, x_min:x_max]
            peak_val = np.max(local_patch)

            stats = report[split][f"class_{cls}"]
            stats["total"] += 1

            if peak_val == 1.0:
                stats["eq_1"] += 1
            elif 0.9 < peak_val < 1.0:
                stats["between_09_1"] += 1
            elif 0.8 < peak_val <= 0.9:
                stats["between_08_09"] += 1
            else:
                stats["lt_08"] += 1
                # Log low confidence
                low_confidence_cells[split][f"class_{cls}"].append({
                    "file": fname,
                    "x": x,
                    "y": y,
                    "class": cls,
                    "value": float(peak_val)
                })

# === DISPLAY REPORT ===
for split in splits:
    print(f"\n===== {split.upper()} =====")
    for cls in ["class_1", "class_2"]:
        stats = report[split][cls]
        print(f"\nClass {cls[-1]}:")
        print(f"- Total cells        : {stats['total']}")
        print(f"- Value = 1.0        : {stats['eq_1']}")
        print(f"- 0.9 < value < 1.0  : {stats['between_09_1']}")
        print(f"- 0.8 < value <= 0.9 : {stats['between_08_09']}")
        print(f"- value <= 0.8       : {stats['lt_08']}")

        # Show sample low-confidence cases
        low_conf = low_confidence_cells[split][cls]
        if low_conf:
            print(f"\n  First 5 low-confidence detections (<0.8):")
            for entry in low_conf[:5]:
                print(f"    File: {entry['file']}, (x, y) = ({entry['x']}, {entry['y']}), Value = {entry['value']:.3f}")

# === OPTIONAL: Export low-confidence to CSV ===
# for split in splits:
#     for cls in ["class_1", "class_2"]:
#         df = pd.DataFrame(low_confidence_cells[split][cls])
#         if not df.empty:
#             out_path = f"/content/{split}_{cls}_low_conf.csv"
#             df.to_csv(out_path, index=False)
#             print(f"Saved: {out_path}")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import peak_local_max
from skimage.draw import disk
import random
import glob

# === Cell detection function ===
def detect_local_maxima(mask, min_distance=3, threshold_abs=0.1):
    coords = peak_local_max(
        mask,
        min_distance=min_distance,
        threshold_abs=threshold_abs,
        exclude_border=False
    )
    return coords

# === Updated Visualization function ===
def plot_detection(mask, detections_ch0, detections_ch1, title="Detected Cells"):
    merged = np.clip(mask[0] + mask[1], 0, 1)

    # Create overlay image with just detected cells on black background
    overlay = np.zeros((merged.shape[0], merged.shape[1], 3), dtype=np.float32)

    # Channel 0: yellow (R=1, G=1, B=0)
    for y, x in detections_ch0:
        rr, cc = disk((y, x), radius=3, shape=merged.shape)
        overlay[rr, cc] = [1, 1, 0]

    # Channel 1: blue (R=0, G=0, B=1)
    for y, x in detections_ch1:
        rr, cc = disk((y, x), radius=3, shape=merged.shape)
        overlay[rr, cc] = [0, 0, 1]

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.title("Merged Mask")
    plt.imshow(merged, cmap='gray')
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title(title)
    plt.imshow(overlay)
    plt.axis("off")
    plt.show()

# === Load a real mask ===
mask_dir = "/content/drive/MyDrive/research/ocelot/dataset/GT/gaussian masks/type5/train/cell"
mask_files = glob.glob(os.path.join(mask_dir, "*.npy"))
mask_real = np.load(random.choice(mask_files))

# === Generate synthetic mask with overlapping cells ===
def generate_synthetic_mask(shape=(1024, 1024), sigma=7, radius=11, n_cells=10):
    mask = np.zeros((2, shape[0], shape[1]), dtype=np.float32)
    centers_ch0 = [(random.randint(100, 924), random.randint(100, 924)) for _ in range(n_cells)]
    centers_ch1 = [(y + random.randint(-5, 5), x + random.randint(-5, 5)) for y, x in centers_ch0]

    for y, x in centers_ch0:
        y_min, y_max = max(0, y - radius), min(shape[0], y + radius + 1)
        x_min, x_max = max(0, x - radius), min(shape[1], x + radius + 1)
        y_grid, x_grid = np.meshgrid(np.arange(y_min, y_max), np.arange(x_min, x_max), indexing='ij')
        gaussian = np.exp(-((x_grid - x) ** 2 + (y_grid - y) ** 2) / (2 * sigma ** 2))
        circle_mask = ((x_grid - x) ** 2 + (y_grid - y) ** 2) <= radius ** 2
        gaussian *= circle_mask
        mask[0, y_min:y_max, x_min:x_max] = np.maximum(mask[0, y_min:y_max, x_min:x_max], gaussian)

    for y, x in centers_ch1:
        y_min, y_max = max(0, y - radius), min(shape[0], y + radius + 1)
        x_min, x_max = max(0, x - radius), min(shape[1], x + radius + 1)
        y_grid, x_grid = np.meshgrid(np.arange(y_min, y_max), np.arange(x_min, x_max), indexing='ij')
        gaussian = 0.7 * np.exp(-((x_grid - x) ** 2 + (y_grid - y) ** 2) / (2 * sigma ** 2))
        circle_mask = ((x_grid - x) ** 2 + (y_grid - y) ** 2) <= radius ** 2
        gaussian *= circle_mask
        mask[1, y_min:y_max, x_min:x_max] = np.maximum(mask[1, y_min:y_max, x_min:x_max], gaussian)

    return mask

mask_synth = generate_synthetic_mask()

# === Detect cells ===
detected_real_0 = detect_local_maxima(mask_real[0])
detected_real_1 = detect_local_maxima(mask_real[1])
detected_synth_0 = detect_local_maxima(mask_synth[0])
detected_synth_1 = detect_local_maxima(mask_synth[1])

# === Plot both ===
plot_detection(mask_real, detected_real_0, detected_real_1, title="Detected Cells in Real Mask (No Background)")
plot_detection(mask_synth, detected_synth_0, detected_synth_1, title="Detected Cells in Synthetic Mask (No Background)")
```

```python
import os
import glob
import numpy as np
import pandas as pd
from scipy.spatial import cKDTree
from skimage.feature import peak_local_max
from tqdm import tqdm

def detect_local_maxima(mask, min_distance=5, threshold_abs=0.1):
    """
    Detect local maxima coordinates in a 2D mask.
    Returns (y,x) list.
    """
    coords = peak_local_max(mask, min_distance=min_distance, threshold_abs=threshold_abs)
    return [tuple(coord) for coord in coords]

def evaluate_detection(detected_points, gt_points, radius):
    """
    Evaluate detection for one class.
    detected_points: list of (y,x)
    gt_points: list of (y,x)
    radius: max distance to match detected to GT

    Returns TP, FP, FN, TN counts.
    """
    if len(detected_points) == 0 and len(gt_points) == 0:
        # no detections and no ground truth
        return 0, 0, 0, 0

    detected_points = np.array(detected_points)
    gt_points = np.array(gt_points)

    if len(detected_points) == 0:
        # No detections, all GT are FN
        return 0, 0, len(gt_points), 0

    if len(gt_points) == 0:
        # No GT, all detections are FP
        return 0, len(detected_points), 0, 0

    # Build KD-tree for efficient nearest neighbor search
    gt_tree = cKDTree(gt_points)
    detected_tree = cKDTree(detected_points)

    # For each detected point, find closest GT point and distance
    distances, indices = gt_tree.query(detected_points, distance_upper_bound=radius)

    # True Positives: detected points that matched a GT within radius
    tp_mask = distances != np.inf
    TP = np.sum(tp_mask)

    # False Positives: detected points without a GT match
    FP = np.sum(~tp_mask)

    # For each GT point, find closest detected point and distance
    distances_gt, indices_gt = detected_tree.query(gt_points, distance_upper_bound=radius)

    # False Negatives: GT points with no detected match
    fn_mask = distances_gt == np.inf
    FN = np.sum(fn_mask)

    # TN is tricky to calculate exactly for detection; here we put zero or omit
    TN = 0

    return TP, FP, FN, TN

def load_gt_points_from_csv(csv_path, cls):
    """
    Load ground truth points of a given class (1 or 2) from csv file.
    Returns list of (y,x) tuples.
    """
    if os.stat(csv_path).st_size == 0:
        return []

    df = pd.read_csv(csv_path, header=None)
    df.columns = ['x', 'y', 'class']
    df_cls = df[df['class'] == cls]
    return [(row['y'], row['x']) for idx, row in df_cls.iterrows()]

def main_evaluate(mask_root, annotation_root, radius=15, min_distance=5, threshold_abs=0.1):
    subsets = ['train', 'val', 'test']

    # Accumulate totals per subset and class
    totals = {
        'train': {1: {'TP':0, 'FP':0, 'FN':0, 'TN':0}, 2: {'TP':0, 'FP':0, 'FN':0, 'TN':0}},
        'val': {1: {'TP':0, 'FP':0, 'FN':0, 'TN':0}, 2: {'TP':0, 'FP':0, 'FN':0, 'TN':0}},
        'test': {1: {'TP':0, 'FP':0, 'FN':0, 'TN':0}, 2: {'TP':0, 'FP':0, 'FN':0, 'TN':0}},
    }

    for subset in subsets:
        mask_folder = os.path.join(mask_root, subset, "cell")
        annotation_folder = os.path.join(annotation_root, subset, "cell")

        mask_paths = sorted(glob.glob(os.path.join(mask_folder, "*.npy")))

        print(f"Evaluating {subset} set with {len(mask_paths)} samples")

        for mask_path in tqdm(mask_paths):
            mask = np.load(mask_path)  # shape (2, H, W)

            filename = os.path.splitext(os.path.basename(mask_path))[0]
            csv_path = os.path.join(annotation_folder, filename + ".csv")

            # Detect local maxima for each channel
            detected_ch0 = detect_local_maxima(mask[0], min_distance=min_distance, threshold_abs=threshold_abs)
            detected_ch1 = detect_local_maxima(mask[1], min_distance=min_distance, threshold_abs=threshold_abs)

            # Load ground truth points from CSV for both classes
            gt_ch0 = load_gt_points_from_csv(csv_path, cls=1)
            gt_ch1 = load_gt_points_from_csv(csv_path, cls=2)

            # Evaluate each class
            tp0, fp0, fn0, tn0 = evaluate_detection(detected_ch0, gt_ch0, radius)
            tp1, fp1, fn1, tn1 = evaluate_detection(detected_ch1, gt_ch1, radius)

            # Accumulate totals
            totals[subset][1]['TP'] += tp0
            totals[subset][1]['FP'] += fp0
            totals[subset][1]['FN'] += fn0
            totals[subset][1]['TN'] += tn0

            totals[subset][2]['TP'] += tp1
            totals[subset][2]['FP'] += fp1
            totals[subset][2]['FN'] += fn1
            totals[subset][2]['TN'] += tn1

    # Print results per subset and class
    for subset in subsets:
        print(f"\nResults for {subset} subset:")
        for cls in [1, 2]:
            metrics = totals[subset][cls]
            TP = metrics['TP']
            FP = metrics['FP']
            FN = metrics['FN']

            # Compute precision, recall, F1
            precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
            recall = TP / (TP + FN) if (TP + FN) > 0 else 0.0
            f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

            print(f" Class {cls} (1=background, 2=tumor):")
            print(f"  TP: {TP}, FP: {metrics['FP']}, FN: {metrics['FN']}, TN: {metrics['TN']}")
            print(f"  Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")

if __name__ == "__main__":
    mask_path = "/content/drive/MyDrive/research/ocelot/dataset/GT/gaussian masks/type5"
    annotation_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations"
    main_evaluate(mask_path, annotation_path, radius=15, min_distance=1, threshold_abs=0.1)
```

```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
import albumentations as A
from albumentations.pytorch import ToTensorV2
import torch

def normalize_albumentations(image):
    # Albumentations normalization matching your pipeline
    transform = A.Compose([
        A.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
        ToTensorV2()
    ])
    transformed = transform(image=image)
    return transformed["image"]

def display_images(folder_path):
    # Get a random image from the folder (only .jpg images)
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.jpg')]
    if not image_files:
        raise ValueError("No .jpg images found in the folder!")
    random_image = random.choice(image_files)
    image_path = os.path.join(folder_path, random_image)

    # Load the image using cv2 (BGR) and convert to RGB
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply Albumentations normalization
    normalized_tensor = normalize_albumentations(image)

    # Print pixel value statistics for the normalized tensor
    print("Normalized tensor stats:")
    print("Min:", torch.min(normalized_tensor).item(), "Max:", torch.max(normalized_tensor).item())

    # Convert normalized tensor to numpy for display (shape becomes HWC)
    normalized_np = normalized_tensor.permute(1, 2, 0).numpy()

    # Denormalize for display: x_display = x * std + mean
    denormalized_np = normalized_np * 0.5 + 0.5
    denormalized_np = np.clip(denormalized_np, 0, 1)

    # Plot the images: original, raw normalized, and denormalized
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(image)
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    axes[1].imshow(normalized_np)  # Raw normalized image may appear off due to its range
    axes[1].set_title("Raw Normalized Image\n(not scaled for display)")
    axes[1].axis("off")

    axes[2].imshow(denormalized_np)
    axes[2].set_title("Denormalized Image")
    axes[2].axis("off")

    plt.show()

    # Print sample pixel values (top-left 3x3 region) for all three versions
    print("\nSample pixel values (top-left 3x3 region):")
    print("Original Image (RGB, 0-255):")
    print(image[:3, :3, :])

    print("\nRaw Normalized Image (may include negative values):")
    print(normalized_np[:3, :3, :])

    print("\nDenormalized Image (values in [0,1]):")
    print(denormalized_np[:3, :3, :])
```

```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# === CONFIGURATION ===
CSV_FOLDER = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/train/cell"
IMG_FOLDER = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images/train/cell"
RADIUS_LIST = [5, 7, 9, 11]
IMAGE_SIZE = (1024, 1024)

# Class color: BGR
CLASS_COLOR = {
    1: (0, 255, 255),   # Yellow
    2: (255, 0, 0),     # Blue
}

def generate_mask(csv_path, radius):
    mask = np.zeros((*IMAGE_SIZE, 3), dtype=np.uint8)
    inner_radius = max(0, radius - 3)
    data = pd.read_csv(csv_path, header=None, names=["x", "y", "class"])
    for _, row in data.iterrows():
        center = (int(row['x']), int(row['y']))
        color = CLASS_COLOR.get(int(row['class']), (255, 255, 255))
        cv2.circle(mask, center, radius, color, thickness=-1)
        cv2.circle(mask, center, inner_radius, (0, 0, 0), thickness=-1)
    return mask

def overlay_image(image, mask, alpha=1):
    overlay = image.copy()
    mask_nonzero = np.any(mask != [0, 0, 0], axis=-1)
    overlay[mask_nonzero] = (image[mask_nonzero] * (1 - alpha) + mask[mask_nonzero] * alpha).astype(np.uint8)
    return overlay

# === MAIN SCRIPT ===
csv_files = random.sample([f for f in os.listdir(CSV_FOLDER) if f.endswith(".csv")], 3)

for csv_file in csv_files:
    image_id = os.path.splitext(csv_file)[0]
    image_path = os.path.join(IMG_FOLDER, f"{image_id}.jpg")
    csv_path = os.path.join(CSV_FOLDER, csv_file)

    image = cv2.imread(image_path)
    image = cv2.resize(image, IMAGE_SIZE)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    fig = plt.figure(figsize=(24, 12), dpi=150)  # Larger canvas + higher DPI for zooming
    fig.suptitle(f"Instance: {csv_file}", fontsize=18)

    for col, radius in enumerate(RADIUS_LIST):
        mask = generate_mask(csv_path, radius)
        mask_rgb = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
        overlay = overlay_image(image, mask)
        overlay_rgb = cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB)

        # Row 1: mask
        plt.subplot(3, 4, col + 1)
        plt.imshow(mask_rgb)
        plt.title(f"Mask (r={radius})")
        plt.axis('off')

        # Row 2: original image
        plt.subplot(3, 4, col + 5)
        plt.imshow(image_rgb)
        plt.title("Original Image")
        plt.axis('off')

        # Row 3: overlay
        plt.subplot(3, 4, col + 9)
        plt.imshow(overlay_rgb)
        plt.title(f"Overlay (r={radius})")
        plt.axis('off')

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
import os
import random
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage import color

def visualize_cell_color_distribution(image_folder, csv_folder, crop_size=50):
    def load_random_patch():
        image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg'))]
        if not image_files:
            raise FileNotFoundError("No image files found in the folder.")
        chosen_file = random.choice(image_files)
        image_path = os.path.join(image_folder, chosen_file)
        csv_path = os.path.join(csv_folder, chosen_file.replace('.png', '.csv').replace('.jpg', '.csv'))

        if not os.path.exists(csv_path):
            print(f"Annotation CSV not found for: {chosen_file}, skipping.")
            return None, None, None

        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        ann = pd.read_csv(csv_path, header=None, names=["x", "y", "class"])
        return img, ann, chosen_file

    def crop_patch(image, center_x, center_y, size):
        half = size // 2
        x1, y1 = max(0, center_x - half), max(0, center_y - half)
        x2, y2 = x1 + size, y1 + size
        return image[y1:y2, x1:x2]

    def convert_color_spaces(patch):
        hsv = cv2.cvtColor(patch, cv2.COLOR_RGB2HSV)
        hed = color.rgb2hed(patch)
        # Normalize and scale HED to 0â€“255
        hed_min = hed.min(axis=(0, 1), keepdims=True)
        hed_max = hed.max(axis=(0, 1), keepdims=True)
        hed_scaled = (hed - hed_min) / (hed_max - hed_min + 1e-5) * 255
        hed_uint8 = np.clip(hed_scaled, 0, 255).astype(np.uint8)
        return hsv, hed_uint8

    def compute_histograms(patch, color_space="HSV"):
        patch = patch.astype(np.uint8)
        hists = []
        if color_space == "HSV":
            ranges = [(0, 180), (0, 256), (0, 256)]  # OpenCV HSV ranges
            bins = [180, 256, 256]
        else:  # Assume HED or RGB
            ranges = [(0, 256), (0, 256), (0, 256)]
            bins = [256, 256, 256]

        for i in range(3):
            hist = cv2.calcHist([patch], [i], None, [bins[i]], ranges[i])
            hists.append(hist)
        return hists


    def plot_histogram(ax, hist, mode="fixed", label=""):
        hist = hist.flatten()
        ax.set_xlim([0, len(hist)])  # Automatically adjust to 180 or 256
        ax.set_ylim([0, hist.max() * 1.1])  # Set the Y-axis limit (frequency)

        ax.set_title(f"{label} Histogram", fontsize=10)
        ax.set_xlabel("Intensity", fontsize=8)
        ax.set_ylabel("Frequency", fontsize=8)

        # Plot histogram with color coding based on channel type
        if mode == "hue":
            for i in range(1, len(hist)):
                color = plt.cm.hsv(i / len(hist))
                ax.plot([i-1, i], [hist[i-1], hist[i]], color=color)
        elif mode == "saturation":
            for i in range(1, 256):
                color = (0, i/255.0, i/255.0)  # Cyan tone increases with x
                ax.plot([i-1, i], [hist[i-1], hist[i]], color=color)
        elif mode == "value":
            for i in range(1, 256):
                color = (i/255.0, i/255.0, i/255.0)  # Brightness increases
                ax.plot([i-1, i], [hist[i-1], hist[i]], color=color)
        else:  # fixed color for H&E, etc.
            fixed_colors = {
                'Hematoxylin': 'purple',
                'Eosin': 'orange',
                'DAB': 'gray',
                'H': 'magenta',
                'S': 'cyan',
                'V': 'yellow'
            }
            color = fixed_colors.get(label, 'blue')
            ax.plot(hist, color=color, linewidth=1.5)

        # Set X and Y ticks
        tick_interval = 30 if len(hist) >= 180 else 20
        xticks = np.arange(0, len(hist), tick_interval)
        ax.set_xticks(xticks)
        ax.set_xticklabels(xticks, fontsize=8)

        ax.set_yticks(np.linspace(0, hist.max(), 6))  # Evenly spaced ticks for Y-axis
        ax.set_yticklabels(np.round(np.linspace(0, hist.max(), 6), 2), fontsize=8)

        # Enable the grid for readability
        ax.grid(True, which='both', linestyle='--', linewidth=0.5)

        # Apply tick params
        ax.tick_params(axis='both', labelsize=8)

    image, annotations, img_name = load_random_patch()
    if image is None:
        return

    print(f"Selected image: {img_name} with {len(annotations)} annotations")

    for label in [1, 2]:  # Non-tumor and Tumor
        class_cells = annotations[annotations['class'] == label]
        if len(class_cells) < 3:
            print(f"Not enough cells for class {label} in {img_name}. Found: {len(class_cells)}")
            continue

        selected_cells = class_cells.sample(n=3)

        for _, row in selected_cells.iterrows():
            cx, cy = int(row['x']), int(row['y'])
            patch = crop_patch(image, cx, cy, crop_size)
            hsv, hed = convert_color_spaces(patch)
            hsv_hists = compute_histograms(hsv)
            hed_hists = compute_histograms(hed)

            # Plot histograms and channel images for HSV
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
            ax1.imshow(hsv[:, :, 0], cmap='hsv')  # Display Hue channel
            ax1.set_title(f"HSV Hue Channel for Cell at ({cx}, {cy})", fontsize=14)
            ax1.axis('off')

            plot_histogram(ax2, hsv_hists[0], mode="hue", label="H (Hue)")
            plt.tight_layout()
            plt.show()

            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
            ax1.imshow(hsv[:, :, 1], cmap='Blues')  # Display Saturation channel
            ax1.set_title(f"HSV Saturation Channel for Cell at ({cx}, {cy})", fontsize=14)
            ax1.axis('off')

            plot_histogram(ax2, hsv_hists[1], mode="saturation", label="S (Saturation)")
            plt.tight_layout()
            plt.show()

            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
            ax1.imshow(hsv[:, :, 2], cmap='gray')  # Display Value channel
            ax1.set_title(f"HSV Value Channel for Cell at ({cx}, {cy})", fontsize=14)
            ax1.axis('off')

            plot_histogram(ax2, hsv_hists[2], mode="value", label="V (Value)")
            plt.tight_layout()
            plt.show()

            # Plot histograms and channel images for HED
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
            ax1.imshow(hed[:, :, 0], cmap='Purples')  # Display Hematoxylin channel
            ax1.set_title(f"HED Hematoxylin Channel for Cell at ({cx}, {cy})", fontsize=14)
            ax1.axis('off')

            plot_histogram(ax2, hed_hists[0], mode="fixed", label="Hematoxylin")
            plt.tight_layout()
            plt.show()

            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
            ax1.imshow(hed[:, :, 1], cmap='Oranges')  # Display Eosin channel
            ax1.set_title(f"HED Eosin Channel for Cell at ({cx}, {cy})", fontsize=14)
            ax1.axis('off')

            plot_histogram(ax2, hed_hists[1], mode="fixed", label="Eosin")
            plt.tight_layout()
            plt.show()

            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
            ax1.imshow(hed[:, :, 2], cmap='Greys')  # Display DAB channel
            ax1.set_title(f"HED DAB Channel for Cell at ({cx}, {cy})", fontsize=14)
            ax1.axis('off')

            plot_histogram(ax2, hed_hists[2], mode="fixed", label="DAB")
            plt.tight_layout()
            plt.show()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from glob import glob

def detect_foreign_colors(image, pixel_thresh=0.001):
    """
    Detects colors in image that are NOT: red, blue, purple, pink, white, black.
    Flags patch if foreign-colored pixels exceed threshold.
    """
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    h, s, v = cv2.split(hsv)

    h = h.astype(np.uint8)
    s = s.astype(np.uint8)
    v = v.astype(np.uint8)

    # Accepted color masks
    red1 = ((h < 10) | (h > 170)) & (s > 50) & (v > 50)          # Red tones
    pink_purple = ((h > 290//2) & (h < 330//2)) & (s > 50)       # Approx magenta-pink (145â€“165)
    blue = ((h > 200//2) & (h < 260//2)) & (s > 50)              # Blue hues
    white = (s < 30) & (v > 180)                                 # Low saturation, high brightness
    black = (v < 50)                                             # Low brightness

    # Combine masks
    accepted = red1 | pink_purple | blue | white | black

    # Anything not in accepted range is considered foreign
    foreign_mask = ~accepted

    foreign_ratio = np.sum(foreign_mask) / foreign_mask.size
    return foreign_ratio > pixel_thresh, foreign_mask



def scan_folder_for_artifacts(folder_path, pixel_thresh=0.001, show=True, save_csv=True):
    image_paths = sorted(glob(os.path.join(folder_path, "*")))
    flagged = []

    for path in image_paths:
        # Read image safely
        bgr_img = cv2.imread(path)
        if bgr_img is None:
            print(f"âš ï¸ Skipped unreadable or non-image file: {os.path.basename(path)}")
            continue

        img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
        flag, mask = detect_foreign_colors(img, pixel_thresh)

        if flag:
            flagged.append(path)

            if show:
                fig, axs = plt.subplots(1, 2, figsize=(10, 5))
                axs[0].imshow(img)
                axs[0].set_title("Original")
                axs[1].imshow(mask, cmap='gray')
                axs[1].set_title("Foreign Color Mask")
                for ax in axs:
                    ax.axis('off')
                plt.suptitle(os.path.basename(path))
                plt.show()

    if save_csv:
        pd.DataFrame(flagged, columns=["filename"]).to_csv("flagged_foreign_color_images.csv", index=False)
        print(f"\nâœ… Saved {len(flagged)} flagged images to CSV.")

    return flagged

```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
import pandas as pd
from skimage.color import separate_stains, hed_from_rgb
from skimage.feature.texture import graycomatrix, graycoprops
from tqdm import tqdm
import time

# === CONFIG ===
image_folder = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images/train/cell"
csv_folder = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/train/cell"
GRAY_LEVELS = 128
DISTANCE = 1
ANGLES = [0, np.pi/4, np.pi/2, 3*np.pi/4]
WINDOW_SIZE = 15
STRIDE = 1

# All Haralick features available in scikit-image
HARALICK_FEATURES = [
    'contrast',
    'dissimilarity',
    'homogeneity',
    'energy',
    'correlation',
    'entropy'  # Adding entropy
]

# === UTILS ===
def quantize(img, levels):
    img = np.clip(img, 0, 1)
    return (img * (levels - 1)).astype(np.uint8)

def compute_single_haralick(img_window, distances, angles, levels, feature):
    glcm = graycomatrix(img_window, distances=distances, angles=angles, levels=levels, symmetric=True, normed=True)

    # Custom implementation for entropy (not directly available in graycoprops)
    if feature == 'entropy':
        # Calculate entropy from GLCM
        # Replace zero values to avoid log(0)
        eps = 1e-10
        glcm_prob = glcm + eps
        entropy = -np.sum(glcm_prob * np.log(glcm_prob), axis=(0, 1))
        return np.mean(entropy)
    else:
        feat_vals = graycoprops(glcm, feature)
        return np.mean(feat_vals)

def sliding_window_feature_map(image, window_size, stride, levels, feature):
    h, w = image.shape
    half = window_size // 2
    out_h = (h - window_size) // stride + 1
    out_w = (w - window_size) // stride + 1
    feature_map = np.zeros((out_h, out_w))

    for i in range(out_h):
        for j in range(out_w):
            y = i * stride
            x = j * stride
            window = image[y:y+window_size, x:x+window_size]
            feature_map[i, j] = compute_single_haralick(window, [DISTANCE], ANGLES, levels, feature)

    return feature_map

def compute_all_features(image, window_size, stride, levels, features_list):
    feature_maps = {}
    for feature in features_list:
        print(f"Computing {feature} feature map...")
        feature_maps[feature] = sliding_window_feature_map(image, window_size, stride, levels, feature)
    return feature_maps

# === MAIN ===
# Select an image
img_list = [f for f in os.listdir(image_folder) if f.lower().endswith(('.png', '.jpg', '.tif'))]
selected_img = np.random.choice(img_list)
base_name = os.path.splitext(selected_img)[0]
img_path = os.path.join(image_folder, selected_img)
csv_path = os.path.join(csv_folder, f"{base_name}.csv")
print(f"Selected image: {selected_img}")

# Load image and cell annotations
image_rgb = io.imread(img_path)
coords = pd.read_csv(csv_path, header=None).values  # x, y, class
tumor_cells = coords[coords[:, 2] == 2]
if len(tumor_cells) == 0:
    raise ValueError(f"No tumor cells found in: {csv_path}")
x, y, _ = tumor_cells[np.random.choice(len(tumor_cells))]
x, y = int(x), int(y)

# Convert to grayscale, H, E
image_gray = color.rgb2gray(image_rgb)
image_hed = separate_stains(image_rgb, hed_from_rgb)
h_channel = np.clip(np.exp(-image_hed[:, :, 0]), 0, 1)
e_channel = np.clip(np.exp(-image_hed[:, :, 1]), 0, 1)

# Extract patch around the tumor cell
half_crop = 50  # Bigger crop to allow feature map computation
gray_crop = image_gray[y - half_crop:y + half_crop, x - half_crop:x + half_crop]
h_crop = h_channel[y - half_crop:y + half_crop, x - half_crop:x + half_crop]
e_crop = e_channel[y - half_crop:y + half_crop, x - half_crop:x + half_crop]

# Quantize all channels
q_gray = quantize(gray_crop, GRAY_LEVELS)
q_h = quantize(h_crop, GRAY_LEVELS)
q_e = quantize(e_crop, GRAY_LEVELS)

# Create output directory for saving results
timestamp = time.strftime("%Y%m%d_%H%M%S")
output_dir = f"haralick_features_{timestamp}"
os.makedirs(output_dir, exist_ok=True)

# Create a figure showing the original crops
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(gray_crop, cmap='gray')
plt.title("Grayscale Crop")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(h_crop, cmap='gray')
plt.title("H Channel Crop")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(e_crop, cmap='gray')
plt.title("E Channel Crop")
plt.axis('off')

plt.suptitle("Original Image Channels")
plt.tight_layout()
plt.savefig(f"{output_dir}/original_crops.png", dpi=300, bbox_inches='tight')
plt.show()

# Compute all feature maps for each channel
print("Computing feature maps for grayscale channel...")
gray_maps = compute_all_features(q_gray, WINDOW_SIZE, STRIDE, GRAY_LEVELS, HARALICK_FEATURES)

print("Computing feature maps for H channel...")
h_maps = compute_all_features(q_h, WINDOW_SIZE, STRIDE, GRAY_LEVELS, HARALICK_FEATURES)

print("Computing feature maps for E channel...")
e_maps = compute_all_features(q_e, WINDOW_SIZE, STRIDE, GRAY_LEVELS, HARALICK_FEATURES)

# Save feature maps statistics to a file
with open(f"{output_dir}/feature_statistics.txt", 'w') as f:
    for feature in HARALICK_FEATURES:
        f.write(f"\n{feature.upper()} Statistics:\n")
        f.write(f"Grayscale: min={gray_maps[feature].min():.4f}, max={gray_maps[feature].max():.4f}, mean={gray_maps[feature].mean():.4f}\n")
        f.write(f"H Channel: min={h_maps[feature].min():.4f}, max={h_maps[feature].max():.4f}, mean={h_maps[feature].mean():.4f}\n")
        f.write(f"E Channel: min={e_maps[feature].min():.4f}, max={e_maps[feature].max():.4f}, mean={e_maps[feature].mean():.4f}\n")

# Create a single figure with all features
# One row per feature, three columns (grayscale, H, E)
num_features = len(HARALICK_FEATURES)
plt.figure(figsize=(15, 4 * num_features))

for i, feature in enumerate(HARALICK_FEATURES):
    # Find common color scale for this feature across channels
    min_val = min(gray_maps[feature].min(), h_maps[feature].min(), e_maps[feature].min())
    max_val = max(gray_maps[feature].max(), h_maps[feature].max(), e_maps[feature].max())

    # Grayscale
    plt.subplot(num_features, 3, i*3 + 1)
    plt.imshow(gray_maps[feature], cmap='hot', vmin=min_val, vmax=max_val)
    plt.title(f"Grayscale: {feature}")
    plt.axis('off')
    plt.colorbar(shrink=0.8)

    # H Channel
    plt.subplot(num_features, 3, i*3 + 2)
    plt.imshow(h_maps[feature], cmap='hot', vmin=min_val, vmax=max_val)
    plt.title(f"H Channel: {feature}")
    plt.axis('off')
    plt.colorbar(shrink=0.8)

    # E Channel
    plt.subplot(num_features, 3, i*3 + 3)
    plt.imshow(e_maps[feature], cmap='hot', vmin=min_val, vmax=max_val)
    plt.title(f"E Channel: {feature}")
    plt.axis('off')
    plt.colorbar(shrink=0.8)

plt.suptitle(f"Haralick Features Comparison - Window: {WINDOW_SIZE}Ã—{WINDOW_SIZE}, Stride: {STRIDE}")
plt.tight_layout()
plt.subplots_adjust(top=0.95)
plt.savefig(f"{output_dir}/all_features_comparison.png", dpi=300, bbox_inches='tight')
plt.show()

# Create a figure showing individual features in a grid
for feature in HARALICK_FEATURES:
    # Save individual feature maps as separate files
    plt.figure(figsize=(15, 5))

    # Find common color scale for this feature across channels
    min_val = min(gray_maps[feature].min(), h_maps[feature].min(), e_maps[feature].min())
    max_val = max(gray_maps[feature].max(), h_maps[feature].max(), e_maps[feature].max())

    plt.subplot(1, 3, 1)
    plt.imshow(gray_maps[feature], cmap='hot', vmin=min_val, vmax=max_val)
    plt.title(f"Grayscale: {feature}")
    plt.axis('off')
    plt.colorbar(shrink=0.8)

    plt.subplot(1, 3, 2)
    plt.imshow(h_maps[feature], cmap='hot', vmin=min_val, vmax=max_val)
    plt.title(f"H Channel: {feature}")
    plt.axis('off')
    plt.colorbar(shrink=0.8)

    plt.subplot(1, 3, 3)
    plt.imshow(e_maps[feature], cmap='hot', vmin=min_val, vmax=max_val)
    plt.title(f"E Channel: {feature}")
    plt.axis('off')
    plt.colorbar(shrink=0.8)

    plt.suptitle(f"Haralick Feature: {feature}")
    plt.tight_layout()
    plt.savefig(f"{output_dir}/{feature}_comparison.png", dpi=300, bbox_inches='tight')
    plt.show()

    # Save the feature maps as numpy arrays
    np.save(f"{output_dir}/{feature}_gray.npy", gray_maps[feature])
    np.save(f"{output_dir}/{feature}_h.npy", h_maps[feature])
    np.save(f"{output_dir}/{feature}_e.npy", e_maps[feature])

print(f"\nAll feature maps saved to directory: {output_dir}")
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage.color import rgb2gray
from skimage.util import img_as_ubyte
from tqdm import tqdm

def get_entropy_mask(image, threshold=2.5):
    gray = img_as_ubyte(rgb2gray(image))
    ent = entropy(gray, disk(5))
    return ent > threshold

def blur_on_mask(image, mask, lap_thresh=100.0):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    masked = gray.copy()
    masked[~mask] = 0
    lap_var = cv2.Laplacian(masked, cv2.CV_64F).var()
    return lap_var < lap_thresh, lap_var

def scan_blur_in_folder(folder_path, entropy_thresh=2.5, lap_thresh=100.0, show=False):
    image_paths = [os.path.join(folder_path, fname)
                   for fname in os.listdir(folder_path)
                   if fname.lower().endswith(('.png', '.jpg', '.jpeg', '.tif'))]

    flagged = []

    for path in tqdm(image_paths):
        img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
        mask = get_entropy_mask(img, threshold=entropy_thresh)
        is_blur, score = blur_on_mask(img, mask, lap_thresh=lap_thresh)

        if is_blur:
            flagged.append((path, score))

            if show:
                plt.figure(figsize=(10, 4))
                plt.subplot(1, 3, 1)
                plt.imshow(img)
                plt.title("Image")
                plt.axis('off')

                plt.subplot(1, 3, 2)
                plt.imshow(mask, cmap='gray')
                plt.title("Entropy Mask")

                plt.subplot(1, 3, 3)
                plt.imshow(cv2.Laplacian(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), cv2.CV_64F), cmap='gray')
                plt.title(f"Laplacian Map\nVar: {score:.2f}")

                plt.tight_layout()
                plt.show()

    print(f"\nFlagged {len(flagged)} blurred images out of {len(image_paths)} total.")
    return flagged
```

```python
import cv2
import matplotlib.pyplot as plt
import os
import random

def denoise_image(input_image_path, h=10, templateWindowSize=7, searchWindowSize=21):
    # Read the image
    img = cv2.imread(input_image_path)

    # Check if image is loaded properly
    if img is None:
        print(f"Error: Unable to load image from path: {input_image_path}")
        return

    # Convert image to RGB (OpenCV reads it in BGR by default)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Apply Non-Local Means Denoising
    denoised_img = cv2.fastNlMeansDenoisingColored(img, None, h, templateWindowSize, searchWindowSize)

    # Convert denoised image to RGB for display
    denoised_rgb = cv2.cvtColor(denoised_img, cv2.COLOR_BGR2RGB)

    # Plot original and denoised side by side
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(denoised_rgb)
    plt.title("Denoised Image")
    plt.axis('off')

    plt.show()

def process_random_images(image_dir, num_images=5):
    # List all image files in the directory
    image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.jpeg', '.png', '.tiff'))]

    # Select random images from the list
    random_images = random.sample(image_files, num_images)

    # Process each selected image
    for img_file in random_images:
        img_path = os.path.join(image_dir, img_file)
        print(f"Processing image: {img_path}")
        denoise_image(img_path)

# Example usage
image_dir = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images/train/cell"
process_random_images(image_dir, num_images=5)
```

```python
import os
import json
import pandas as pd

# Paths
metadata_path = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/metadata.json"
annotation_root = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations"

# Load metadata
with open(metadata_path, "r") as f:
    metadata = json.load(f)

sample_pairs = metadata["sample_pairs"]

# Dictionary to store counts {organ: {subset: {class_1: count, class_2: count}}}
organ_counts = {}

# Iterate through subsets
for subset in ["train", "val", "test"]:
    annotation_dir = os.path.join(annotation_root, subset, "cell")

    for filename in os.listdir(annotation_dir):
        if filename.endswith(".csv"):
            file_id = filename.replace(".csv", "")

            if file_id in sample_pairs:
                organ = sample_pairs[file_id]["organ"]

                if organ not in organ_counts:
                    organ_counts[organ] = {s: {1: 0, 2: 0} for s in ["train", "val", "test"]}

                file_path = os.path.join(annotation_dir, filename)
                df = pd.read_csv(file_path, names=["x", "y", "class"])

                class_counts = df["class"].value_counts().to_dict()

                organ_counts[organ][subset][1] += class_counts.get(1, 0)
                organ_counts[organ][subset][2] += class_counts.get(2, 0)

# Compute total cells per subset (to use for % calculation)
subset_totals = {
    s: sum(organ_counts[org][s][1] + organ_counts[org][s][2] for org in organ_counts)
    for s in ["train", "val", "test"]
}

# Prepare DataFrame
summary_data = []
for organ, subsets in organ_counts.items():
    row = {"Organ": organ}
    for subset in ["train", "val", "test"]:
        c1 = subsets[subset][1]
        c2 = subsets[subset][2]
        total = c1 + c2
        percent1 = 100 * c1 / subset_totals[subset] if subset_totals[subset] > 0 else 0
        percent2 = 100 * c2 / subset_totals[subset] if subset_totals[subset] > 0 else 0
        row[subset] = f"{c1} | {c2} ({percent1:.1f}% | {percent2:.1f}%)"
    summary_data.append(row)

# Create summary DataFrame
summary_df = pd.DataFrame(summary_data)

# Add total row
total_counts = {"Organ": "Total"}
for subset in ["train", "val", "test"]:
    c1 = sum(organ_counts[org][subset][1] for org in organ_counts)
    c2 = sum(organ_counts[org][subset][2] for org in organ_counts)
    total = c1 + c2
    percent1 = 100 * c1 / total if total > 0 else 0
    percent2 = 100 * c2 / total if total > 0 else 0
    total_counts[subset] = f"{c1} | {c2} ({percent1:.1f}% | {percent2:.1f}%)"

summary_df = pd.concat([summary_df, pd.DataFrame([total_counts])], ignore_index=True)

# Display table
summary_df

# Save to CSV
output_path = "/content/cell_class_distribution_by_organ.csv"
summary_df.to_csv(output_path, index=False)
print(f"CSV saved to: {output_path}")

```

```python
import os
import numpy as np
import pandas as pd
import cv2

# Parameters
IMG_SIZE = 1024
RADIUS = 15  # Fixed circle radius
IMAGE_EXTENSIONS = {".png", ".jpg", ".jpeg"}

# Paths
#BASE_DIR = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1"
#OUTPUT_DIR = os.path.join(BASE_DIR, "dot masks")
os.makedirs(OUTPUT_DIR, exist_ok=True)

def generate_circle_mask(csv_path, img_size=IMG_SIZE, radius=RADIUS):
    """Generate one-hot encoded circle mask for a given CSV annotation."""
    df = pd.read_csv(csv_path, header=None, names=["x", "y", "class"])
    if df.empty:
        print(f"Skipping {csv_path}: Empty file.")
        return None

    # Initialize a zero-valued float32 mask
    mask = np.zeros((img_size, img_size, 2), dtype=np.float32)

    for _, row in df.iterrows():
        x, y, cls = int(row["x"]), int(row["y"]), int(row["class"])
        channel = 0 if cls == 1 else 1  # Map Class 1 to Channel 0, Class 2 to Channel 1

        # Extract single-channel mask and ensure it's correctly indexed
        single_channel_mask = mask[:, :, channel].copy()

        # Draw filled circle (value = 1) on the extracted single-channel mask
        cv2.circle(single_channel_mask, (x, y), radius, 1, -1)

        # Reassign modified channel back to the main mask
        mask[:, :, channel] = single_channel_mask

    return mask

# Iterate through train, val, and test splits
for split in ["train", "val", "test"]:
    csv_dir = os.path.join(BASE_DIR, "annotations", split, "cell")
    mask_output_dir = os.path.join(OUTPUT_DIR, split, "cell")
    os.makedirs(mask_output_dir, exist_ok=True)

    csv_files = [f for f in os.listdir(csv_dir) if f.endswith(".csv")]

    for csv_file in csv_files:
        csv_path = os.path.join(csv_dir, csv_file)
        img_base = os.path.splitext(csv_file)[0]
        mask_path = os.path.join(mask_output_dir, img_base + ".npy")

        mask = generate_circle_mask(csv_path)
        if mask is not None:
            np.save(mask_path, mask)
            print(f"âœ… Saved mask: {mask_path}")
        else:
            print(f"âš ï¸ Skipping mask for {csv_file}")
```

## gaussian mask

```python
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import random

# Parameters
IMG_SIZE = 1024
IMAGE_EXTENSIONS = {".png", ".jpg", ".jpeg"}
RADIUS_VALUES = [1, 5, 15, 60]  # Different radius values to compare
GAUSSIAN_STDDEV = 12

# Paths
CSV_DIR = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/train/cell"
IMG_DIR = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images/train/cell"

# List available CSV files
csv_files = [f for f in os.listdir(CSV_DIR) if f.endswith(".csv")]
if not csv_files:
    raise FileNotFoundError("No CSV annotation files found in the specified directory!")

# Randomly select a CSV file
csv_file = random.choice(csv_files)
csv_path = os.path.join(CSV_DIR, csv_file)

# Find corresponding image file (same name, different extension)
img_base = os.path.splitext(csv_file)[0]
img_file = None
for ext in IMAGE_EXTENSIONS:
    potential_img_path = os.path.join(IMG_DIR, img_base + ext)
    if os.path.exists(potential_img_path):
        img_file = potential_img_path
        break

if not img_file:
    print(f"âš ï¸ No matching image found for {csv_file}. Skipping...")
else:
    print(f"Processing: {csv_file} with image {os.path.basename(img_file)}")

    # Load cell image
    cell_img = cv2.imread(img_file)
    cell_img = cv2.cvtColor(cell_img, cv2.COLOR_BGR2RGB)

    # Load CSV annotations
    try:
        df = pd.read_csv(csv_path, header=None, names=["x", "y", "class"])
        if df.empty:
            print(f"âŒ {csv_file} is empty. Skipping...")
            exit()

        masks = {}

        for radius in RADIUS_VALUES:
            mask = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.float32)  # Use 3 channels for color

            for _, row in df.iterrows():
                x, y, cls = int(row["x"]), int(row["y"]), int(row["class"])
                channel = 0 if cls == 1 else 2  # Class 1 (Blue), Class 2 (Red)

                for dx in range(-radius, radius + 1):
                    for dy in range(-radius, radius + 1):
                        nx, ny = x + dx, y + dy
                        if 0 <= nx < IMG_SIZE and 0 <= ny < IMG_SIZE:
                            dist_sq = dx**2 + dy**2
                            if dist_sq <= radius**2:
                                weight = np.exp(-dist_sq / (2 * GAUSSIAN_STDDEV**2))
                                mask[ny, nx, channel] += weight

            # Normalize mask to 0-255 for visualization
            mask = (mask / mask.max() * 255).astype(np.uint8)
            masks[radius] = mask

        # Display results
        fig, axes = plt.subplots(1, 5, figsize=(20, 5))
        axes[0].imshow(cell_img)
        axes[0].set_title("Original Cell Image")
        axes[0].axis("off")

        for i, radius in enumerate(RADIUS_VALUES):
            axes[i + 1].imshow(masks[radius])
            axes[i + 1].set_title(f"Gaussian Mask with Radius {radius}")
            axes[i + 1].axis("off")

        plt.show()

    except Exception as e:
        print(f"âŒ Error processing {csv_file}: {str(e)}. Continuing...")
```

```python
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import random

# Parameters
IMG_SIZE = 1024
IMAGE_EXTENSIONS = {".png", ".jpg", ".jpeg"}
RADIUS = 15  # Fixed radius for all masks
GAUSSIAN_STDDEVS = [3, 6, 12]  # Different standard deviations to compare

# Paths
CSV_DIR = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/annotations/train/cell"
IMG_DIR = "/content/drive/MyDrive/research/ocelot/dataset/ocelot2023_v1.0.1/images/train/cell"

# List available CSV files
csv_files = [f for f in os.listdir(CSV_DIR) if f.endswith(".csv")]
if not csv_files:
    raise FileNotFoundError("No CSV annotation files found in the specified directory!")

# Randomly select a CSV file
csv_file = random.choice(csv_files)
csv_path = os.path.join(CSV_DIR, csv_file)

# Find corresponding image file (same name, different extension)
img_base = os.path.splitext(csv_file)[0]
img_file = None
for ext in IMAGE_EXTENSIONS:
    potential_img_path = os.path.join(IMG_DIR, img_base + ext)
    if os.path.exists(potential_img_path):
        img_file = potential_img_path
        break

if not img_file:
    print(f"âš ï¸ No matching image found for {csv_file}. Skipping...")
else:
    print(f"Processing: {csv_file} with image {os.path.basename(img_file)}")

    # Load cell image
    cell_img = cv2.imread(img_file)
    cell_img = cv2.cvtColor(cell_img, cv2.COLOR_BGR2RGB)

    # Load CSV annotations
    try:
        df = pd.read_csv(csv_path, header=None, names=["x", "y", "class"])
        if df.empty:
            print(f"âŒ {csv_file} is empty. Skipping...")
            exit()

        masks = {}

        for stddev in GAUSSIAN_STDDEVS:
            mask = np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.float32)  # Use 3 channels for color

            for _, row in df.iterrows():
                x, y, cls = int(row["x"]), int(row["y"]), int(row["class"])
                channel = 0 if cls == 1 else 2  # Class 1 (Blue), Class 2 (Red)

                for dx in range(-RADIUS, RADIUS + 1):
                    for dy in range(-RADIUS, RADIUS + 1):
                        nx, ny = x + dx, y + dy
                        if 0 <= nx < IMG_SIZE and 0 <= ny < IMG_SIZE:
                            dist_sq = dx**2 + dy**2
                            if dist_sq <= RADIUS**2:
                                weight = np.exp(-dist_sq / (2 * stddev**2))
                                mask[ny, nx, channel] += weight

            # Normalize mask to 0-255 for visualization, and adjust it to have the same number of pixels as before
            mask = (mask / mask.max() * 255).astype(np.uint8)
            masks[stddev] = mask

        # Display results
        fig, axes = plt.subplots(1, len(GAUSSIAN_STDDEVS) + 1, figsize=(20, 5))
        axes[0].imshow(cell_img)
        axes[0].set_title("Original Cell Image")
        axes[0].axis("off")

        for i, stddev in enumerate(GAUSSIAN_STDDEVS):
            axes[i + 1].imshow(masks[stddev])
            axes[i + 1].set_title(f"Gaussian Mask with StdDev {stddev}")
            axes[i + 1].axis("off")

        plt.show()

    except Exception as e:
        print(f"âŒ Error processing {csv_file}: {str(e)}. Continuing...")
```
